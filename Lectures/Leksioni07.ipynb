# **Transistorët**

## **Çfarë është një transistor?**

Një transistor është një pajisje elektronike gjysmëpërçuese e cila sillet ose si një çelës i komanduar elektrikisht ose si një amplifikator kontrolluar elektrikisht. Në thelb të funksionit të transistorit është kontrolli i rrjedhjes së rrymës, analoge me kontrollin e rrjedhjes së ujit nga një rubinet. Ndryshe nga pajisjet pasive që kemi parë deri tani që kanë patur vetëm dy terminale, transistorët kanë edhe një terminal shtesë për kontrollin, pra janë pajisje me tre terminale. Një sinjal i vogël tensioni apo rryme që aplikohet mbi terminalin e kontrollit, mundësohet një ndryshim të madh rryme mbi dy terminalet e tjera të tij. Për shkak të numrit të madh të aplikimeve që ato kanë, transistorët gjënden praktikisht në çdo qark elektrik të pajisjeve moderne.


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përkufizimi**

**Një transistor është një pajisje elektronike gjysmëpërçuese me tre terminale që kontrollon rrjedhën e rrymës në një qark, nëpërmjet një sinjali kontrollues të jashtëm (rrymë ose tension).**

- Nëse kontrolli bëhet me rrymë (injeksion bartësish) ➔ kemi transistor bipolar (Bipolar Junction Transistor, BJT shkurt).
- Nëse kontrolli bëhet me tension (kontroll elektrostatik të kanaleve) ➔ kemi transistor me efekt fushe (Field Effect Transistor, FET shkurt).

</div>

- Transistori kontrollon rrymën me një sinjal të vogël elekrik (rrymë apo tension) dhe e **amplifikon** efektin e këtij sinjali në një sinjali shumë më të madh në dalje. Pra **transistori është një amplifikator**.
- Për sinjale mjaftueshmërisht të mëdha, **transistori kalon në regjim ngopje** dhe **kthehet efektivisht në një çelës elektrik të komandueshëm**.
- Ai është **zemra** e elektronikës moderne: mikroprocesorët, sensorët, komunikimi digjital, etj — të gjitha varen nga transistorët.


<br>

👉 **Në thelb**, një transistor është një **valvul elektrike**: përdor një **sinjal të vogël** për të kontrolluar **një sinjal më të madh**.

## Përmbledhje historike e transistorit

- **1925:** Lilienfeld imagjinoi një pajisje qe mund ti komandohej rezistenca elektrike nëpërmjet fushës elektrike, por materialet nuk ishin mjaftueshëm të pastra.
- Në vitet 1940, teknologjia bazohej në **tubat vakum** për përforcim dhe komutacion — këto ishin të mëdha, konsumonin shumë energji dhe ishin të paqëndrueshme. Kërkohej **një zëvendësues me përmasa të vogla, i besueshëm dhe energjitikisht efikas**.
- **1947:** Shpikja e parë praktike: BJT (Bardeen, Brattain, Shockley në Bell Labs).
- **1959:** Pasi u zgjidhën problemet me oksidet (SiO₂), u bë i mundur ndertimi i MOSFET.
- Sot: MOSFET dominon elektronikën digjitale dhe IC-të (integrated circuits), BJT përdoret kryesisht në përforcues ose qarkë analogë me rrymë të madhe.

**Ironi historike**:  
Pavarësisht se transistori BJT (1947) u zbulu përpara FET (1959), në fakt, ideja e parë ishte për një **Field Effect Transistor (FET)** (1925) — por atë kohë, nuk ishte e mundur për shkak të papërsosmërisë së materialeve.

## **Si funksionon një BJT?**

Këtu po fokusohemi përgjithësisht te transistori npn edhe për shkak se është edhe më i përdorur në praktikë. Megjithatë, në rastet kur nuk përmëndet në mënyrë eksplicite, analiza për një transistor pnp është identike vetëm se polaritetet e tensioneve të terminaleve dhe drejtimet e rrymave janë të kunderta me një npn.

### Struktura:

Transistori npn ka
- **Emiteri (E)**: Gjysëmpërcjellës i tipit n (shumë i dopuar)
- **Baza (B)**: Gjysëmpërcjellës i tipit p (shumë e hollë dhe lehtë e dopuar)
- **Kolektori (C)**: Gjysëmpërcjellës i tipit n (më pak i dopuar)

<img src="figs/lec7/BJT_pnp_npn.svg" alt="bjt" width="50%" style="background-color:#fff;" > <img src="figs/lec7/npn_Cross-section.svg" alt="npn cross section" width="30%" style="background-color:#fff;" >


### Operimi:

- **Lidhje e drejtë (Forward bias)** në Emiter-Bazë → injektim i elektroneve në bazë.
- **Lidhje e kundërt (Reverse bias)** në Bazë-Kolektor → fushë që rrallon elektronet në kolektor.
- **Rezultati**: një **rrymë shumë e vogël në base** kontrollon një **rrymë të madhe emitter-collector**.

* Transistori **npn** përbëhet nga një bazë $p$ shumë e hollë që është sandwic me dy zonave $n$.
* **Pa aplikuar tension në bazë**, bashkimi p-n ndalon kalimin e elektroneve → **nuk ka rrymë**.
* **Tension negativ në bazë** e përkeqëson situatën: polarizim i kundërt → rrit **zonë e varfëruar**.
* **Tension pozitiv ≥ 0.6 V** në bazë → polarizim drejt:
  * Elektronet kalojnë nga emiteri në bazë.
  * Baza është shumë e hollë, kështu që shumica e elektroneve **kalojnë direkt në kolektor**.
* Rritja e tensionit në bazë **rrit rrymën kolektor-emiter**.
* Nga pikëpamja konvencionale: **rryma hyn në bazë dhe del nga kolektori drejt emiterit**.

### **Regjimet e Funksionimit të Transistorit Bipolar (BJT)**

Transistori bipolar (BJT) ka katër regjime kryesore funksionimi, të përcaktuara nga polarizimi i bashkimeve p-n midis terminaleve të tij:

1. **Regjimi Aktiv (Forward-Active Mode):**
   * **Polarizimi:** Bashkimi bazë-emiter është i polarizuar drejt, ndërsa bashkimi bazë-kolektor është i polarizuar në kundërt.
   * **Funksioni:** Ky është regjimi ku transistori vepron si përforcues. Një rrymë e vogël në bazë ($I_B$) kontrollon një rrymë më të madhe në kolektor ($I_C$), me një raport përafërsisht të barabartë me fitimin e rrymës $\beta$.


2. **Regjimi i Saturimit (Saturation Mode):**
   * **Polarizimi:** Të dy bashkimet, bazë-emiter dhe bazë-kolektor, janë të polarizuara drejt.
   * **Funksioni:** Transistori sillet si një çelës i mbyllur, duke lejuar rrjedhjen e lirë të rrymës nga emiteri në kolektor. Ky regjimi përdoret zakonisht në aplikime digjitale për të përfaqësuar gjendjen "ON".


3. **Regjimi i Prerjes (Cut-off Mode):**
   * **Polarizimi:** Të dy bashkimet janë të polarizuara në kundërt.
   * **Funksioni:** Transistori sillet si një çelës i hapur, duke bllokuar rrjedhjen e rrymës. Ky regjim përdoret për të përfaqësuar gjendjen "OFF" në qarkun digjital.


4. **Regjimi Aktiv i Kundërt (Reverse-Active Mode):**
   * **Polarizimi:** Bashkimi bazë-emetues është i polarizuar në kundërt, ndërsa bashkimi bazë-kolektor është i polarizuar drejt.
   * **Funksioni:** Në këtë regjim, rolet e emiterit dhe kolektorit janë të ndërruara. Megjithatë, ky regjimi përdoret rrallë, pasi performanca e transistorit është shumë më e ulët në këtë mënyrë për shkak të dizajnit të optimizuar për regjimin aktiv të drejtë.


Këto regjime janë thelbësore për të kuptuar se si një transistor bipolar mund të përdoret si përforcues në qarkun analog apo si çelës në qarkun digjital.

### Parametrat $\alpha$ dhe $\beta$ per rritjen e rrymes (current gain)

Në gjëndje stabël, vlen:
$$ I_E = I_B + I_C$$

dhe $ I_B\ll I_C$. Transistori BJT ka këto parametra që janë konstant,
$$
\alpha =\frac{I_C}{I_E}; \;\; \beta =\frac{I_C}{I_B}; \;\Longrightarrow \alpha = \frac{\beta}{1+\beta}
$$

Ku $\alpha\approx1$ (merr vlera midis 0.98 dhe 0.998) është rritja në rrymë (current gain) për lidhjnet me bazë të përbashkët. Në këtë lidhje, rritja në rrymë është nga emiteri në kolektor. Ndërsa $\beta\gg 1$ (e rendit 100) është një numër i rëndësishëm i performancës së BJT, dhe tregon amplifikimin e rrymës së bazës psh në lidhjen me emiter të përbashkët. Parametri $\beta$ siguron që rryma e bazës ështv e neglizhueshme në krahasim me atë të emiterit dhe kolektorit. Pavaresisht se këto parametra luajnë një rol shumë të rëndësishëm në analizën dhe dizejnimin e qarqeve me transistorë, ato nuk janë parametra fundamentale që psh të rrjedhin nga fizika e gjysëmpërcuesve.

### LLojet e lidhjeve te BJT


| Lidhja                      | Sjellja                                          |
| --------------------------- | ------------------------------------------------ |
| **Emiter të Përbashkët**    | Amplifikator i mirë tensioni me invertim.        |
| **Bazë të Përbashkët**      | Amplifikator për frekuencat e larta pa invertim. |
| **Kolektor të Përbashkët**  | Buffer, përshtatës impedance, nuk amplifikon.    |

## Modelimi i Sjelljes së Transistorit BJT


### **Modeli Ebers–Moll (Modeli i sinjalit të madh)**

Ky model përshkruan sjelljen e BJT-së në të gjitha regjimet e funksionimit duke përdorur dy dioda të lidhura në mënyrë të përshtatshme si ne figurë. Ai përfshin rrymat e emiterit ($I_E$), bazës ($I_B$) dhe kolektorit ($I_C$) si funksion të tensioneve përkatëse.

<img src="figs/lec7/Ebers-Moll_model_schematic_(NPN).svg" alt="npn cross section" width="40%" style="background-color:#fff;" > <img src="figs/lec7/Ebers-Moll_model_schematic_(PNP).svg" alt="npn cross section" width="40%" style="background-color:#fff;" >

Figura e parë (majtas) është pvr transistorin NPN kurse e dyta (djathtas) për atë PNP. $I_{CD}$ dhe $I_{ED}$ janë rrymat e diodës në kolektor dhe emiter, respektivisht. $\alpha_F$ dhe $\alpha_R$ janë përforcimet e rrymës në drejtimin e drejtë (forward) dhe të kundërt (reverse).

**Ekuacionet:**

$$
\begin{aligned}
I_E &= I_{ES} \left( e^{\frac{V_{BE}}{V_T}} - 1 \right) - \alpha_R I_{CS} \left( e^{\frac{V_{BC}}{V_T}} - 1 \right) \\
I_C &= \alpha_F I_{ES} \left( e^{\frac{V_{BE}}{V_T}} - 1 \right) - I_{CS} \left( e^{\frac{V_{BC}}{V_T}} - 1 \right) \\
I_B &= I_E - I_C
\end{aligned}
$$

ku:
* $I_{ES}$ dhe $I_{CS}$ janë rrymat e ngopjes për kalimet p-n emiter-bazë dhe kolektor-bazë, përkatësisht dhe jane te rendit ($10^{-15}-10^{-12}$ A).
* $\alpha_F$ dhe $\alpha_R$ janë fitimet përpara dhe mbrapa të rrymës së bazës.
* $V_T$ është tensioni termik.


### **Efekti Early**

Efekti Early përshkruan ndryshimin e gjërësisë efektive të bazës së BJT me ndryshimin e tensionit midis bazës dhe kolektorit $V_{CB}$. Me rritjen e tensionit të kundërt midis bazës dhe kolektorit (duke mbajtur konstant tensionin midis bazës dhe emiterit $V_{BE}=const$), rriet gjërësia e zonës së varfëruar të bazës duke zvogëluar numrin e ngarkesave që mund të rrinë në bazë, që përkthehet në një rritje të rrymës së kolektorit.

<img src="figs/lec7/Early_effect_(NPN).svg" alt="npn cross section" width="40%" style="background-color:#fff;" > 

E thene ndryshe, efekti Early shpjegon rritjen e rrymës së kolektorit me rritjen e tensionit kolektor-emiter ($V_{CE}$) për shkak të ngushtimit e gjërësisë efektive të bazës së BJT.

$$
I_C = I_S e^{\frac{V_{BE}}{V_T}} \left(1 + \frac{V_{CE}}{V_A}\right)
$$

ku $V_A$ është tensioni Early, një parametër që përfaqëson efektin e modifikimit të gjerësisë së bazës. Termi i parë deri te eskponenciali përfaqëson rrymën që rrjedh nëë kalimin p-n te bazë-emiter (rryma e bazës $I_B$), kurse termi në kllapa është përforcimi i rrymës së bazës $\beta$. 
Pra nëse $\beta_0$ është përforcimi për $V_{CE}=0$, atëhere
$$\beta= \beta_0\,\left(1 + \frac{V_{CE}}{V_A}\right)$$

Ku $V_A$ eshte tensioni Early qe varion tipikisht nga 15 - 150V, dhe eshte me i vogel per pajisje me te vogla.

<img src="figs/lec7/Early_effect_ICvsVCE).svg" alt="npn cross section" width="80%" style="background-color:#fff;" >



### **Modeli Hybrid-π (Modeli i sinjalit të vogël)**

Ky model përdoret për analizën e sinjaleve të vogla në regjimin aktiv të BJT-së. Ai përfshin elementë si rezistenca e hyrjes ($r_\pi$), transkonduktanca ($g_m$) dhe rezistenca e daljes ($r_o$).

<img src="figs/lec7/H_pi_model.svg" alt="npn cross section" width="40%" style="background-color:#fff;" >

**Ekuacionet:**

$$
\begin{aligned}
i_b &= \frac{v_{be}}{r_\pi} \\
i_c &= g_m v_{be} \\
r_o &= \frac{V_A + V_{CE}}{I_C}
\end{aligned}
$$

ku: 
* $r_\pi = \frac{\beta}{g_m}$
* $g_m = \frac{I_C}{V_T}$
* $r_o$ përfaqëson rezistencën e daljes për shkak të efektit Early.


### **Modeli me parametra h (Modeli i sinjalit të vogël)**

Ky model përshkruan BJT-në si një rrjet me dy porta duke përdorur parametra h për të lidhur rrymat dhe tensionet hyrëse dhe dalëse.

**Ekuacionet për konfigurimin me emiter të përbashkët:**

$$
\begin{aligned}
v_b &= h_{ie} i_b + h_{re} v_c \\
i_c &= h_{fe} i_b + h_{oe} v_c
\end{aligned}
$$

ku:
* $h_{ie}$ është rezistenca e hyrjes
* $h_{fe}$ është fitimi i rrymës
* $h_{re}$ është raporti i tensionit të kthimit
* $h_{oe}$ është konduktanca e daljes



### **Modeli Gummel–Poon (Modeli i sinjalit të madh)**

Ky model është një zgjerim i modelit Ebers–Moll dhe përfshin efekte si modifikimi i fitimit të rrymës me rrymën dhe injektimi në nivele të larta. Ai përdoret për simulime më të sakta të BJT-së në qarqe të ndryshme.


### **Modeli VBIC (Vertical Bipolar Inter-Company)**

Modeli VBIC është një model i avancuar që përmirëson saktësinë e simulimeve të BJT-së duke përfshirë efekte termike dhe sjellje në frekuenca të larta. Ai është një zgjerim i modelit Gummel–Poon dhe përdoret në simulime të avancuara të qarqeve.


### **Modeli HICUM (High Current Model)**

Modeli HICUM është një model fizik i avancuar për BJT-të që përfshin efekte të ndryshme si efektet termike, kapacitetet parazitare dhe sjelljen në frekuenca të larta. Ai është i përshtatshëm për simulime të saktë të qarqeve të avancuara.



### **Përmbledhje e Modeleve Themelore të Transistorit Bipolar (BJT)**

| **Modeli**              | **Përshkrimi i shkurtër**                                                                                                         |
| ----------------------- | --------------------------------------------------------------------------------------------------------------------------------- |
| **Ebers–Moll**          | Model i sinjalit të madh. Përshkruan të gjitha regjimet e funksionimit. Bazohet në dy dioda dhe dy burime rryme.                  |
| **Efekti Early**        | Ndryshon rrymën e kolektorit me rritjen e $V_{CE}$. Përdoret për të përfshirë efektin e ngushtimit të bazës.                   |
| **Hybrid-π**            | Model për sinjale të vogla. Përdor transkonduktancën $g_m$, rezistencën e bazës $r_\pi$, dhe daljen $r_o$.               |
| **Modeli h-parametrik** | Përdor parametrat eksperimentalë $h_{ie}$, $h_{fe}$, $h_{oe}$, $h_{re}$ për analiza në topologji të ndryshme.         |
| **Gummel–Poon**         | Model i zgjeruar për simulime më të sakta. Përfshin efektet jo-lineare dhe mbingarkesën e bazës.                                  |

$$
I_C \approx \beta I_B
$$

ku $\beta$ është fitimi i rrymës (zakonisht 50-300).

## Si funksionon një FET? (Shembull: n-channel MOSFET)

### Struktura:

- **Source**: furnizon elektrone.
- **Drain**: largon elektrone.
- **Gate**: kontrollon përmes një izolatori (zakonisht SiO₂).

### Operimi:

- **Pa tension në gate**: kanali nuk ekziston (në MOSFET të modës së rritjes).
- **Me tension pozitiv në gate**: elektrone mblidhen në sipërfaqen e trupit → formohet kanal për përçueshmëri.
- **Rezultati**: **tensioni në gate** kontrollon **rrymën Source-Drain**.


## 8. Modelet teorike bazë

### Për BJT:

- **Ebers-Moll Model**:
  - Përshkruan rrjedhën e rrymës në të dyja drejtimet.
- **Në regjimin aktiv**:
  $$
  I_C \approx I_S e^{\frac{V_{BE}}{V_T}}
  $$

ku:
- $I_S$ = rryma e saturimit,
- $V_T$ = tensioni termik ($\approx 26$ mV në $300$K).

### Për FET:

- **Modeli gradual i kanalit**:
  - Në regjimin linear:
    $$
    I_D = \mu_n C_{ox} \frac{W}{L} \left( (V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right)
    $$
  - Në regjimin e saturimit:
    $$
    I_D = \frac{1}{2} \mu_n C_{ox} \frac{W}{L} (V_{GS} - V_{th})^2
    $$



## 9. Rëndësia dhe Përdorimi i Transistorëve

- **BJT** përdoret në:
  - Përforcues audio
  - Përforcues të fuqisë
  - Komutatorë në rryma të mëdha
- **FET** përdoret në:
  - Mikroprocesorë
  - Memories RAM
  - Komutacion shumë efikas energjie



## 10. Konkluzion

- Një **transistor** është një **kontrollues i rrymës** me bazë në **fizikën e bartësve të ngarkesës** në materiale gjysmëpërçuese.
- Kontrolli mund të jetë:
  - **Me rrymë** (BJT, me injeksion dhe difuzion)
  - **Me tension** (FET, me fushë elektrike)
- Është ndërtuar mbi:
  - **Doping-un e gjysmëpërçuesve**
  - **Bashkimet p-n**
  - **Transportin e bartësve** (difuzion dhe drift)

**Në thelb:**  
Një transistor është një **pajisje që mundëson elektronikën moderne duke kontrolluar energjinë elektrike në mënyrë inteligjente, efikase dhe të shpejtë.**

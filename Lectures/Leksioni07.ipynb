# **I. Hyrje**

## **Çfarë është një transistor?**

Një transistor është një pajisje elektronike gjysmëpërçuese e cila sillet ose si një çelës i komanduar elektrikisht ose si një amplifikator kontrolluar elektrikisht. Në thelb të funksionit të transistorit është kontrolli i rrjedhjes së rrymës, analoge me kontrollin e rrjedhjes së ujit nga një rubinet. Ndryshe nga pajisjet pasive që kemi parë deri tani që kanë patur vetëm dy terminale, transistorët kanë edhe një terminal shtesë për kontrollin, pra janë pajisje me tre terminale. Një sinjal i vogël tensioni apo rryme që aplikohet mbi terminalin e kontrollit, mundësohet një ndryshim të madh rryme mbi dy terminalet e tjera të tij. Për shkak të numrit të madh të aplikimeve që ato kanë, transistorët gjënden praktikisht në çdo qark elektrik të pajisjeve moderne.


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përkufizimi**

**Një transistor është një pajisje elektronike gjysmëpërçuese me tre terminale që kontrollon rrjedhën e rrymës në një qark, nëpërmjet një sinjali kontrollues të jashtëm (rrymë ose tension).**

- Nëse kontrolli bëhet me rrymë (injeksion bartësish) ➔ kemi transistor bipolar (Bipolar Junction Transistor, BJT shkurt).
- Nëse kontrolli bëhet me tension (kontroll elektrostatik të kanaleve) ➔ kemi transistor me efekt fushe (Field Effect Transistor, FET shkurt).

</div>

- Transistori kontrollon rrymën me një sinjal të vogël elekrik (rrymë apo tension) dhe e **amplifikon** efektin e këtij sinjali në një sinjali shumë më të madh në dalje. Pra **transistori është një amplifikator**.
- Për sinjale mjaftueshmërisht të mëdha, **transistori kalon në regjim ngopje** dhe **kthehet efektivisht në një çelës elektrik të komandueshëm**.
- Ai është **zemra** e elektronikës moderne: mikroprocesorët, sensorët, komunikimi digjital, etj — të gjitha varen nga transistorët.


<br>

👉 **Në thelb**, një transistor është një **valvul elektrike**: përdor një **sinjal të vogël** për të kontrolluar **një sinjal më të madh**.

## Përmbledhje historike e transistorit

- **1925:** Lilienfeld imagjinoi një pajisje qe mund ti komandohej rezistenca elektrike nëpërmjet fushës elektrike, por materialet nuk ishin mjaftueshëm të pastra.
- Në vitet 1940, teknologjia bazohej në **tubat vakum** për përforcim dhe komutacion — këto ishin të mëdha, konsumonin shumë energji dhe ishin të paqëndrueshme. Kërkohej **një zëvendësues me përmasa të vogla, i besueshëm dhe energjitikisht efikas**.
- **1947:** Shpikja e parë praktike: BJT (Bardeen, Brattain, Shockley në Bell Labs).
- **1959:** Pasi u zgjidhën problemet me oksidet (SiO₂), u bë i mundur ndertimi i MOSFET.
- Sot: MOSFET dominon elektronikën digjitale dhe IC-të (integrated circuits), BJT përdoret kryesisht në përforcues ose qarkë analogë me rrymë të madhe.

<img src="figs/lec7/Triode_symbol.svg" alt="Triode" width="44%" style="background-color:#fff;" > <img src="figs/lec7/Triode.svg" alt="Triode" width="45%" style="background-color:#fff;" >


**Ironi historike**:  
Pavarësisht se transistori BJT (1947) u zbulu përpara FET (1959), në fakt, ideja e parë ishte për një **Field Effect Transistor (FET)** (1925) — por atë kohë, nuk ishte e mundur për shkak të papërsosmërisë së materialeve.

# **II.A — Struktura dhe Funksionimi Fizik i Transistorit Bipolar (BJT)**

## **1. Struktura Bazë e BJT**

<img src="figs/lec7/npn_Cross-section.svg" alt="npn cross section" width="30%" style="background-color:#fff;" > <img src="figs/lec7/BJT_pnp_npn.svg" alt="bjt" width="50%" style="background-color:#fff;" >

Një **transistor bipolar me bashkim** (BJT) përbëhet nga **tre rajone gjysmëpërçuese**:
* **Emiter (E)** – shumë i dopuar për të siguruar shumë bartës.
* **Bazë (B)** – e hollë dhe lehtë e dopuar, e cila lejon vetëm një pjesë të vogël të rekombinimit.
* **Kolektor (C)** – më pak i dopuar dhe më i gjerë, për të thithur bartësit.

BJT-të ndahen në dy lloje:
* **NPN**: n–p–n
* **PNP**: p–n–p

Në vijim do të fokusohemi tek tipi **NPN**, pasi është më i përdorur dhe më intuitiv për analizë. Analiza për transistorin PNP është identike, vetëm ndryshojmë polaritetin e tensioneve dhe kahun e rrymës. Cdo gjë tjetër në analizë është identike.


## **2. Injektimi dhe Transporti i Bartësve**

Në një transistor **NPN** që operon në **regjimin aktiv përpara**, kemi:
* Bashkimi **emiter-bazë** është i **polarizuar drejt**.
* Bashkimi **bazë-kolektor** është i **polarizuar në kundërt**.

Kjo gjeneron këtë seri ngjarjesh:
1. Elektronet injektohen nga **emiteri** (n) në **bazën** (p).
2. Baza është **shumë e hollë**, kështu që **shumica e elektroneve nuk rekombinohen**, por kalojnë përmes bazës.
3. Në anën tjetër, **fusha elektrike e bashkimit bazë-kolektor** i tërheq këto elektrone drejt **kolektorit**.

<img src="figs/lec7/npn_Operation.svg" alt="npn cross section" width="80%" style="background-color:#fff;" >

<img src="figs/lec7/NPN_Band_Diagram_Active.svg" alt="npn cross section" width="50%" style="background-color:#fff;" >


## **3. Rrymat përmes të tre terminaleve**

Në këtë proces, lindin tre rryma kryesore:
* **$I_E$**: rryma e emiterit — rrjedh jashtë nga emiteri.
* **$I_B$**: rryma e bazës — shumë e vogël (rreth 1% e $I_E$).
* **$I_C$**: rryma e kolektorit — gati e barabartë me $I_E$.

Për rrymat vlen:
$$ I_E = I_B + I_C $$

Rryma e kolektorit është:
$$ I_C = \beta I_B $$
ku $\beta$ është **fitimi i rrymës** (current gain), zakonisht $50 \leq \beta \leq 300$.


## **4. Thelbi fizik i përforcimit**

Amplifikimi ndodh sepse:
* Një **rrymë shumë e vogël** në bazë ($I_B$)
* **Kontrollon rrjedhën e një rryme shumë më të madhe** në kolektor ($I_C$)

Ky raport krijon efektin e **përforcimit të rrymës**.


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përmbledhje Strukturës & Funksionit**

* BJT përdor **injektimin e bartësve** dhe **fushën elektrike** për të kontrolluar rrjedhjen e rrymës.
* **Struktura asimetrike** (baza e hollë, emiteri i dopuar) është çelësi për amplifikim efikas.
* Ky funksion fizik bën të mundur modelimin si përforcues apo komutator.

# **II.B — Regjimet e Funksionimit të Transistorit Bipolar (BJT)**

Transistori bipolar (BJT) ka **katër regjime kryesore funksionimi**, të përcaktuara nga **polarizimi i dy bashkimeve p-n**:
* **Bashkimi emiter–bazë (BE)**
* **Bashkimi bazë–kolektor (BC)**

Këto regjime janë të rëndësishme për të kuptuar **si përdoret BJT si përforcues apo si çelës**.

<img src="figs/lec7/bjt_char.png" alt="npn cross section" width="50%" style="background-color:#fff;" >


## **Tabela përmbledhëse e regjimeve**

| Regjimi         | Polarizimi i BE | Polarizimi i BC | Përshkrimi                              |
| --------------- | --------------- | --------------- | --------------------------------------- |
| Aktiv përpara   | Drejt           | Kundërt         | Përforcues: $I_C \approx \beta I_B$ |
| Saturim         | Drejt           | Drejt           | Çelës i mbyllur: $I_C$ maksimal      |
| Cutoff  | Kundërt         | Kundërt         | Çelës i hapur: $I_C \approx 0$       |
| Aktiv i kundërt | Kundërt         | Drejt           | Funksion i anasjelltë (rrallë përdoret) |


## **1. Regjimi Aktiv Përpara (Forward Active Region)**

* **BE** është i polarizuar **drejt**:
  $$ V_{BE} \gtrsim 0.6\ \text{V (për silic)} $$
* **BC** është i polarizuar **në kundërt**:
  $$ V_{CB} > 0 $$

**Rezultati:**
* Elektronet injektohen nga emiteri në bazë.
* Kolektori i tërheq → rrjedh $I_C \approx \beta I_B$.
* Ky është regjimi **i dëshiruar për amplifikim**.

## **2. Regjimi i Saturimit**

* Të dy bashkimet **BE dhe BC** janë **të polarizuara drejt**.
  $$ V_{BE} \gtrsim 0.6\, \text{V},\quad V_{CB} < 0 $$

**Rezultati:**
* Zona e bazës është “e mbushur me bartës”.
* Transistori sillet si një **çelës i mbyllur**.
* Rryma $I_C$ është e kufizuar nga qarku → nuk ndjek më $I_C = \beta I_B$.


## **3. Regjimi i Prerjes (Cutoff Region)**

* Të dy bashkimet janë **të polarizuara në kundërt**:
  $$ V_{BE} < 0,\quad V_{CB} < 0 $$

**Rezultati:**
* Asnjë bartës nuk injektohet.
* Rryma $I_C \approx 0$ → **transistori është i fikur**.
* Përdoret si çelës **OFF** në logjikën digjitale.

## **4. Regjimi Aktiv i Kundërt**

* **BE** është në **kundërt** (reverse bias) dhe **BC** në **drejt** (forward bias).
* Struktura është e anasjelltë → kolektori vepron si emiter.

**Rezultati:**
* Funksionon, por **me efikasitet shumë më të ulët**:
  * $\beta_{\text{kthim}} \ll \beta_{\text{përpara}}$
* Përdoret rrallë → vetëm në analiza teorike ose tranzicion gjatë komutimit.

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përmbledhje e Regjimeve**

* Regjimi **aktiv përpara** është **zemra e përforcimit analog**.
* Regjimet **saturim** dhe **cutoff** janë themelore në **logjikën digjitale** (ON/OFF).
* Kuptimi i polarizimeve për secilën gjendje është thelbësor për projektim qarku.

# **II.C — Modelimi Matematikor i BJT: Modeli Ebers–Moll dhe Përafrimet**

## **Qëllimi i modelimit**

Modeli matematikor i transistorit bipolar shërben për të përshkruar **rrymat nëpër terminalet e tij** ($I_E$, $I_B$, $I_C$) në funksion të **tensioneve të aplikuara** dhe të **vetive fizike** të pajisjes.

Modeli **Ebers–Moll** është një nga më të rëndësishmit, pasi:

* Përshkruan të **gjitha regjimet** e funksionimit të BJT-së.
* Është bazuar në **transportin e bartësve** dhe **sjelljen e bashkimeve p-n**.
* Përdor konceptin e dy **rrymave të ngopjes** dhe dy **faktorëve të fitimit** (gain) 


## **1. Ekuacionet e plota të modelit Ebers–Moll**

<img src="figs/lec7/Ebers-Moll_model_schematic_(NPN).svg" alt="npn cross section" width="40%" style="background-color:#fff;" > <img src="figs/lec7/Ebers-Moll_model_schematic_(PNP).svg" alt="npn cross section" width="40%" style="background-color:#fff;" >

Për një transistor **NPN**, rrymat përshkruhen si më poshtë:

$$
\begin{aligned}
I_E &= I_{ES} \left( e^{\frac{V_{BE}}{V_T}} - 1 \right) - \alpha_R I_{CS} \left( e^{\frac{V_{BC}}{V_T}} - 1 \right) \\
I_C &= \alpha_F I_{ES} \left( e^{\frac{V_{BE}}{V_T}} - 1 \right) - I_{CS} \left( e^{\frac{V_{BC}}{V_T}} - 1 \right) \\
I_B &= I_E - I_C
\end{aligned}
$$

**Simbolet dhe kuptimet fizike**

| Simbol        | Kuptimi                                                           |
| ------------- | ----------------------------------------------------------------- |
| $I_{ES}$   | Rryma e ngopjes së bashkimit emiter–bazë                          |
| $I_{CS}$   | Rryma e ngopjes së bashkimit bazë–kolektor                        |
| $V_{BE}$   | Tensioni midis bazës dhe emiterit                                 |
| $V_{BC}$   | Tensioni midis bazës dhe kolektorit                               |
| $V_T$      | Tensioni termik, $V_T = \frac{kT}{q} \approx 25.85\, \text{mV}$ |
| $\alpha_F$ | Fitimi përpara (forward gain) i bartësve (zakonisht $0.98 - 0.998$)            |
| $\alpha_R$ | Fitimi kthim (reverse gain) (zakonisht shumë më i vogël: $0.2 - 0.5$)          |


## **2. Përafrime në regjimin aktiv përpara**

<img src="figs/lec7/Approximated_Ebers_Moll.svg" alt="npn cross section" width="40%" style="background-color:#fff;" >

Kur BJT është në **regjimin aktiv përpara**:
* $V_{BE} \gg V_T$ → $e^{\frac{V_{BE}}{V_T}} \gg 1$
* $V_{BC}$ është negativ (polarizim i kundërt) → $e^{\frac{V_{BC}}{V_T}} \approx 0$

### ➤ Thjeshtimet bëhen:
$$
\begin{aligned}
I_C &\approx \alpha_F I_{ES} e^{\frac{V_{BE}}{V_T}} \\
I_E &\approx I_{ES} e^{\frac{V_{BE}}{V_T}} \\
I_B &\approx I_E - I_C = (1 - \alpha_F) I_{ES} e^{\frac{V_{BE}}{V_T}}
\end{aligned}
$$


### Marrëdhënia me parametërin $\beta$

Shpesh preferohet të përdorim:
$$
\beta = \frac{\alpha_F}{1 - \alpha_F}
\quad \text{dhe} \quad
\alpha_F = \frac{\beta}{1 + \beta}
$$

Duke përdorur këto, kemi:
$$
I_C = \beta I_B \quad \text{dhe} \quad I_E = (1 + \beta) I_B
$$


## **3. Efekti Early dhe modeli i përmirësuar**

Kur rritet $V_{CE}$ në regjimin aktiv, zona e bazës zvogëlohet → rritet $I_C$ pak më shumë se sa do duhej sipas modelit eksponencial.

<img src="figs/lec7/Early_effect_(NPN).svg" alt="npn cross section" width="40%" style="background-color:#fff;" >


### Korrigjohet me:
$$
I_C = I_S e^{\frac{V_{BE}}{V_T}} \left(1 + \frac{V_{CE}}{V_A}\right)
$$
ku:
* $V_A$ është **tensioni Early** (zakonisht $50 - 150$ V)
* Ky efekt është i rëndësishëm në **modelet e sinjalit të vogël**, duke përcaktuar rezistencën e daljes:
$$
r_o = \frac{V_A}{I_C}
$$

* Modeli **Ebers–Moll** përshkruan plotësisht sjelljen e një BJT-je në të gjitha regjimet.
* Për analiza të zakonshme (regjimi aktiv përpara), përdorim thjeshtuara që fokusohen vetëm te bashkimi emiter–bazë.
* **Efekti Early** përshkruan ndryshimin e gjërësisë efektive të bazës së BJT me ndryshimin e tensionit midis bazës dhe kolektorit $V_{CB}$ (kur $V_{BE}=\text{const}$).
* Efekti **Early** është përmirësim që saktëson përllogaritjet e rrymave te transistorit në qark real.

# **II.D — Modelimi i Sinjalit të Vogël për BJT**


Modelimi i sinjalit të vogël është thelbësor për të kuptuar dhe analizuar se si BJT vepron **si përforcues**, kur sinjalet që i aplikohen janë **shumë më të vogla se tensionet DC të polarizimit**.


## **Qëllimi**

* **Përshkrim linear** i transistorit rrotull pikës së punës (dmth Q-point).
* Llogaritja e: **fitimit të tensionit** (gain), **impedancës së hyrjes/daljes**, dhe **reagimit në frekuencë**.
* Modelet bazohen në **diferencimin e modeleve jo-lineare**.

## **1. Modeli hybrid-$\pi$ ($\pi$ sepse forma e qarkut ngjan me shkronjën greke)**

### **Komponentët e modelit hybrid-$\pi$ për NPN në konfigurim me emiter të përbashkët:**

| Elementi               | Përkufizimi                                                 | Fizika                                  |
| ---------------------- | ----------------------------------------------------------- | --------------------------------------- |
| $r_\pi$             | Impedanca e hyrjes bazë-emiter                              | $r_\pi = \frac{\beta}{g_m}$         |
| $g_m v_{be}$       | Burimi i rrymës së kontrolluar nga tensioni                 | $g_m = \frac{I_C}{V_T}$            |
| $r_o$               | Rezistenca e daljes (nga efekti Early)                      | $r_o = \frac{V_A}{I_C}$            |
| $C_\pi$, $C_\mu$ | Kapaciteti i bashkimeve BE dhe BC (në frekuenca të larta) | Jo gjithmonë përfshihen në modelin bazë |

### **Ekuacionet themelore:**
$$
g_m = \frac{I_C}{V_T}, \quad
r_\pi = \frac{\beta}{g_m}, \quad
r_o = \frac{V_A}{I_C}
$$


### **Vlera tipike të parametrave:**

* $I_C = 1$ mA
* $\beta = 100$
* $V_T = 25$ mV
* $V_A = 100$ V

Atëherë:
$$
g_m = \frac{1\text{ mA}}{25\text{ mV}} = 40\text{ mS}, \quad
r_\pi = \frac{100}{40\text{ mS}} = 2.5\ \text{k}\Omega, \quad
r_o = \frac{100\text{ V}}{1\text{ mA}} = 100\ \text{k}\Omega
$$

## **2. Modeli me parametra h (h-parameter model)**

Modeli h përfaqëson transistorin si një **rrjet me dy porta**, duke lidhur tensionin dhe rrymën në hyrje me ato në dalje.

### Ekuacionet (konfigurimi me emiter të përbashkët):

v_b &= h_{ie} i_b + h_{re} v_c \\
i_c &= h_{fe} i_b + h_{oe} v_c
\end{aligned}
$$

| Parametri   |                     Përkufizimi                   |
| ----------- | ----------------------------------------------- |
| $h_{ie}$ | Impedanca e hyrjes                                |

* Përforcimi (gain) rritet me $g_m$ dhe me $R_C$.
* Për tension të vogël hyrës, dalja është e përforcuar dhe e përmbysur.

# **II.E — Aplikime Praktike të Transistorit BJT në Qark**


## **Qëllimi i këtij seksioni**

Të tregojmë **si përdoret një BJT në praktikë**, si **përforcues i sinjalit** ashtu edhe si **çelës** në qark elektronik. Këto dy funksione përbëjnë bazën e përdorimit të BJT në elektronikë **analoge** dhe **digjitale**.


## **1. BJT si Përforcues**

<img src="figs/lec7/common_EBC.jpeg" alt="npn cross section" width="50%" style="background-color:#fff;" >

Transistori përdoret si **përforcues i tensionit** në konfigurime të ndryshme:

### **1.1 Përforcuesi me emiter të përbashkët (Common Emitter Amplifier)**

* Konfigurimi më i zakonshëm.
* Jep **përforcimin më të madh të tensionit** dhe **përmbysje faze** (180°).

#### Skema e përgjithshme:

* Hyrja aplikohet në **bazë**.
* Dalja merret nga **kolektori**.
* **Emiteri** është tokë për sinjalin AC (direkt ose përmes një kondensatori by-pass).

#### Fitimi i tensionit:
$$
A_v = \frac{v_{out}}{v_{in}} \approx -g_m (R_C \parallel r_o)
$$

#### Karakteristika:
* Fitim i lartë.
* Përmbysje faze.
* Impedancë hyrëse $r_\pi$.
* Impedancë daljeje $r_o$.


### **1.2 Përforcuesi me kolektor të përbashkët (Common Collector / Buffer)**

* Hyrja në bazë, dalja nga emiter.
* **Fitimi i tensionit afër 1** (pra, nuk përforcon tensionin por përforcon rrymën).
* Impedancë hyrëse e lartë, impedancë daljeje e ulët.

#### Përforcimi:
$$
A_v \approx \frac{r_E}{r_E + r_\pi / (\beta + 1)} \approx 1
$$

#### Përdorim:
* Si **buffer** për ndarjen/deciftimin e qarqeve me ngarkesa të ndryshme.


### **1.3 Përforcuesi me bazë të përbashkët (Common Base Amplifier)**

* Hyrja në **emiter**, dalja në **kolektor**, baza është tokë për AC.
* Përforcim i lartë tensioni, **pa përmbysje faze**.
* Impedancë hyrëse shumë e ulët.

#### Përdorim:
* Përdoret në qarqe me **frekuenca të larta** (RF-radio frequency).


## **2. BJT si Komutator (Çelës)**

Kur përdoret si **çelës**, BJT operon në **dy regjime**:

* **Cutoff** → transistori është **i fikur**.
* **Saturim** → transistori është **i ndezur**.

### Aplikim tipik:

#### Qark komandues i LED:

* **Baza** lidhet përmes një rezistence me një sinjal kontrolli.
* Kur sinjali është **i lartë**: BJT futet në **saturim** → LED ndizet.
* Kur sinjali është **i ulët**: BJT në **cutoff** → LED fiket.


#### Përllogaritje praktike:

1. Zgjidh **rezistencën e bazës** për të dhënë rrymën e nevojshme $I_B$:
   $$ R_B = \frac{V_{in} - V_{BE}}{I_B}, \quad \text{me} \quad I_B = \frac{I_C}{\beta_{forced}} $$
2. Ku $\beta_{forced}$ është zakonisht më i vogël se vlera nominale, për siguri.


## **Shembuj të zakonshëm praktikë**

| Aplikimi               | Roli i BJT          | Regjimi          |
| ---------------------- | ------------------- | ---------------- |
| Qark përforcuesi audio | Përforcues tensioni | Aktiv përpara    |
| LED driver             | Çelës ON/OFF        | Saturim / Cutoff |
| Qark logjik TTL        | Portë logjike       | Komutacion       |
| Sensor aktiv           | Amplifikon sinjalin | Aktiv përpara    |


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje**

* BJT mund të përdoret për **përforcim** (aktiv përpara) dhe **komutacion** (cutoff/saturim).
* Konfigurimi me **emiter të përbashkët** është më i fuqishmi për amplifikim.
* **Fitimi** dhe **impedanca** varen nga konfigurimi dhe polarizimi.
* Në përdorim si çelës, dimensionimi i rrymës së bazës është kyç për kontroll të saktë.

# **III.A — Transistori me Efekt të Fushës (FET): Llojet dhe Struktura Fizike**

## **Qëllimi i këtij seksioni**

Të prezantojmë transistori **FET (Field-Effect Transistor)** si një pajisje që, ndryshe nga BJT, kontrollon rrymën nëpërmjet një **fusha elektrike** në vend të injektimit të bartësve.

FET është bazë për shumë sisteme moderne, veçanërisht **MOSFET-i**, që përbën pjesën më të madhe të tranzistorëve në elektronikën digjitale (p.sh., në mikroprocesorë).

## **1. Parimi themelor i FET**

* FET është një pajisje **me tre terminale**:
  * **Gate (G)**: porta që kontrollon fushën elektrike
  * **Source (S)**: burimi i bartësve
  * **Drain (D)**: destinacioni i bartësve
* Bartësit rrjedhin në një **kanal përçues** (conduction channel), i cili kontrollohet nga tensioni në **Gate (G)**.
* Nuk ka nevojë për rrymë në portë (rasti ideal), vetëm **tension** → **pajisje e kontrolluar nga tensioni**.

Seksionin tërthorë i një JFET

<img src="figs/lec7/JFET_cross_section.svg" alt="npn cross section" width="40%" style="background-color:#fff;" >


Seksionin tërthorë i një MOSFET

<img src="figs/lec7/FET_cross_section.svg" alt="npn cross section" width="40%" style="background-color:#fff;" >


## **2. Llojet kryesore të FET**

### **JFET (Junction FET)**

* Kanali është gjithmonë i pranishëm (depletion-mode).
* Tensioni negativ në Gate ngushton kanalin (në n-JFET).
* Struktura bazohet në një **bashkim p-n** për të kontrolluar përçueshmërinë.

### **MOSFET (Metal-Oxide-Semiconductor FET)**

* Ka një shtresë **oksid izolues** midis Gate dhe trupit gjysmëpërçues.
* Gate **nuk ka kontakt të drejtpërdrejtë** me kanalin → **impedancë e lartë hyrëse**.

#### Dy mënyrat kryesore:

| Tipi i MOSFET-it     | Sjellje pa tension në Gate                                                |
| -------------------- | ------------------------------------------------------------------------- |
| **Enhancement-mode** | Kanali **nuk ekziston** pa $V_{GS}$ → duhet fushë për ta krijuar       |
| **Depletion-mode**   | Kanali ekziston dhe mund të **mbyllet** me tension negativ në $V_{GS}$ |


<img src="figs/lec7/FET_Symbols.svg" alt="npn cross section" width="90%" style="background-color:#fff;" >


## **3. Struktura e MOSFET-it (n-channel, enhancement-mode)**


### Kur $V_{GS} > V_{th}$:

* Elektronet tërhiqen në sipërfaqe → krijohet një **kanal n-type**.
* Rryma $I_D$ rrjedh nga **Drain** në **Source**.


## **4. Përmbledhje krahasuese: JFET vs MOSFET**

| Veçori                | JFET                | MOSFET                     |
| --------------------- | ------------------- | -------------------------- |
| Izolimi i Gate        | Bashkim p-n         | Izolim oksid (SiO₂)        |
| Impedancë hyrëse      | Mesatare            | Shumë e lartë (10⁶–10¹² Ω) |
| Mënyra e funksionimit | Depletion           | Enhancement ose Depletion  |
| Prodhimi masiv        | Rrallë përdoret sot | Përdorim universal në IC   |

IC = Integrated Circuits!


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje**

* FET kontrollon rrjedhjen e rrymës përmes **një fushe elektrike**, jo me rrymë si BJT.
* Lloji më i përdorur është **MOSFET**, për shkak të:
  * Impedancës së lartë hyrëse
  * Konsumit të ulët të energjisë
  * Shkallëzueshmërisë për qarkë të integruar

# **III.B — Regjimet e Funksionimit të MOSFET-it**

## **Qëllimi i këtij seksioni**

Të shpjegojmë **si ndryshon sjellja e një MOSFET-i në varësi të tensionit $V_{GS}$ (Gate–Source)** dhe **$V_{DS}$ (Drain–Source)**. Këto regjime përcaktojnë **nëse kanali është i hapur, i pjesshëm ose plotësisht i formuar**, dhe çfarë karakteristikash ka rryma $I_D$.

Do të fokusohemi në **MOSFET n-channel në modën e rritjes (enhancement-mode)**, pasi është më i përdoruri në praktikë.


## **1. Tensionet përkatëse**

* $V_{GS}$ — kontrollon formimin e kanalit.
* $V_{th}$ — tensioni i pragut (threshold voltage), zakonisht 1–2 V.
* $V_{DS}$ — aplikimi i tensionit për të nxitur rrjedhjen e rrymës nga Drain në Source.


## **2. Tre regjimet kryesore të funksionimit**


### **1. Regjimi i Prerjes (Cutoff Region)**

**Kusht:**
$$ V_{GS} < V_{th} $$

**Çfarë ndodh:**
* Nuk formohet kanal përçues.
* Rryma $I_D \approx 0$.
* Transistori është **i fikur**.

**Përdorim:** si **çelës OFF** në qark digjital.


### **2. Regjimi Linear (Triode / Linear Region)**

**Kusht:**
$$ V_{GS} > V_{th} \quad \text{dhe} \quad V_{DS} < V_{GS} - V_{th} $$

**Çfarë ndodh:**
* Formohet një kanal **i plotë përçues**.
* Rryma varet nga $V_{DS}$ dhe është **e kontrollueshme linearisht**.
* Sjellja është si një **rezistencë e kontrolluar me tension**.

**Rryma:**
$$ I_D = \mu_n\, C_{ox}\, \frac{W}{L}\, \left[ (V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right] $$

**Përdorim:** si **rezistencë e ndryshueshme** ose për përforcues në frekuenca të ulëta.

### **3. Regjimi i Saturimit (Active Region)**

**Kusht:**
$$ V_{GS} > V_{th} \quad \text{dhe} \quad V_{DS} \geq V_{GS} - V_{th} $$

**Çfarë ndodh:**
* Kanali "ngushtohet" në afërsi të Drain → ndodh **pinch-off**.
* Rryma **bëhet e pavarur nga $V_{DS}$**.
* Transistori punon si **burim rryme i kontrolluar nga tensioni**.

**Rryma:**
$$ I_D = \frac{1}{2} \mu_n\, C_{ox}\, \frac{W}{L}\, (V_{GS} - V_{th})^2 $$

**Përdorim:** në **qarkë përforcuese analogë**.


<img src="figs/lec7/IvsV_mosfet.svg" alt="npn cross section" width="70%" style="background-color:#fff;" >



<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje e Regjimeve MOSFET n-channel enhancement-mode**

| Regjimi         | Kushti                             | Sjellje         | Përdorim   |
| --------------- | ---------------------------------- | --------------- | ---------- |
| Cutoff          | $V_{GS} < V_{th}$              | Kanal i mbyllur | Çelës OFF  |
| Linear (Triode) | $V_{DS} < V_{GS} - V_{th}$    | Kanal përçues   | Rezistencë |
| Saturim         | $V_{DS} \geq V_{GS} - V_{th}$ | Rrymë konstante | Përforcues |


* MOSFET ka tre regjime bazë, të përcaktuara nga $V_{GS}$ dhe $V_{DS}$.
* Regjimi i **saturimit** është ai ku përdoret për **amplifikim**.
* Regjimi **linear (triode)** është si **rezistencë e komanduar me tension**.
* Regjimi **cutoff** e bën transistorin të mos përçojë rrymë.


**Paralelizëm me BJT**

| BJT                          | MOSFET                          |
| ---------------------------- | ------------------------------- |
| $I_B$ kontrollon $I_C$ | $V_{GS}$ kontrollon $I_D$ |
| Regjimi aktiv                | Regjimi i saturimit             |
| Regjimi i prerjes            | Regjimi Cutoff                  |
| Komandim me rrymë            | Komandim me tension             |

# **III.C — Modelimi Matematikor i MOSFET-it**

## **Qëllimi i këtij seksioni**

Të formulojmë **ekuacionet matematikore për rrymën e drain-it $I_D$**, në funksion të tensioneve të kontrollit $V_{GS}$ dhe $V_{DS}$, për të tre regjimet e funksionimit të MOSFET-it.

Sic kemi përmendur, fokusi do të jetë te **n-channel enhancement-mode MOSFET**, për shkak të përdorimit të gjerë në elektronikën digjitale dhe analoge.


## **1. Parametrat fizikë të modelit**

| Parametër    | Kuptimi                                          |
| ------------ | ------------------------------------------------ |
| $\mu_n$   | Mobiliteti i elektroneve në kanalin n          |
| $C_{ox}$  | Kapaciteti për njësi sipërfaqe i shtresës okside |
| $W$, $L$ | Gjerësia dhe gjatësia e kanalit                  |
| $V_{th}$  | Tensioni i pragut (threshold voltage)            |
| $V_{GS}$  | Tensioni portë–burim                             |
| $V_{DS}$  | Tensioni drain–burim                             |
| $I_D$     | Rryma nga drain në source                        |


## **2. Ekuacionet sipas regjimit të funksionimit**

### a) **Regjimi i Cutoff**

Kusht:
$$ V_{GS} < V_{th} $$

Rryma:
$$ I_D \approx 0 $$


### b) **Regjimi Linear (Triode)**

Kusht:
$$ V_{GS} > V_{th} \quad \text{dhe} \quad V_{DS} < V_{GS} - V_{th} $$

Rryma:
$$ I_D = \mu_n C_{ox} \frac{W}{L} \left[ (V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right] $$

Ky është regjimi ku MOSFET sillet si **rezistencë e komanduar me tension**.


### c) **Regjimi i Saturimit (Active)**

Kusht:
$$ V_{GS} > V_{th} \quad \text{dhe} \quad V_{DS} \geq V_{GS} - V_{th} $$

Rryma:
$$ I_D = \frac{1}{2}\, \mu_n\, C_{ox} \,\frac{W}{L}\, (V_{GS} - V_{th})^2 $$

Në këtë regjim, rryma është **e pavarur nga $V_{DS}$** dhe kontrollohet vetëm nga $V_{GS}$ — për këtë arsye, MOSFET quhet **burim rryme i komanduar nga tensioni**.


## **3. Model i përmbledhur i rrymës $I_D$**

Për të gjithë regjimet në një formulë të vetme:
$$
I_D = 
\begin{cases}
0, & V_{GS} < V_{th} \\
\mu_n C_{ox} \frac{W}{L} \left[(V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right], & V_{DS} < V_{GS} - V_{th} \\
\frac{1}{2} \mu_n C_{ox} \frac{W}{L} (V_{GS} - V_{th})^2, & V_{DS} \geq V_{GS} - V_{th}
\end{cases}
$$


## **4. Përkufizimi i faktorëve të projektimit**

Përdoret shpesh faktori:
$$
k' = \mu_n C_{ox}
\quad \text{dhe} \quad
k = k' \cdot \frac{W}{L}
$$

Kështu formulat bëhen më kompakte:

* **Regjimi linear**:
$$ I_D = k\; \left[(V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right] $$

* **Regjimi i saturimit**:
$$ I_D = \frac{1}{2}\, k \,(V_{GS} - V_{th})^2 $$


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje**

| Regjimi | Kushti                             | \$I\_D\$                               |
| ------- | ---------------------------------- | -------------------------------------- |
| Cutoff  | $V_{GS} < V_{th}$              | $I_D \approx 0$                     |
| Linear  | $V_{DS} < V_{GS} - V_{th}$    | $I_D \propto V_{DS}$               |
| Saturim | $V_{DS} \geq V_{GS} - V_{th}$ | $I_D \propto (V_{GS} - V_{th})^2$ |

</div>

# **III.D — Modelimi i Sinjalit të Vogël për MOSFET**

## **Qëllimi i këtij seksioni**

Të përshkruajmë **modelin linear të MOSFET-it** kur përdoret në regjimin e saturimit për të analizuar **sinjale të vogla AC** — përforcuesit analogë, qarkët e filtrimit, amplifikatorë të tensionit apo rrymës.

Ky model është ekuivalent me modelin **hybrid-$\pi$** të përdorur për BJT.

## **1. Supozimet themelore**

* MOSFET-i është në **regjimin e saturimit**:
  $$ V_{GS} > V_{th},\quad V_{DS} > V_{GS} - V_{th} $$
* Qarku është **në pikën e punës (Q-point)** dhe mbi këtë pikë aplikohet një **sinjal i vogël AC**.
* Variacionet janë mjaftueshmërisht të vogla për të linearizuar ekuacionin $I_D = f(V_{GS})$.

## **2. Elementët e modelit të sinjalit të vogël për MOSFET**

| Simbol      | Përshkrimi                                       | Formula                                                                              |
| ----------- | ------------------------------------------------ | ------------------------------------------------------------------------------------ |
| $g_m$    | Transkonduktanca (si "fitimi" gain)                   | $g_m = \frac{\partial I_D}{\partial V_{GS}} = \frac{2 I_D}{V_{GS} - V\_{th}}$ |
| $r_{ds}$ | Rezistenca e daljes (nga mod. gjatësisë kanalit) | $r_{ds} = \frac{1}{\lambda I_D}$                                                 |
| $v_{gs}$ | Tensioni diferencial Gate–Source                 | —                                                                                    |
| $i_d$    | Rryma e kanalit në sinjalin të vogël              | $i_d = g_m v_{gs}$                                                              |

> Ku $\lambda$ është parametri i modulimit të gjatësisë së kanalit, tipikisht shumë i vogël ($\lambda \approx 0.01$ V⁻¹)


## **3. Qarku përfaqësues i modelit të sinjalit të vogël**



* **Nuk ka rrymë hyrëse në Gate** → impedancë hyrëse ideale: $R_{in} = \infty$
* **Burimi i rrymës së kontrolluar nga $v_{gs}$** përcakton daljen.
* **$r_{ds}$** vendoset paralel me ngarkesën në drain.


## **4. Fitimi i tensionit për qark përforcues Common Source**

Për një përforcues të thjeshtë me **rezistencë $R_D$ në drain**, fitimi është:
$$ A_v = \frac{v_{out}}{v_{in}} = -g_m (R_D \parallel r_{ds}) $$

Nëse supozohet $r_{ds} \gg R_D$, atëherë:
$$ A_v \approx -g_m \, R_D $$

* Shenja minus tregon **përmbysje faze**
* Përforcimi varet vetëm nga $g_m$ dhe rezistenca ngarkesës (load resistor)


## **5. Shembull praktik numerik**

Le të themi:
* $I_D = 1$ mA
* $V_{GS} - V_{th} = 1$ V
* $\lambda = 0.02$ V⁻¹

Atëherë:
$$ g_m = \frac{2 I_D}{V_{GS} - V_{th}} = \frac{2 \cdot 1\text{ mA}}{1\text{ V}} = 2\text{ mS} $$

$$ r_{ds} = \frac{1}{\lambda I_D} = \frac{1}{0.02 \cdot 1\text{ mA}} = 50\text{ k}\Omega  $$

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje**

* **Modeli i sinjalit të vogël për MOSFET** është thelbësor për analizë AC në përforcues.
* Fitimi përcaktohet nga \$g\_m\$ dhe ngarkesa në drain.
* Impedanca hyrëse është shumë e lartë → ideal për buffer.
* Efekti i modulimit të kanalit (\$\lambda\$) futet përmes \$r\_{ds}\$, i cili mund të përmirësojë saktësinë në frekuenca të larta.

</div>

# **III.E — Aplikime Praktike të MOSFET-it në Qark**


## **Qëllimi i këtij seksioni**

Të tregojmë se si përdoret **MOSFET-i** në qark për **përforcim analog** dhe **komutacion digjital**, duke analizuar konfigurimet klasike:

* **Common Source (CS)** — përforcues tensioni me përmbysje faze
* **Common Drain (CD)** — buffer tensioni
* **Common Gate (CG)** — përdoret në frekuenca të larta


## **1. MOSFET si Përforcues**

### **1.1 Përforcuesi me Source të Përbashkët (Common Source Amplifier)**

* **Hyrja në Gate**, **dalja në Drain**, **Source është i tokëzuar**.
* Sjellje si përforcues tensioni me **përmbysje faze**.
* Rritja e $V_{in}$ → rrit $I_D$ → rrit rënien në $R_D$ → ul $V_{out}$.

#### Përforcimi i tensionit:

$$ A_v = \frac{v_{out}}{v_{in}} = -g_m \; (R_D \parallel r_{ds}) $$

#### Karakteristika:
* Përforcimi i lartë (high gain).
* Impedancë hyrëse shumë e lartë.
* Impedancë daljeje rreth $R_D$.


### **1.2 Përforcuesi me Drain të Përbashkët (Common Drain / Buffer)**

* **Hyrja në Gate**, **dalja në Source**, **Drain është fiksuar në tension ($V_{DD}$)**.
* Përforcimi i tensionit është **afër 1**.
* Nuk përforcon tensionin, por përforcon **rrymën**.

#### Fitimi:
$$ A_v \approx \frac{g_m R_S}{1 + g_m R_S} \approx 1 \quad \text{nëse } g_m R_S \gg 1
$$

#### Karakteristika:
* Impedancë hyrëse shumë e lartë.
* Impedancë daljeje e ulët.
* Përdoret si **buffer tensioni**.


### **1.3 Përforcuesi me Gate të Përbashkët (Common Gate)**

* **Hyrja në Source**, **dalja në Drain**, **Gate është tokë për AC**.
* Nuk ka përmbysje faze.
* Përdoret në **frekuenca të larta** për shkak të kapacitetit të ulët hyrës.

#### Fitimi:
$$ A_v = g_m\; (R_D \parallel r_{ds}) $$

#### Karakteristika:
* Impedancë hyrëse e ulët.
* Impedancë daljeje mesatare.
* Frekuencë kufi më e lartë se CS (common source).


## **2. MOSFET si Çelës (Switch)**

### Regjime të përdorura:

* **OFF**: $V_{GS} < V_{th}$ → kanali nuk formohet → asnjë rrymë.
* **ON**: $V_{GS} > V_{th}$ dhe $V_{DS} \ll V_{GS} - V_{th}$ → përçueshmëri e lartë (regjimi triode).

### Aplikime tipike:

* **Inverter logjik (NOT gate)** me MOSFET.
* **Komandim i LED, motorëve, relëve**.

#### Përllogaritje për dimensionim:
* Zgjedhje e $V_{GS}$ për të kaluar nga cutoff në triode.
* Vlerësim i $R_{DS(on)}$, për të minimizuar humbjet.


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

####  **Përmbledhje e Konfigurimeve**

| Konfigurimi          | Fitimi \$A\_v\$                    | Faza | $R_{in}$       | $R_{out}$ | Përdorim                    |
| -------------------- | ---------------------------------- | ---- | ------------- | -------- | --------------------------- |
| Common Source (CS)   | $-g_m (R_D \parallel r_{ds})$ | 180° | Shumë i lartë | $R_D$     | Përforcues tensioni         |
| Common Drain (CD) | Afër 1                             | 0°   | Shumë i lartë | I ulët   | Buffer (impedance matching) |
| Common Gate (CG)     | $g_m (R_D \parallel r_{ds})$  | 0°   | I ulët        | Mesatar  | Frekuenca të larta          |


* **Common Source** jep përforcim tensioni me përmbysje faze.
* **Source Follower** është buffer — përdoret për ndarje fazash.
* **Common Gate** jep përforcim pa përmbysje, përdoret për frekuenca të larta.
* Si **çelës**, MOSFET është superior për shkak të kontrollit pa konsum energjie.


**Krahasim me BJT**

| Aspekti             | BJT                 | MOSFET                     |
| ------------------- | ------------------- | -------------------------- |
| Kontrolli           | Me rrymë ($I_B$) | Me tension ($V_{GS}$)   |
| Impedanca hyrëse    | e ulët              | shumë e lartë              |
| Konsumi i energjisë | më i lartë          | shumë i ulët               |
| Shpejtësia          | më e ulët në IC     | shumë e lartë (CMOS, GHz+) |

# **IV — Krahasimi Midis BJT dhe MOSFET**

## **Qëllimi i këtij seksioni**

Të bëjmë një krahasim të plotë ndërmjet **transistorit bipolar (BJT)** dhe **transistorit me efekt fushe (MOSFET)** në aspektin fizik, funksional dhe praktik, për të ndihmuar në **zgjedhjen e duhur të pajisjes** në projektim elektronik.


## **1. Krahasimi i parimit të funksionimit**

| Aspekti               | BJT                                         | MOSFET                                           |
| --------------------- | ------------------------------------------- | ------------------------------------------------ |
| Mekanizmi kryesor     | Injektim dhe difuzion i bartësve            | Modulim i kanalit me fushë elektrike             |
| Kontrolli             | Me **rrymë** ($I_B$ kontrollon $I_C$) | Me **tension** ($V_{GS}$ kontrollon $I_D$) |
| Transporti i bartësve | **Bipolar** (elektrone dhe vrima)           | **Unipolar** (vetëm bartësit kryesorë)           |
| Tipi i bartësve       | Elektrone + vrima                           | Elektrone (n-MOS) ose vrima (p-MOS)              |

## **2. Parametrat Elektrikë**

| Parametri              | BJT                               | MOSFET                              |
| ---------------------- | --------------------------------- | ----------------------------------- |
| Impedanca hyrëse       | E ulët (kΩ)                       | Shumë e lartë (MΩ–TΩ)               |
| Konsumi në hyrje       | I pranishëm ($I_B$)            | I papërfillshëm (Gate current ≈ 0)  |
| Përforcimi i rrymës        | $I_C = \beta I_B$             | Nuk ka përforcimi rryme                  |
| Përforcimi i tensionit     | Lartë në konfigurime të caktuara  | Lartë në konfigurime të caktuara    |
| Rezistenca e daljes    | E moderuar ($r_o$)             | E lartë ($r_{ds}$, por ndryshon) |
| Tensioni i aktivizimit | $V_{BE} \approx 0.6-0.7$ V | $V_{th}$ tipikisht 1–2 V         |


## **3. Sjellja dinamike dhe në frekuencë**

| Aspekti                   | BJT                           | MOSFET                             |
| ------------------------- | ----------------------------- | ---------------------------------- |
| Shpejtësia e komutacionit | Më e ngadaltë (ruan ngarkesë) | Më e shpejtë (kapacitiv në hyrje)  |
| Kapacitete parazitare     | Të ulëta                      | Të pranishme: $C_{gs}, \; C_{gd}$ |
| Shpejtësia e komutimit     | Deri në 100 MHz               | Deri në GHz                        |


## **4. Përdorimi në praktikë**

| Aplikimi                                              | Pajisja më e përshtatshme |
| ----------------------------------------------------- | ------------------------- |
| Përforcues audio (me tension të lartë, zhurmë të ulët)  | **BJT**                   |
| Komutacion digjital, CMOS logic                       | **MOSFET**                |
| Qarkë të integruara me densitet të lartë (IC)         | **MOSFET** (CMOS)         |
| Përforcues të sinjalit të vogël në RF                 | **BJT** (CB) / **MOSFET CG**   |
| Qarkë të thjeshtë analoge me polarizim të qëndrueshëm | **BJT**                   |


## **5. Tabela përmbledhëse krahasuese**

| Karakteristikë     | BJT                             | MOSFET                                      |
| ------------------ | ------------------------------- | ------------------------------------------- |
| Lloji i kontrollit | Rrymë                           | Tension                                     |
| Tipi i bartësit    | Bipolar (elektrone + vrima)     | Unipolar (elektrone ose vrima)              |
| Konsum në hyrje    | Ka rrymë në bazë                | Pothuajse zero                              |
| Impedancë hyrëse   | E ulët                          | Shumë e lartë                               |

| Shpejtësia         | E kufizuar nga ngarkesa e bazës | E lartë (për shkak të kontrollit kapacitiv) |
| Përmasat në IC     | Më i madh                       | I vogël, lehtësisht i integrueshëm          |
| Dominon në:        | Elektronikë analoge, fuqie      | Elektronikë digjitale, CMOS                 |

* **MOSFET** dominon në komutacion, në **elektronikë digjitale**, dhe për **efikasitet energjitik**.

# **V — Modele të Avancuara dhe Simulimi i Transistorëve**

## **Qëllimi i këtij seksioni**

Të prezantojmë **modelet më të avancuara fizike dhe empirike** që përdoren në **simulime kompjuterike**, veçanërisht për qarqet e integruara dhe elektroniken me performancë të lartë.


## **1. Pse nevojiten modele të avancuara?**

Modelet idealë si **Ebers–Moll**, **hybrid-$\pi$** apo **$g_m$ për MOSFET** janë të vlefshëm për:
* Analizë të sinjalit të vogël
* Dizajne themelore (jo të avancuara).

**Por nuk mjaftojnë** për:
* **Qarkë komplekse**, me shumë pajisje në IC.
* **Variacione termike**, jo-linearitete të rrymave të mëdha.
* **Frekuenca shumë të larta** apo **rryma të ngarkesës dinamike**


## **2. Modelet e avancuara për BJT**

### **2.1 Modeli Gummel–Poon**

Zgjerim i modelit Ebers–Moll me:
* **Fitim rryme jo konstant** (varet nga $I_C$)
* **Efektet e ngopjes së përçueshmërisë**
* **Më shumë saktësi për simulime SPICE**

### **2.2 Modeli VBIC (Vertical Bipolar Inter-Company)**

* Adreson kufizimet e modelit Gummel–Poon.
* Përfshin:
  * **Efektet termike**
  * **Early effect të avancuar**
  * **Ngopje jo-lineare**
* Është **standard për simulime të saktë të BJT** në IC.

### **2.3 Modeli HICUM (High Current Model)**

* Optimizuar për:
  * **Frekuenca shumë të larta (RF)**
  * **Teknologji BiCMOS**
* I bazuar në fizikë të plotë me transport 2D.

## **3. Modelet e avancuara për MOSFET**

### **3.1 Modeli BSIM (Berkeley Short-channel IGFET Model)**

* Modelet BSIM3, BSIM4 dhe BSIM6 janë standard në CMOS.
Përfshin:
* **Efekte të kanalit të shkurtër**
* **Degradimin e mobilitetit**
* **Efekte termike dhe zhurmën**
* **Rryma të nënpragut (subthreshold conduction)**

### **3.2 Modelet për frekuenca të larta**

* **EKV (Enz–Krummenacher–Vittoz)**: për modelim të përpiktë analog.
* **HiSIM (Hiroshima-University)**: i bazuar në fizikën e nano-teknologjive.


## **4. Simulimi SPICE**

* **SPICE (Simulation Program with Integrated Circuit Emphasis)** është standard për simulimin e qarkut.

### Sintaksë për BJT:

```
Q1 C B E QNPN
.model QNPN NPN (IS=1e-15 BF=100 VAF=100)
```

### Sintaksë për MOSFET:

```
M1 D G S 0 NMOS
.model NMOS NMOS (VTO=1 KP=50u LAMBDA=0.01)
```

Mund të përdorësh programe si:

* **LTspice**
* **Ngspice**
* **Qucs**
* **TINA-TI**
* **Spectre (Cadence)** për industri


## **5. Krahasim i modeleve për SPICE**

| Pajisja | Model i thjeshtë | Model i avancuar           | Teknologji që e përdor |
| ------- | ---------------- | -------------------------- | ---------------------- |
| BJT     | Ebers–Moll       | Gummel–Poon, VBIC, HICUM   | BiCMOS, HBT            |
| MOSFET  | Shkallë e dytë   | BSIM3, BSIM4, BSIM6, HiSIM | CMOS, FinFET, FD-SOI   |


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përmbledhje**

* Modelet e avancuara janë **jetike për simulime reale dhe dizajn të integruar**.
* Zgjedhja e modelit ndikon në **saktësinë**, **kohën e simulimit**, dhe **rezultatet praktike**.
* Projektimi modern bazohet në këto modele për optimizim termik, konsum energjie dhe performancë frekuencore.

# **VI — Projektimi i Pikës së Qetësisë (Q-point) dhe Teknikat e Polarizimit të Transistorëve**


## **Qëllimi i këtij seksioni**

Të kuptojmë:
* Çfarë është **Pika e Qetësisë (Q-point)**,
* Si zgjidhet për një funksionim të qëndrueshëm,
* Metodat më të përdorura për **polarizimin e transistorit** (biasing), për të vendosur atë Q-point në regjimin e dëshiruar.


## **1. Çfarë është Pika e Qetësisë (Q-point)?**

**Pika e Qetësisë (Q)** është gjendja e transistorit kur:
* Nuk ka sinjal hyrës (vetëm DC),
* Të gjitha tensionet dhe rrymat janë të **qëndrueshme (steady-state)**.

### Për BJT:
* Q-pika përcaktohet nga:
  * $V_{CE}$ — tensioni kolektor–emiter
  * $I_C$ — rryma e kolektorit

### Për MOSFET:
* Q-pika përcaktohet nga:
  * $V_{DS}$ — tensioni drain–source
  * $I_D$ — rryma e kanalit

**Qëllimi:** Ta vendosim transistorin në **regjimin aktiv/saturim**, larg nga cutoff dhe saturimi i thellë, që ai të **punojë në regjim linear dhe të qëndrueshëm**.


## **2. Teknikat e polarizimit për BJT**

### **2.1 Polarizimi me ndarës tensioni (Voltage Divider Bias)**

**Skema:**
* Dy rezistenca $R_1$ dhe $R_2$ formojnë një ndarës që vendos $V_B$.
* $R_E$ stabilizon rrymën përmes feedback termik.

**Avantazhe:**
* Qëndrueshmëri termike
* Q-pikë pak e varur nga $\beta$

### Përllogaritja:
$$
V_B = V_{CC} \cdot \frac{R_2}{R_1 + R_2}, \quad
V_E = V_B - V_{BE}, \quad
I_E = \frac{V_E}{R_E}
$$

$$
I_C \approx I_E, \quad
V_{CE} = V_{CC} - I_C (R_C + R_E)
$$


### **2.2 Polarizimi me rezistencë bazë (Base Resistor Bias)**

**Skema më e thjeshtë**, vetëm një rezistencë në bazë ($R_B$).

**Disavantazhe:**
* Varet shumë nga $\beta$
* Ndjeshëm ndaj ndryshimeve termike


## **3. Teknikat e polarizimit për MOSFET**

### **3.1 Polarizimi me rezistencë source (Self-Bias / Source Degeneration)**

**Përdorim i një rezistence $R_S$** për të përcaktuar $I_D$:
$$ V_{GS} = V_G - I_D\, R_S $$

Përdoret për **qëndrueshmëri më të madhe**.


### **3.2 Polarizimi me ndarës tensioni**

Ashtu si me BJT:
* Përdoret ndarës tensioni në portë për të vendosur $V_G$.
* $R_S$ stabilizon $I_D$.

$$ 
V_{GS} = V_G - I_D\, R_S
\quad \Rightarrow \quad
I_D = \frac{1}{2} \, k\, (V_{GS} - V_{th})^2
$$

Kërkon **zgjidhje numerike** për $I_D$.


## **4. Qëndrueshmëria e Q-pikës**

Pika e qetësisë duhet të jetë:
* **Larg cutoff dhe saturimit** (për të siguruar punën në regjimin linear)
* **E qëndrueshme ndaj ndryshimeve të temperaturës**
* **Jo shumë e varur nga variacionet e $\beta$ apo $V_{th}$**

### Përdorimi i rezistencës në emiter/source:

* Krijon **negative feedback** → rregullon automatikisht $I_C$/$I_D$


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përmbledhje**

* **Q-point** përcakton funksionimin e tranzistorit: vendoset përmes **polarizimit DC**.
* Teknikat si **ndarësi tensioni** me **rezistencë në emiter/source** ofrojnë **qëndrueshmëri të lartë**.
* Projektimi i polarizimit është **një nga hapat më të rëndësishëm** në dizajn qarku.

# **VII — Përdorim Praktik dhe Ushtrime Laboratorike me Transistorë**


## **Qëllimi i këtij seksioni**

Të përforcojmë të gjitha konceptet teorike të mëparshme përmes:
* Matjeve eksperimentale,
* Konfigurimeve reale të qarkut,
* Përdorimit të pajisjeve laboratorike si **oshiloskopi**, **gjeneratori i funksioneve**, **furnizuesi DC**, dhe **multimetri**.


## **1. Matja e karakteristikave $I-V$ të BJT dhe MOSFET**

### BJT – Karakteristika $I_C$ vs $V_{CE}$ për $I_B$ të ndryshme

#### Pajisjet:
* BJT NPN (p.sh. 2N3904/BC547 etj)
* Rezistencë ngarkese $R_C$
* Burim DC për $V_{CE}$ dhe $I_B$
* Ampermetër dhe voltmetër

#### Procedura:
1. Fikso $I_B\$ në një vlerë të caktuar (p.sh., 10 μA).
2. Varion $V_{CE}$ dhe mat $I_C$.
3. Përsërit për disa vlera të $I_B$.

#### Rezultati:
* Marrim **një grup kurbash karakteristike** që shfaqin **regjimet cutoff, aktiv dhe saturim**.


### MOSFET – Karakteristika $I_D$ vs $V_{DS}$ për vlera të ndryshme të $V_{GS}$

#### Pajisjet:
* MOSFET n-channel (p.sh. IRF540N)
* Burim DC për $V_{GS}$ dhe $V_{DS}$
* Ampermetër për $I_D$

#### Procedura:
1. Fikso $V_{GS}$ mbi $V_{th}$ (p.sh., 2.5 V, 3.3 V, 4 V).
2. Varion $V_{DS}$ dhe mat $I_D$.
3. Përsërit për çdo vlerë të $V_{GS}$.

#### Rezultati:
* Kurba $I_D$ vs $V_{DS}$ që tregojnë **regjimin triode dhe saturim**.


## **2. Konfigurimi dhe testimi i përforcuesve**

### Përforcuesi me emiter të përbashkët (BJT)

* Përdor një rezistencë në kolektor ($R_C$) dhe ndarës tensioni për polarizim.
* Apliko një sinjal të vogël sinus ($\sim$10 mV) në bazë.
* Mat $V_{out}$ në kolektor me osciloskop.

**Analiza:**
* Vëzhgo përmbysjen e fazës.
* Llogarit fitimin $A_v = v_{out} / v_{in}$.


### Përforcuesi common source (MOSFET)

* Polarizo portën me ndarës tensioni.
* Source i tokëzuar, drain përmes $R_D$.
* Apliko sinjal AC në Gate dhe shiko përgjigjen në Drain.

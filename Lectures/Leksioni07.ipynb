# **Transistorët**

## **Çfarë është një transistor?**

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

#### **Përkufizimi**

**Një transistor është një pajisje elektronike gjysmëpërçuese me tre terminale që kontrollon rrjedhën e rrymës në një qark, nëpërmjet një sinjali kontrollues të jashtëm (rrymë ose tension).**

- Nëse kontrolli bëhet me rrymë (injeksion bartësish) ➔ kemi transistor bipolar (Bipolar Junction Transistor, BJT shkurt).
- Nëse kontrolli bëhet me tension (kontroll elektrostatik të kanaleve) ➔ kemi transistor me efekt fushe (Field Effect Transistor, FET shkurt).

</div>

- Transistori kontrollon rrymën me një sinjal të vogël elekrik (rrymë apo tension) dhe e **amplifikon** efektin e këtij sinjali në një sinjali shumë më të madh në dalje. Pra **transistori është një amplifikator**.
- Për sinjale mjaftueshmërisht të mëdha, **transistori kalon në regjim ngopje** dhe **kthehet efektivisht në një çelës elektrik të komandueshëm**.
- Ai është **zemra** e elektronikës moderne: mikroprocesorët, sensorët, komunikimi digjital, etj — të gjitha varen nga transistorët.


<br>

👉 **Në thelb**, një transistor është një **valvul elektrike**: përdor një **sinjal të vogël** për të kontrolluar **një sinjal më të madh**.

## Përmbledhje historike e transistorit

- **1925:** Lilienfeld imagjinoi një pajisje qe mund ti komandohej rezistenca elektrike nëpërmjet fushës elektrike, por materialet nuk ishin mjaftueshëm të pastra.
- Në vitet 1940, teknologjia bazohej në **tubat vakum** për përforcim dhe komutacion — këto ishin të mëdha, konsumonin shumë energji dhe ishin të paqëndrueshme. Kërkohej **një zëvendësues me përmasa të vogla, i besueshëm dhe energjitikisht efikas**.
- **1947:** Shpikja e parë praktike: BJT (Bardeen, Brattain, Shockley në Bell Labs).
- **1959:** Pasi u zgjidhën problemet me oksidet (SiO₂), u bë i mundur ndertimi i MOSFET.
- Sot: MOSFET dominon elektronikën digjitale dhe IC-të (integrated circuits), BJT përdoret kryesisht në përforcues ose qarkë analogë me rrymë të madhe.

**Ironi historike**:  
Pavarësisht se transistori BJT (1947) u zbulu përpara FET (1959), në fakt, ideja e parë ishte për një **Field Effect Transistor (FET)** (1925) — por atë kohë, nuk ishte e mundur për shkak të papërsosmërisë së materialeve.

## 3. Struktura Bazë dhe Tipet e Transistorëve

Dy jane tipet kryesore te transistoreve: 
1. Transistore Bipolare (Bipolar Junction Transistor, BJT shkurt)
2. Transistoret me Efekt Fushe (Field Effect Transistor, FET, shkurt)
   - JFET (Junction Field Effect Transistor)
   - MOSFET (

| Tipi        | Parimi kontrolli | Materiali kryesor | Përdorimi kryesor |
|-------------|------------------|-------------------|-------------------|
| **BJT**     | Kontroll i rrymës (current injection) | Si, Ge | Amplifikim dhe komutacion |
| **FET**     | Kontroll i tensionit (electric field) | Si, SiO₂ | Elektronikë digjitale dhe përforcim me fuqi të ulët |

### Tre terminalet e një transistori:

- **Hyrje e kontrollit** - Base (B) në BJT, Gate (G) në FET
- **Hyrje e energjisë** - Collector (C) në BJT, Drain (D) në FET
- **Dalje e energjisë** - Emitter (E) në BJT, Source (S) në FET



## 4. Parimet e Fizikës së Gjysmëpërçuesve

### a) Sjellja e materialit gjysmëpërçues

- Në një gjysmëpërçues të pastër (Si, Ge):
  - Në zero Kelvin: izolator
  - Në temperatura dhome: fluktuacinet termike mundësojne disa elektrone të ngjiten në bandën e përçueshmërisë (shpërndarja Boltzmann).
- **Doping**:
  - **n-type**: elektronë të tepërt (donorë pentavalent)
  - **p-type**: vrima të tepërta (akseptorë trivalent)

### b) Bashkimi p-n

- Kur bashkohen materialet **p** dhe **n** → formohet një **zonë e varfëruar** pa bartës të lirë → krijohet **fusha e brendshme elektrike**.
- Kalimi p-n është **baza** për diodën dhe për strukturën e transistorit.


## 5. Si funksionon një BJT?

### Struktura:
Shqyrtojmë rastin e npn

- **Emitter**: n-type (shumë i dopuar)
- **Base**: p-type (shumë e hollë dhe lehtë e dopuar)
- **Collector**: n-type (më pak i dopuar)

### Operimi:

- **Forward bias** në Emitter-Base → injektim i elektroneve në bazë.
- **Reverse bias** në Base-Collector → fushë që thith elektronet në kolektor.
- **Rezultati**: një **rrymë shumë e vogël në base** kontrollon një **rrymë të madhe emitter-collector**.

$$
I_C \approx \beta I_B
$$

ku $\beta$ është fitimi i rrymës (zakonisht 50-300).



## 6. Si funksionon një FET? (Shembull: n-channel MOSFET)

### Struktura:

- **Source**: furnizon elektrone.
- **Drain**: largon elektrone.
- **Gate**: kontrollon përmes një izolatori (zakonisht SiO₂).

### Operimi:

- **Pa tension në gate**: kanali nuk ekziston (në MOSFET të modës së rritjes).
- **Me tension pozitiv në gate**: elektrone mblidhen në sipërfaqen e trupit → formohet kanal për përçueshmëri.
- **Rezultati**: **tensioni në gate** kontrollon **rrymën Source-Drain**.



## 7. Diferencat Themelore midis BJT dhe FET

| Aspekti        | BJT                              | FET                         |
|----------------|-----------------------------------|------------------------------|
| Kontroll       | Rrymë ($I_B$)                    | Tension ($V_{GS}$)           |
| Bartës         | Elektrone dhe vrima (bipolar)     | Elektrone ose vrima (unipolar) |
| Mekanizmi      | Injeksion dhe difuzion             | Kontroll me fushë elektrike  |
| Impedanca Hyrëse | E ulët                           | Shumë e lartë                |
| Konsumi i energjisë | Më i lartë                   | Shumë i ulët                 |



## 8. Modelet teorike bazë

### Për BJT:

- **Ebers-Moll Model**:
  - Përshkruan rrjedhën e rrymës në të dyja drejtimet.
- **Në regjimin aktiv**:
  $$
  I_C \approx I_S e^{\frac{V_{BE}}{V_T}}
  $$

ku:
- $I_S$ = rryma e saturimit,
- $V_T$ = tensioni termik ($\approx 26$ mV në $300$K).

### Për FET:

- **Modeli gradual i kanalit**:
  - Në regjimin linear:
    $$
    I_D = \mu_n C_{ox} \frac{W}{L} \left( (V_{GS} - V_{th}) V_{DS} - \frac{V_{DS}^2}{2} \right)
    $$
  - Në regjimin e saturimit:
    $$
    I_D = \frac{1}{2} \mu_n C_{ox} \frac{W}{L} (V_{GS} - V_{th})^2
    $$



## 9. Rëndësia dhe Përdorimi i Transistorëve

- **BJT** përdoret në:
  - Përforcues audio
  - Përforcues të fuqisë
  - Komutatorë në rryma të mëdha
- **FET** përdoret në:
  - Mikroprocesorë
  - Memories RAM
  - Komutacion shumë efikas energjie



## 10. Konkluzion

- Një **transistor** është një **kontrollues i rrymës** me bazë në **fizikën e bartësve të ngarkesës** në materiale gjysmëpërçuese.
- Kontrolli mund të jetë:
  - **Me rrymë** (BJT, me injeksion dhe difuzion)
  - **Me tension** (FET, me fushë elektrike)
- Është ndërtuar mbi:
  - **Doping-un e gjysmëpërçuesve**
  - **Bashkimet p-n**
  - **Transportin e bartësve** (difuzion dhe drift)

**Në thelb:**  
Një transistor është një **pajisje që mundëson elektronikën moderne duke kontrolluar energjinë elektrike në mënyrë inteligjente, efikase dhe të shpejtë.**

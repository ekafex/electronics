# **Amplifikatorët Elektronikë: Amplifikatorët Operacionalë (op-amp)**

## **Amplifikatorët Elektronikë**

<img src="figs/lec8/960px-McIntosh_MC2505.jpg" alt="Vdiode" width="50%" style="background-color:#fff;" >


<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">
    
#### 1. Çfarë është një amplifikator?

Një **amplifikator elektronik** (ose thjesht “amp”) është një qark me dy porta që përdor energjinë e furnizimit për të **rritur amplitudën** e një sinjali në dalje krahasuar me hyrjen. Shkalla e rritjes quhet **amplifikimi (gain)** – raporti i tensionit, rrymës ose fuqisë mes daljes dhe hyrjes. **Një pajisje quhet amplifikator vetëm nëse amplifikimi i fuqisë është më i madh se 1**.

</div>


#### 2. Klasifikimet kryesore

| Bazuar te                | Shembull / Diapazon                                        | Përdorimi kryesor                                              |
| ------------------------ | ---------------------------------------------------------- | -------------------------------------------------------------- |
| **Frekuenca**            | Audio (< 20 kHz), RF (20 kHz – 300 GHz), DC/servo          | Pajisje hi-fi, radio-përcjellës, motor-drive |
| **Pozicioni në zinxhir** | *Preamplifikator* (para përpunimit), *power-amp* (në fund) | Kondicionim sensori vs. shtytje altoparlanti |
| **Klasa e përçimit**     | A, B, AB, C (lineare); D, E (me komutacion)                | Kompromis midis efikasitetit dhe linearitetit |
| **Konfigurimi fazor**    | Invertues, jo-invertues                                    | Vendos marrëdhënien e fazës hyrje-dalje       |


#### 3. Pajisjet aktive që bëjnë amplifikimin

* **Tubat vakum (triode, pentode)** – zanafilla e amplifikatorëve (1906-1960s).
* **Transistorët BJT & MOSFET** – sot përbëjnë shumicën e amplifikatorëve.
* Pajisje të veçanta RF (klystron, TWT) ose me **rezistencë negative** (tunel-diodë) për përdorime specifike.


#### 4. Parametrat kryesorë të performancës

| Parametri                          | Çfarë mat                                         | Pse ka rëndësi                           |
| ---------------------------------- | ------------------------------------------------- | ---------------------------------------- |
| **Amplifikimi ($A_v$, $A_i$, $A_p$)** | Raport dalje/hyrje                                | Sa “rritet” sinjali                      |
| **Gjerësia e brezit (BW)**         | Frekuenca ku amplifikimi bie 3 dB                 | Shpejtësia e sinjalit                    |
| **Efikasiteti**                    | Fuqia dalëse / konsum total                       | Kritik për amplifikatorë fuqie           |
| **Lineariteti & Shformimi**     | Sa i saktë kopjohet forma                         | Audio hi-fi kërkon THD < 0.01 %          |
| **Zhurma**                         | Volt/rrymë zhurme hyrëse                          | Limiti i sinjaleve shumë të vogla        |
| **Slew Rate, stabiliteti**         | Kufizimi i shpejtësisë & shmangia e vet-oshilimit | Cakton kufijtë praktikë |



#### 5. Parimi i funksionimit

Amplifikatori përfaqësohet si **burim i varur tensioni ose rryme**. Në praktikë, shumica përdorin **feedback negativ**: një pjesë e daljes kthehet në hyrje me fazë të kundërt për të ulur amplifikimin e hapur, por për të **rritur bandwidth-in dhe për të zvogëluar distorsionin**.


#### 6. Lloje të zakonshme

* **Audio power-amp** – ngre sinjalin e dobët audio në watt-e për të drejtuar altoparlantë.
* **RF power-amp** – faza finale e një transmetuesi, optimizuar për transferim fuqie në antenë.
* **Preamplifikator/instrumentacion** – amplifikim i tensioneve μV–mV me zhurmë ultra-të ulët.
* **Amplifikator operacional (op-amp)** – “bllok amplifikimi” me amplifikim të hapur shumë të madh, diferencial; karakteristikat përcaktohen nga qarku i jashtëm i feedback-ut.


#### 7. Aplikime tipike

Telefoni, audio hi-fi, përforcues kitarash, radarë, matës instrumentesh, kompjuterë analogë, servomotorë, të gjitha varen nga ndonjë formë amplifikatori për të përpunuar ose gjeneruar sinjale me nivele të përdorshme.


#### 8. Pikat kyçe për t’u mbajtur mend

1. **Amplifikimi** nuk krijon energji – energjia shtesë merret nga burimi i jashtëm.
2. Performanca përcaktohet nga **amplifikimi, bandwidth-i, lineariteti, zhurma, efikasiteti**.
3. **Klasa e përçimit** dhe **feedback-u negativ** janë mjetet kryesore për të balancuar efikasitetin me cilësinë e sinjalit.
4. **Transistorët** e kanë zëvendësuar tubin vakum, por teknikat themelore mbeten të njëjta.

## **Përforcuesi operacional (op-amp)**

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">

### **Çfarë është një op-amp?**

Një **op-amp** (amplifikator operacional) është një komponent elektronik (qark i integruar) që përdoret për të **amplifikuar diferencën e tensionit** midis dy pikave. Ai nuk punon vetëm, por gjithmonë është pjesë e një **qarku të jashtëm** që përcakton funksionin e tij.

> Quhet "operacional" sepse u përdor për herë të parë në **kumpjuterat analogë** për të kryer **operacione matematikore si mbledhje, zbritje, integrim, derivim, etj.**

</div>

<img src="figs/lec8/LM741CN.jpg" width="30%" style="background-color:#fff;" >


### **Si operon një op-amp?**

**Ka dy hyrje dhe një dalje**

<img src="figs/lec8/Generic_741_pinout_top.png" width="50%" style="background-color:#fff;" >

* **$V^+$** — hyrja **jo-invertuese**
* **$V^-$** — hyrja **invertuese**
* **$V_\text{out}$** — dalja (zakonisht tension)
* **$V_{s+}$** — ushqimi pozitiv (shpesh nuk shënohet në skema)
* **$V_{s-}$** — ushqimi negativ (shpesh nuk shënohet në skema)

Simboli:
<img src="figs/lec8/Op-amp_symbol.svg" width="30%" style="background-color:#fff;" >

<br>

Dalja prodhon një tension të barabartë me:
$$
V_\text{out} = A_0\, (V^+ - V^-)
$$

ku $A_0$ është **amplifikimi i hapur** – zakonisht **shumë i madh** (p.sh. e rendit $10^5$ ose më shumë).


#### **Funksioni i op-amp-it të hapur**

Nëse vemë në punë një op-amp, funksioni i tij është që të amplifikojë diferencën e tensioneve $V^+ - V^-$. Meqënëse $A_0$ është shumë e madhe (idealisht infinit), edhe një ndryshim i vogël mes $V^+$ dhe $V^-$ dërgon $V_\text{out}$ në $+\infty$ ose $-\infty$. Mëqe op-amp është lidhur me burimet e jashtme $V_{s+}$ dhe $V_{s-}$ vlerat e tensionit në dalje nuk mund të bëhen më të mëdha apo më të vogla se vlerat e tensionit të buimit, pra vlerat e $V_\text{out}$ shkon në saturim (maksimumi $V_{s+}$ ose minimumi $V_{s-}$. 
Pra:
* Nëse $V^+ - V^- > 0 \Rightarrow V_\text{out} = V_{s+}$
* Nëse $V^+ - V^- < 0 \Rightarrow V_\text{out} = V_{s-}$

Për të krijuar aplikime të dobishme me op-amp njerëzit kanë shpikur **feedback-un**, me konkretisht **feedback-un negativ**.

## **Feedback-u**?

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">
Në një sistem, **feedback** është procesi i kthimit mbrapsht të një pjese të sinjalit dalës në hyrje.

Kjo lidhje “mbrapsht” ndihmon sistemin të **vetë-rregullohet**, **të përforcohet** apo **të stabilizohet** — sipas mënyrës së implementimit.
</div>

> Në një qark elektronik me op-amp, feedback-u është mënyra se si ne përgatisim op-amp-in të sillet si amplifikator, integrues, filtër etj.


### Llojet e feedback-ut

1. **Feedback negativ**
    > Dalja ushqehet mbrapsht me **fazë të kundërt** në hyrje.
    * Synon të **zvogëlojë diferencën** ndërmjet sinjalit hyrës dhe sinjalit të rikthyer.
    * Efekti neto: **stabilitet**, **linearitet**, **kontroll i amplifikimit**, **ulje shformimit dhe zhurmës**.
    * **Përdoret në:**
      * Amplifikatorë linearë
      * Regullatorë tensioni
      * Filtra aktive
      * Sisteme kontrolli (servo, audio) etj.

2. **Feedback pozitiv**
    > Dalja ushqehet mbrapsht me **fazë të njëjtë** në hyrje.
    * **Rrit diferencën** mes hyrjes dhe reagimit.
    * Kjo e çon sistemin drejt **vetë-shtimit** dhe **saturimit** → shkakton **ndezje ose kalim të shpejtë në një gjendje tjetër**.
    * **Efekti kryesor:** e bën sistemin **jolinear**, të **vetë-aktivizohet** sapo kapërcen një prag.
    * **Përdoret në:**
      * Korahasues me histerezi (Schmitt trigger)
      * Multivibratorë (gjenerator sinjali katror)
      * Oshilatorë (RC, LC, Wein Bridge) etj.

### Pse është i rëndësishëm feedback-u?

| Efekt               | Feedback negativ             | Feedback pozitiv        |
| ------------------- | ---------------------------- | ----------------------- |
| **Stabiliteti**     | Përmirësohet                 | Mund të destabilizojë   |
| **Lineariteti**     | Përmirësohet                 | Ulet                    |
| **Zhurma**          | Zvogëlohet                   | Mund të rritet          |
| **Vetë-aktivizimi** | Jo                           | Po (qark ndizet vetë)   |
| **Aplikimi tipik**  | Amplifikatorë, filtra, servo | Oscilatorë, komparatorë |


**Feedback-u** është një nga idetë më themelore në elektronikë:
* **Feedback-u negativ** është **zemra e elektronikës së kontrolluar** — na lejon të ndërtojmë sisteme stabël dhe të sakta.
* **Feedback-u pozitiv** është **motori i aktivizimeve** — nga sinjale logjike te sinjale të vetëgjeneruara (oshilatoret).

**Në qarqet me op-amp, çdo funksion lidhet me mënyrën se si përdorim feedback-un.** Prandaj, njohja dhe kontrolli i feedback-ut është çelësi për të kuptuar dhe dizenjuar qarqe elektronike.

## **Rregullat e analizimit te qarkut me op-amp ideal**

<div style="border:5px solid #0a0; background-color:#ca0; color:#003; border-radius: 15px; padding: 10px;">


1. **Rregulli 1: Amplifikimi i hapur në op-amp-in ideal $A_0 \to \infty$**
   Për një **op-amp ideal**, amplifikimi i tensionit në regjim të hapur është **i pafundmë** $ A_0 = \infty $. Për një **op-amp real**, amplifikimi është një vlerë e madhe por **e fundme**, zakonisht në intervalin: $ 10^4 \text{ deri } 10^6 $

2. **Rregulli 2: Impedanca hyrëse ($ R_\text{in} \to \infty $) dhe impendaca dalëse ($ R_\text{out} \to 0 $)**
    Për një **op-amp ideal**, **impedanca hyrëse** është e pafundme $ R_\text{in} = \infty $ ndërsa ajo dalëse zero $ R_\text{out} = 0 $. Për një **op-amp real**, impedanca hyrëse është e lartë por **e fundme**, zakonisht $10^6\, \Omega$ (p.sh. op-amp bipolar) deri $10^{12} \, \Omega$ (p.sh. op-amp me hyrje JFET). Ndersa impendanca dalese varion midis $ R_\text{out} \approx 10 \text{ deri } 1000\, \Omega $.

3. **Rregulli 3: Rryma hyrëse në terminalet e op-amp-it është zero**
   Në një **op-amp ideal**, terminalet hyrëse **nuk tërheqin asnjë rrymë** $ I^+ = I^- = 0 $. Në praktikë, kjo është **e vërtetë** sepse rymat hyrëse janë zakonisht shumë të vogla: **~pA** (p.sh. në op-amp me hyrje JFET) dhe **~nA** (p.sh. në op-amp bipolar) Prandaj në shumicën e analizave, këto rryma **injorohen pa gabim të madh**.

4. **Rregulli 4: Sjellja e op-amp-it me feedback negativ ($ V^+ = V^-$)**
    Një op-amp me feedback negativ **ndjen një diferencë tensioni** midis hyrjes jo-invertuese $V^+$ dhe hyrjes invertuese $V^-$, dhe **reagon** duke **dërguar aq sa duhet tension ose rrymë nëpër rrjetin e feedback-ut**, në mënyrë që të **eliminojë këtë diferencë** $ V^+ - V^- = 0 $. Në këtë rast, sistemi krijon një **ekuilibër dinamik**, ku op-amp-i “detyron” veten që hyrjet të jenë me potencial pothuajse të njëjtë.

</div>

Këto rregulla janë themeli i analizës së të gjitha qarqeve me op-amp në regjim linear — nga amplifikatorët deri te filtrat dhe integratorët etj.

### **Si funksionon feedback në praktikë?**

#### Shembull 1: **Amplifikator jo-invertues**

<img src="figs/lec8/Op-Amp_Non-Inverting_Amplifier.svg" width="40%" style="background-color:#fff;" >

Në këtë konfigurim, hyrja vendoset në $V^+ =V_{in}$ dhe një pjesë e daljes kthehet te $V^-$ me një ndarës tensioni.
$$ V^- = \frac{R_1}{R_1+R_2}\,V_{out}= \beta\,V_{out} $$

Nga formula e amplifikimit te op-amp kemi
$$
V_{out} = A_0\, (V^+ - V^-) = A_0\, (V_{in} - \beta\,V_{out}) = \left( \frac{1}{\beta + \frac{1}{A_0}} \right)\, V_{in}
$$

Meqë $A_0$ është shumë e madhe, atëherë $1/A_0\approx 0$ rrjedhimisht

$$
V_{out} \approx \frac{V_{in}}{\beta} = \left( 1 + \frac{R_2}{R_1} \right)\, V_{in}
$$

**Koeficienti i amplifikimit:**
$$
A_\text{closed-loop} = 1 + \frac{R_2}{R_1}
$$

Përfitimi: Mund të krijosh amplifikator me **amplifikim të qëndrueshëm**, që përcaktohet vetëm nga **rezistencat** dhe jo nga karakteristikat e op-amp.



### Shembull 2: **Amplifikator invertues**

<img src="figs/lec8/Op-Amp_Inverting_Amplifier.svg" width="40%" style="background-color:#fff;" >

Hyrja vendoset në $V^-$, ndërsa $V^+=0$ lidhet në tokë si në figurë.

$$ V^- = \frac{1}{R_f+R_{in}}\,(R_f\,V_{in} + R_{in}\,V_{out})$$

Nga formula e amplifikimit të op-amp kemi
$$
V_{out} = A_0\, (V^+ - V^-) = -A_0\, V^- = -V_{in}\,\frac{A_0\,R_f}{R_f+R_{in}(1+A_0)}
$$

Sërisht, meqë $A_0\to \infty$ është shumë i madh, kemi
$$
V_{out} \approx -V_{in}\,\frac{R_f}{R_{in}}
$$

**Koeficienti i amplifikimit:**
$$
A_\text{closed-loop} = - \frac{R_f}{R_{in}}
$$

Përfitimi: Jep **faze të përmbysur**, kontroll të saktë të amplifikimit, dhe mund të përdoret për **mbledhje tensionesh**.



### **Për çfarë është i dobishëm një op-amp?**

| Funksioni                       | Përshkrimi                                           | Përdorim                                 |
| ------------------------------- | ---------------------------------------------------- | ---------------------------------------- |
| **Amplifikator**                | Rrit sinjale shumë të dobëta                         | Sensore, mikrofona                       |
| **Bufer (follower)**            | Ruajtje tensioni me impedancë të ulët dalje          | Ndërfaqe sinjalesh delikate              |
| **Filtra aktive**               | Filtrim i sinjaleve sipas frekuencës                 | Audio, sensorë                           |
| **Krahasues**                  | Vendos nëse një sinjal është më i madh se një tjetër | Sisteme alarmi, zero-crossing            |
| **Integrues / Diferencues** | Funksione matematikore analoge                       | Kontroll automatik, përpunim sinjali      |
| **Mbledhës tensionesh**      | Mbledh disa sinjale hyrëse në një dalje           | Mix audio, instrumente               |
| **Oshilatorë**                  | Gjenerues sinjali                      | Gjenerator sinjali, psh muzikë elektronike |


### **Pse përdoret kaq shumë?**

* I **lirë** dhe shumë i disponueshëm
* I **parashikueshëm** dhe i **lehtë për t'u modeluar**
* Mund të bëjë shumë funksione të ndryshme me ndryshime të thjeshta në qark
* Zë **pak vend** dhe është i integrueshëm në çdo sistem elektronik


### **Përmbledhje**
Një op-amp është një pajisje që përdor të gjithë fuqinë që ka në dispozicion nga burimi që të barazojë tensionet në dy hyrjet e tij përmes feedback-u.
*Op-amp-i nuk e përforcon vetë sinjalin hyrës — ai përdor energjinë nga burimi i furnizimit për të ndërtuar një sinjal të ri që imiton hyrjen me një amplifikim të kontrolluar.*

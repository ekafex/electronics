Kjo diagramë paraqet katër elementët themelorë dy-terminalësh në elektronikë: **rezistenca**, **kondensatori**, **bobina** dhe **memristori**. Këto elementë lidhin katër madhësi elektrike bazë: **tensionin ($v$)**, **rrymën ($i$)**, **ngarkesën ($q$)** dhe **fluksin magnetik ($\varphi$)**.

**Lidhjet midis madhësive elektrike për secilin element:**

1. **Rezistenca (R):**
   - **Lidhja:** Tension - rrymë
   - **Ekuacioni:** $v = R \cdot i $

2. **Kondensatori (C):**
   - **Lidhja:** Tension - ngarkesë
   - **Ekuacioni:** $q = C \cdot v $

3. **Bobina (L):**
   - **Lidhja:** Rrymë - fluks magnetik
   - **Ekuacioni:** $\varphi = L \cdot i $

4. **Memristori (M):**
   - **Lidhja:** Ngarkesë - fluks magnetik
   - **Ekuacioni:** $\varphi = M(q) \cdot q $

Në këtë leksion do merremi me qarqet që përmbajnë R, L dhe C. Duhet te dini qe ekziston nje element i katert pasiv qe quhet memristor. Memristori fillimisht u parashikua si një element që plotëson lidhjen e munguar midis ngarkesës dhe fluksit magnetik, duke e bërë atë elementin e katërt themelor në teorinë e qarqeve elektrike. Ndërsa rezistenca, kondensatori dhe bobina janë përdorur gjerësisht në praktikë, memristori është një koncept më i ri me aplikime potenciale në fushën e memories dhe sistemeve neuromorfike.

Memristori është parashikuar në vitin 1971 nga Leon Chua dhe i realizuar eksperimentalisht për herë të parë rreth vitit 2008 nga HP Labs. Emri vjen nga bashkimi i fjalëve **"memory"** dhe **"resistor"**, pra rezistencë me memorie.

#### **Karakteristikat fizike të memristorit**

- **Rezistenca e tij ndryshon në varësi të historikut të rrymës** që ka kaluar nëpër të. Ai “mban mend” se sa ngarkesë ka kaluar.
- Nuk kërkon energji për të ruajtur gjendjen e vet – pra është **jo-fluturues (non-volatile)**.
- Nëse e ndërpret qarkun dhe e rindez më vonë, **ruan gjendjen rezistive të mëparshme**.

#### **Sjellje e thjeshtë e memristorit**

Në mënyrë të thjeshtë:

- Nëse kalon rrymë në një drejtim, **rezistenca zvogëlohet**.
- Nëse kalon rrymë në drejtimin e kundërt, **rezistenca rritet**.
- Kur ndalon rryma, **rezistenca ngel konstante** – pra komponenti ka “kujtesë”.

#### **Zbatime dhe potenciale**

1. **Memorie jo-fluturues (non-volatile memory)** memorie e qëndrueshme (alternative ndaj flash memory).
2. **Rrjete neuromorfike** – memristorët imitojnë sinapset në trurin biologjik.
3. **Kompjuter analoge** – ku gjendja e sistemit përfaqëson zgjidhjen e një problemi.
4. **Hardware për Machine Learning** – realizim fizik i "weight matrices" (disa startups që po zhvillojnë këtë teknologji këto 5-vitet e fundit për të kursyer energji dhe për të bërë AI më eficient.
5. etj...


#### **Pse është interesant në elektronikë dhe fizikë**

- Memristori **shton një dimension të ri** në analizën e qarqeve – veçanërisht kur përdorim sisteme **jo-lineare** dhe me **memorie**.
- Është një **komponent jo-linear** dhe që **varion në kohë** (time-variant), i cili sfidon modelet klasike të analizës së qarqeve.

Ref: [HP 2008 Nature paper](https://web.ece.ucsb.edu/~strukov/papers/2008/Nature2008.pdf), [Wikipedia](https://en.wikipedia.org/wiki/Memristor)

# RLC

## **Kondensatori**
Kondensatorët janë elementë që **ruajnë energji në fushën elektrike**. Ato janë si “depo tensioni”, që rezistojnë ndaj ndryshimeve të shpejta të tensionit.

- Në qark, ato **bllokojnë DC** dhe **kalojnë AC**, në varësi të frekuencës.
- Përdoren për **filtrim** (p.sh. në qarqe furnizimi për të hequr valëzimet), **kuplim sinjali** (midis dy nivele tensioni), **integrues** dhe **kondensatorë të kthimit në zero**.
- Sjellja e tyre përshkruhet nga ligji:
  $$ i(t) = C \frac{dv(t)}{dt} $$ 
  ku çdo ndryshim në tension shkakton një rrymë.

## **Bobina**
Bobinat (induktorët) janë komponentë që **ruajnë energji në fushën magnetike**. Janë si “depo rryme”, që rezistojnë ndaj ndryshimeve të shpejta të rrymës.

- **Kalojnë DC** por **bllokojnë AC**, për frekuenca të larta.
- Janë të dobishme në filtrat e furnizimeve, rezonatorë, transformatorë dhe qarqe rezonante.
- Sjellja e tyre përshkruhet nga:
  $$ v(t) = L \frac{di(t)}{dt} $$
  ku ndryshimet në rrymë gjenerojnë tension.

## Qarqet me Kondensatorë

### Ngarkimi dhe shkarkimi i një kondensatori të vetëm

#### 1. **Modeli fizik i kondensatorit real**

Një **kondensator ideal** ruan ngarkesën pa asnjë humbje. Por në realitet, çdo kondensator ka **rrjedhje të ngarkesës** me kalimin e kohës.

Prandaj modeli i tij **real** është:

- **Kondensator ideal** $C$
- **Rezistencë parazitare** $R_p$ **në paralel**, e quajtur edhe **rezistenca e rrjedhjes (leakage)**

**Qëllimi i $R_p$**: përfaqëson humbjet për shkak të materialit dielektrik apo rrymave të rrjedhjes nëpër izolator.


#### 2. **Ngarkimi i kondensatorit**

**Qarku**: Një burim DC $V_0$, një rezistencë $R$, dhe një kondensator $C$ në seri.

**Ekuacioni diferencial**:
$$
V_0 = v_C(t) + R \cdot i(t), \quad i(t) = C \cdot \frac{dv_C(t)}{dt}
$$

**Zgjidhja**:
$$
v_C(t) = V_0 \left(1 - e^{-t/RC} \right)
\quad
i(t) = \frac{V_0}{R} e^{-t/RC}
$$

**Interpretim fizik**:
- Në $t = 0$, kondensatori është bosh, rryma është maksimale.
- Me kohën, tensioni mbi kondensatorin rritet dhe rryma bie.
- Në $t \to \infty$, kondensatori arrin $V_0$ dhe rryma bëhet zero.


#### 3. **Shkarkimi i kondensatorit – roli i $R_p$**

Në mungesë të një rruge të jashtme për shkarkim, kondensatori **shkarkohet përmes $R_p$**.

**Ekuacioni për shkarkim natyror**:
$$
v_C(t) = V_0 e^{-t/(R_p C)}
$$

**Interpretim**:
- Në kondensatorët me $R_p $ shumë të madh (p.sh. 10 MΩ), shkarkimi ndodh **ngadalë**.
- Ky është efekti që vëzhgojmë te **kondensatorët elektrolitikë (film ose qeramike)** që ruajnë ngarkesën për minuta apo orë.

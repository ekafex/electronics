```python
# per te vizatuar qarqet elekctrike po perdor SchemDraw,
# shiko ketu https://schemdraw.readthedocs.io/en/latest/index.html

import schemdraw
import schemdraw.elements as elm
from schemdraw import pictorial
elm.style(elm.STYLE_IEC)
```

```python
with schemdraw.Drawing() as d:
    U=elm.SourceV().up().label(r"$U$",loc='bottom')
    elm.Line().right()
    elm.CurrentLabelInline().label(r'$I$',loc='bottom')
    elm.Resistor().down().label(r"$R$",loc='top')
    elm.Line().left()
    elm.Dot(open=True).at(U.end).label(r'$A$',loc='top')
    elm.Dot(open=True).at(U.start).label(r'$B$',loc='bottom')
    d.save('figs/fig1.svg',transparent=False)
```

```python
with schemdraw.Drawing() as d:
    elm.Resistor()#.label('Rezistenca',)
    d.move(dx=0.2*d.unit)
    elm.ResistorIEEE()
    d.save('figs/R.svg',transparent=False)

with schemdraw.Drawing() as d:
    elm.Capacitor()
    d.move(dx=0.2*d.unit)
    elm.Capacitor2()
    d.save('figs/C.svg',transparent=False)

with schemdraw.Drawing() as d:
    elm.Inductor()
    d.move(dx=0.2*d.unit)
    elm.Inductor2()
    d.save('figs/L.svg',transparent=False)

with schemdraw.Drawing() as d:
    elm.SourceI().right()
    d.move(dx=0.2*d.unit)
    elm.SourceControlledI().right()
    d.save('figs/Is.svg',transparent=False)

with schemdraw.Drawing() as d:
    elm.SourceV().right()
    d.move(dx=0.2*d.unit)
    elm.SourceControlledV().right()
    d.save('figs/Us.svg',transparent=False)

with schemdraw.Drawing() as d:
    bjt=elm.BjtPnp(circle=True).flip().label('BJT',loc='right').label('baza',loc='base').label('kolektori',loc='collector').label('emiteri',loc='emitter')
    d.move_from(bjt.collector, dx=1.2*d.unit)
    elm.JFetP(circle=True).reverse().label('JFet',loc='right').label('porta',loc='gate').label('burimi',loc='source').label('mbledhësi',loc='drain')
    d.save('figs/BJT_FET.svg',transparent=False)
```

```python
with schemdraw.Drawing() as d:
    elm.SourceV().up().label(r'$U_0$',ofst=(0.6,-1.3))#.fill('lightgray')
    elm.Line().up().length(1.5)
    elm.Line().right()
    elm.CurrentLabelInline().label(r'$I$',loc='bottom')
    R1 = elm.Resistor().label(r'$R_1$').down()
    R2 = elm.Resistor().label(r'$R_2$').down()
    elm.Line().at(R1.start).right()
    R4 = elm.Resistor().label(r'$R_4$').down()
    elm.Resistor().label(r'$R_5$').down()
    elm.Line().left().length(6)
    elm.Line().up().length(1.5)
    elm.Resistor().label(r'$R_3$').at(R1.end).right()
    elm.Dot().at(R1.start).label('a',loc='top')
    elm.Dot().at(R1.end).label('b',loc='left')
    elm.Dot().at(R2.end).label('c',loc='bottom')
    elm.Dot().at(R4.end).label('d',loc='right')
    d.save('figs/circ1.svg',transparent=False)
```

```python
with schemdraw.Drawing() as d:
    elm.SourceV().up().label(r'$U_0$',ofst=(0.6,-1.3))#.fill('lightgray')
    elm.Line().up().length(1.5)
    elm.Line().right()
    elm.CurrentLabelInline().label(r'$I_0$',loc='bottom')
    R1 = elm.Resistor().down().label(r'$R_1$',loc='top')
    elm.CurrentLabelInline().at(R1).label(r'$I_1$',loc='bottom')
    elm.CurrentLabelInline(direction='out',start=False).at(R1).label(r'$I_1$',loc='bottom')
    R2 = elm.Resistor().down().label(r'$R_2$',loc='top')
    elm.CurrentLabelInline().at(R2).label(r'$I_2$',loc='top')
    elm.CurrentLabelInline(direction='out',start=False).at(R2).label(r'$I_2$',loc='top')
    elm.Line().at(R1.start).right()
    R4 = elm.Resistor().down().label(r'$R_4$',loc='top')
    elm.CurrentLabelInline().at(R4).label(r'$I_4$',loc='bottom')
    elm.CurrentLabelInline(direction='out',start=False).at(R4).label(r'$I_4$',loc='bottom')
    R5 = elm.Resistor().down().label(r'$R_5$',loc='top')
    elm.CurrentLabelInline().at(R5).label(r'$I_5$',loc='bottom')
    elm.CurrentLabelInline(direction='out',start=False).at(R5).label(r'$I_5$',loc='bottom')
    elm.Line().left().length(6)
    elm.CurrentLabelInline().label(r'$I_0$',loc='top')
    elm.Line().up().length(1.5)
    R3 = elm.Resistor().right().at(R1.end).label(r'$R_3$',loc='top')
    elm.CurrentLabelInline().at(R3).label(r'$I_3$',loc='bottom')
    elm.CurrentLabelInline(direction='out',start=False).at(R3).label(r'$I_3$',loc='bottom')
    elm.Dot().at(R1.start).label('a',loc='top')
    elm.Dot().at(R1.end).label('b',loc='left')
    elm.Dot().at(R2.end).label('c',loc='bottom')
    elm.Dot().at(R4.end).label('d',loc='right')

    d.move(dx=1*d.unit)
    #elm.Arrow(length=0.3)
    elm.Label(r'$\Longleftarrow$ KCL dhe KVL $\Longrightarrow$')
    d.move(dx=1.2*d.unit,dy=-0.5*d.unit)
    
    elm.SourceV().up().label(('–',r'$U_0$','+'),loc='top')#.fill('lightgray')
    elm.Line().up().length(1.5)
    elm.Line().right()
    R1a = elm.Resistor().down().label(('+',r'$U_1$','–'))
    R2a = elm.Resistor().down().label(('+',r'$U_2$','–'))
    elm.Line().at(R1a.start).right()
    R4a = elm.Resistor().down().label(('+',r'$U_4$','–'),loc='bottom')
    elm.Resistor().down().label(('+',r'$U_5$','–'),loc='bottom')
    elm.Line().left().length(6)
    elm.Line().up().length(1.5)
    elm.Resistor().right().at(R1a.end).label(('+',r'$U_3$','–'))
    elm.Dot().at(R1a.start).label('a',loc='top')
    elm.Dot().at(R1a.end).label('b',loc='left')
    elm.Dot().at(R2a.end).label('c',loc='bottom')
    elm.Dot().at(R4a.end).label('d',loc='right')
    elm.LoopArrow(direction='cw',width=0.8,height=3.5).at((4.7*d.unit,0.5*d.unit)).label(r'$L_1$')
    elm.LoopArrow(direction='cw',width=0.8,height=1.5).at((5.6*d.unit,1.1*d.unit)).label(r'$L_2$')
    elm.LoopArrow(direction='cw',width=0.8,height=1.5).at((5.6*d.unit,-0.1*d.unit)).label(r'$L_3$')
    
    d.save('figs/circ1_kirch.svg',transparent=False)
```

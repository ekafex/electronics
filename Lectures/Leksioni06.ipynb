## **Historku: Diodat e Vakumit**

Diodat me vakum (shpesh të quajtura edhe “valvola” në gjuhën popullore) janë paraardhëset e para teknologjike të diodave gjysmëpërçuese që njohim sot. Ato u zhvilluan rreth fillimit të shekullit XX dhe ishin thelbësore në elektronikën e hershme, para epokës së transistorëve. Më poshtë një pasqyrë e shkurtër:

1. **Struktura Themelore:**
   - Në thelb, një diodë me vakum përbëhet nga një **katodë** (e nxehtë) dhe një **anodë** (pllakë metalike) brenda një vakumi hermetik.
   - Kur katoda nxehet (zakonisht me një filament), ajo lëshon elektrone (efekti termoelektronik). Ato elektrone mund të lëvizin drejt anodes nëse ajo është në potencial pozitiv në krahasim me katodën, duke mundësuar rrjedhjen e rrymës vetëm në një drejtim.

2. **Parimi i Funksionimit:**
   - Në rast se aplikohet një diferencë potenciale e drejtë (+ në anodë dhe - në katodë), elektronet e prodhuara në katodë mund të lëvizin pa fërkim/rezistencë në tub drejt anodes për shkak të mungesës së molekulave/atomeve (vakumit).
   - Në se aplikohet një diferencë potencial në kahun e kundërt (- në anodë dhe + në katodë), fusha elektrike brënda tubit pengon elektronet të lëvizin drejt anodës, si rrjedhojë bllokohet rrjedhja e rrymës.

3. **Rëndësia Historike:**
   - Dioda me vakum ishte komponenti i parë që bënte drejtimin e rrymës (rectification) për sinjale AC (rryma elektrike alternative).
   - Ajo hapi rrugën për zbulimin e tubave të tjerë elektronikë, si triodat, tetrodat, pentodat, etj., të cilat u përdorën gjerësisht në radiokomunikacion, amplifikim, dhe më vonë në sistemet televizive.

4. **Kufizimet Kryesore:**
   - Konsum i madh energjie, pasi filamenti duhej të nxehej vazhdimisht.
   - Pajisje të brishta dhe me dimensione relativisht të mëdha.
   - U zëvendësuan kryesisht nga diodat gjysmëpërçuese, që janë më efikase, më të vogla dhe nuk kanë nevojë për ngrohje.

## **Diodat Gjysmëpërçuese**

Le të hyjmë tani në **diodat gjysmëpërçuese** dhe t’i hedhim një vështrim bazave të tyre. Në këtë pikë, është e rëndësishme të kuptojmë njëherazi si përgatitet materiali gjysmëpërçues (silici, germaniumi) dhe si formohet kalimi p-n. Kjo do të na mundësojë të shpjegojmë funksionimin kryesor të diodës si “valvul elektrike” njëdrejtimëshe.


## 1. Struktura e Materialit Gjysmëpërçues

- **Gjysmëpërçuesit e pastër (intrinsic)**: Në kushtet e temperaturës së dhomës, silici ose germaniumi ka një numër mjaft të vogël elektronësh (dhe vrimash) të lirë. Përcjellshmëria e tyre është e kufizuar.
- **Gjysmëpërçuesit të dopuar (extrinsic)**: Për ta rritur përcjellshmërinë, në strukturën e silicit shtohen atome “papastërtish”:
  - **Tipi n**: Doping me elementë pentavalentë (p.sh. fosfor). Këta atome sjellin **elektrone shtesë**.  
  - **Tipi p**: Doping me elementë trivalentë (p.sh. bor). Këta atome krijojnë **vrima shtesë** (mungesë e një elektroni).

## 2. Formimi i Kalimit p-n

Në momentin kur **dy rajone** (n dhe p) bashkohen:

1. **Difuzioni i transportuesve**: Elektronet nga rajoni n tentojnë të kalojnë në rajonin p, ndërsa vrimat nga rajoni p shkojnë drejt rajonit n.  
2. **Zonë e varfërimit (Depletion Region)**: Krijohet një zonë e mesme ku elektronet dhe vrimat kanë “neutralizuar” njëra-tjetrën, duke lënë pas jone fikse të ngarkuara (negative në anën p dhe pozitive në anën n).  
3. **Tensioni ndërtues (Built-in Voltage)**: Zona e varfërimit shoqërohet me një diferencë potenciali të brëndshme, e cila pengon difuzionin e bartësave më tej. Ky tension është rreth 0.6-0.7V për diodën e silicit dhe rreth 0.3V për atë të germaniumit.



## 3. Karakteristika e Polaritetit të Tensionit (Forward/Reverse Bias)

- **Polarizimi i drejtë (Forward Bias)**: Anoda (p) lidhet në potencial pozitiv, katoda (n) në negativ. Kjo redukton barrierën e zonës së varfërimit, duke lejuar rrymën të rrjedhë.
- **Polarizimi i kundërt (Reverse Bias)**: Anoda (p) lidhet në potencial negativ, katoda (n) në pozitiv. Zona e varfërimit zgjerohet, duke penguar rrjedhjen e rrymës (përveç një rryme shumë të vogël inverse).

## 4. Karakteristika V–I

Për një diodë tipike silici, kur aplikojmë **tension përpara**:

- Deri në rreth **0.7 V** (për silic) rryma është mjaft e vogël.  
- Mbi këtë prag (tensioni i barrierës për silic ~0.7 V), rryma rritet shpejt (ekspotencialisht).  

Në **polarizim të kundërt**, rryma mbetet pothuajse konstante dhe shumë e vogël (rryma e ngopjes), derisa mund të arrihet një pikë “thyerjeje” (breakdown) në tensione të larta ku dioda mund të dëmtohet (ose të përdoret qëllimisht, si te diodat Zener).

# **Diodat Gjysmëpërçuese**

## 1. Parimi Themelor i Diodës

Një **diodë gjysmëpërçuese** është një komponent elektronik që **lejon rrymën të rrjedhë në një drejtim**, por e bllokon në drejtimin tjetër. Kjo sillet për shkak të **strukturës p-n**, e cila krijon një **zonë varfërimi** (depletion zone) ku nuk ka transportues të lirë ngarkese.

### Struktura p-n
- Rajoni **n** ka elektrone të tepërta (doping me elementë pentavalentë).
- Rajoni **p** ka vrima të tepërta (doping me elementë trivalentë).
- Kur bashkohen, elektronet dhe vrimat difuzojnë dhe krijojnë një **zonë pa ngarkesa të lira**, e cila paraqet një **barrierë potenciali** (~0.6–0.7 V për silic).


## 2. Kurba V–I dhe **Ligji i Shockley**

#### Lidhje e Drejtë e Diodës ($V>0$)

Sic dhe e pame me lart, aplikimi i një tensioni të drejtë mbi diodë krijon një fushë elektrike brënda që kundërshton ate të zonës së varfëruar në kalimin p-n. Kjo ngushton zonën e varfëruar dhe lehtëson rrjedhjen e rrymës së difuzionit që ishte bllokuar me parë. Observohet se rryma që rrjedh në diodë rritet **eksponencialisht** me tensionin e aplikuar dhe përshkruhet nga ligji Shockley që shkruhet:
$$
I(V) = I_S \left(e^{\frac{V}{nV_T}} - 1\right)
$$

ku:
- $I_S$ është **rryma e ngopjes (leakage current)** e kundërt (shumë e vogël, < 1 nA).
- $n$ është faktori idealitetit (1-2). Kur $n\approx 1$ dioda sillet si diodë ideale.
- $V_T = \frac{kT}{q} \approx 25 mV$ në temperaturë dhome.

#### Lidhje e Kundërt e Diodës ($V<=$)

Lidhja me polaritet të kundërt shkakton një fushë të brëndeshme që i mbivendoset asaj të zonës së varfëruar të kalimit p-n duke rritur gjërësinë e kësaj të fundit. Kjo pengon edhe me shumë rrjedhjen e rrymës përvec rrymes së dobët të ngopjes $I_S$.

#### Zona e Prishjes së Qëndrueshmërisë (Breakdown Region)

Për tensione të kundërta mjaftueshmërisht të mëdha, dioda pëson:
- **Efektin Zener** Zener Breakdown (për dioda shumë të dopuara, për shkak të efektit të tunelimit kuantik).
- **Efektit Ortek** Avalanche Breakdown (për dioda pak të dopuara, për shkak të shumëfishimit të mbartësave të lirë të ngargesës).



## 3. Diodat më të Rëndësishme dhe Fizika e Tyre

| Dioda         | Struktura | Doping-u | Karakteristika | V–I karakteristikë (grafiku) |
|---------------|-----------|----------|----------------|------------------------------|
| **1N4148**    | Silic p–n | moderuar | Komutim i shpejtë, tension ~0.7 V | Rritje eksponenciale në forward, bllokim në reverse |
| **Zener**| p–n shumë i dopuar | shumë i lartë | Prishje e qëndrueshme në polarizim të kundërt | Prishje në tensionin Zener, e përdorur për stabilizim tensioni |
| **LED (e kuqe)** | p–n me GaAs | band gap direkt ~1.9 eV | Emeton dritë në rekombinim, tension përpara ~1.8–2.2 V | V–I e ngjashme me diodën p–n por tension më i lartë |
| **Schottky**  | Metal–n | pa p-rajone | Barrierë më e ulët, komutim shumë i shpejtë | Rënie e vogël në tension (0.2–0.4 V), prishje më e hershme në reverse |
| **Tunnel**    | p–n shumë i dopuar | ekstremisht i lartë | Tunelim kuantik, rajon me rezistencë negative | Kurba jo-monotone me kulm në rrymë dhe pastaj rënie |


## 4. Grafikët e Karakteristikës V–I

Në grafikun që ke parë më sipër, shfaqen këto tipare **(shiko figurën më poshtë)**:

- **1N4148 (silic)**: rënie tipike ~0.7 V, rritje eksponenciale.
- **Zener**: sillet si diodë normale përpara, por në polarizim të kundërt ka një pikë “prishjeje” të kontrolluar (~5.1 V).
- **LED (e kuqe)**: rënie më e lartë në tension përpara, pasi materiali ka band gap më të madh (shih GaAs).
- **Schottky**: tension përpara më i vogël (~0.3 V), ideale për efikasitet të lartë.
- **Tunnel diode**: ka një **rajon me rezistencë negative** në forward bias, ku rryma bie me rritjen e tensionit (veçori e unike).



````python
import numpy as np
import matplotlib.pyplot as plt

# Tensionet përkatëse për çdo diodë
V = np.linspace(-1.5, 1.5, 400)

# Karakteristika V-I për dioda të ndryshme
def diode_iv(V, Is=1e-12, n=1, VT=0.025):
    return Is * (np.exp(V / (n * VT)) - 1)

# Parametra për disa lloje diodash
curves = {
    "Silicon (1N4148)": {"Is": 1e-12, "n": 1, 'col':'k'},
    "Zener 5.1V": {"Is": 1e-12, "n": 1, 'col':'b'},
    "LED (red)": {"Is": 1e-12, "n": 2, 'col':'r'},
    "Schottky": {"Is": 1e-10, "n": 1, 'col':'g'},
    "Tunnel": None  # trajtohet veçmas më poshtë
}

# Inicializimi i figurës
plt.figure(figsize=(10, 6))

# Vizato të gjitha karakteristikat përveç tunnel
for label, params in curves.items():
    if params is not None:
        I = diode_iv(V, Is=params["Is"], n=params["n"])
        plt.plot(V, I*1e3, lw=2, color=params["col"], label=label)  # rryma në mA

# Dioda Tunnel (me rezistencë negative)
V_tunnel = np.linspace(-0.6, 0.6, 400)
I_tunnel = 20 * (V_tunnel * np.exp(-100 * (V_tunnel - 0.1)**2))
plt.plot(V_tunnel, I_tunnel, '--c', lw=4, label="Diode Tunel")

# Përshtatje vizuale
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Karakteristikat V-I për Dioda të Ndryshme")
plt.xlabel("Tensioni (V)")
plt.ylabel("Rryma (mA)")
plt.grid(True)
plt.legend()
plt.ylim(-5, 30)
plt.xlim(-1.5, 1.5)
plt.tight_layout()
plt.show()
````

## 1. **Si modelohet një diodë në praktikë?**

### a) **Modeli ideal** (për analizë konceptuale)

Dioda sillet si:
- **Përçues perfekt** kur është në **forward bias** dhe $V_D > 0$.
- **Izolues perfekt** në **reverse bias**, kur $V_D < 0$.

$$
I = \begin{cases}
\text{përcjell çfarëdo rryme} & V_D > 0 \\
0 & V_D \leq 0
\end{cases}
$$

### ⚙️ b) **Modeli me prag (threshold model)**

Përdor një **tension të fiksuar të përçueshmërisë**, p.sh. $V_\text{on} = 0.7\ \text{V}$ për silic dhe $V_\text{on} = 0.3\ \text{V}$ për germanium:

- Kur $V_D < V_\text{on}$ → dioda është e mbyllur (dioda modelohet si qark i hapur).
- Kur $V_D \geq V_\text{on}$ → dioda hapet dhe sillet si përçues (me rënie tensioni fikse 0.7 V). Dioda modelohet si thjeshte nje percjelles por me nje renie potenciali $V_\text{on}$.

🔧 **Modeli i përdorur në shumicën e analizave manuale** (analizë DC, qark klasik, clipper, rregullator, etj.)



### c) **Modeli eksponencial (Shockley model)**

$$
I_D = I_S \left( e^{\frac{V_D}{n V_T}} - 1 \right)
$$

ku:
- $I_S$ = rryma e ngopjes (e vogël, zakonisht $\sim 10^{-12}$ A).
- $V_T$ = tensioni termik ($\approx 25.9$ mV në $25^\circ$C).
- $n$ = faktori idealitetit (1–2).

**Ky model përdoret në simulime numerike si SPICE.** Jep sjellje shumë të saktë për tensionin dhe rrymën.


### d) **Modeli i qarkut me sinjale të vogla**

Përdoret për **analiza sinjalesh të vogjel rreth pikës së punës DC**.

- Supozohet një **rezistencë dinamike**:

$$ r_d = \frac{n V_T}{I_D} $$

- Modelet përfshijnë ndonjëherë edhe **kapacitete nyje** ($C_j$) dhe **rezistencë në seri $r_s$**.


## 2. **Si analizohen diodat në qark?**

### Metoda klasike: **Supozo dhe verifiko**

1. Supozojmë që dioda është **hapur (on)** ose **mbyllur (off)**.
2. Zëvendësojmë diodën në qark:
   - **Hapur** → zëvendësohet me nje **burim rënës tensioni** (psh 0.7 V për silicin).
   - **Mbyllur** → zëvendësohet me **qark të hapur**.
3. Zbatojmë **ligjin e Ohm-it, Kirchoff-it** për të gjetur rrymën.
4. **Verifikojmë** nëse supozimi është i vlefshëm:
   - Nëse dioda supozohet e hapur dhe rryma del negative → supozimi gabim → është e mbyllur.
   - Nëse supozohet ishte i mbyllur dhe tensioni del $>0.7$ V → gabim → është hapur.



### Shembull 1: Diodë me rezistencë dhe burim tensioni

```
(Vin) ──►─|>|─── R ──── GND
```

- P.sh. Vin = 5V, R = 1kΩ  
- Supozojmë që dioda është **hapur** ⇒ ka rënie 0.7V ⇒ tension në rezistencë: $5V - 0.7V = 4.3V$  
- Rryma: $I = 4.3V / 1k\Omega = 4.3$ mA  
- Rryma është pozitive ⇒ supozimi është i saktë.


### Shembull 2: Diodë clipper

```
Sinjal AC ──►─|>|─── GND
               D
```

- Kur sinjali është pozitiv dhe > 0.7 V, dioda **hapet** dhe tensioni klipohet në 0.7 V.  
- Kur sinjali është negativ, dioda **mbetet e mbyllur**, tensioni shkon drejt vlerës së sinjalit.

# Modelimi numerik e V-I

Grafiku më sipër tregon **karakteristikat realiste V–I** për:

### **Diodën tipike silici (1N4148)**

- **Rritje eksponenciale** e rrymës për tensione > 0.5 V.
- **Efekti i rezistencës serike** ($R_s = 10\ \Omega$) shfaqet si "butësi" në rritje për rrymat më të larta.
- Në polarizim të kundërt (reverse bias), rryma qëndron shumë afër zeros – nuk ndodh prishje (nuk arrijmë tensione të larta negative për këtë rast).


### **Dioda Zener (1N4733A, 5.1 V)**

- **Në polarizim të drejtë**, sillet si diodë e zakonshme, me rritje eksponenciale pas 0.6–0.7 V.
- **Në polarizim të kundërt**, rryma qëndron shumë e vogël derisa **tensioni negativ arrin vlerën -5.1 V** → pastaj hyn në loje **efekti Zener** dhe rryma rritet shpejt në kah të kundërt.
- Rryma Zener kufizohet në **50 mA**, duke simuluar kufijtë e lejuar fizikë.


## Çfarë mësojmë?

1. **Modelimi realist** jep sjellje të detajuar dhe të saktë në të dyja drejtimet.
2. **Zeneri sillet shumë ndryshe në polarizim të kundërt**, duke mbajtur tension pothuaj konstant pas një vlere.
3. Ky model është shumë i përshtatshëm për **simulim të qarqeve rregulluese** (p.sh. clamping dhe stabilizim tensioni).


````python
import numpy as np
import matplotlib.pyplot as plt

# Parametra për një diodë tipike silici (1N4148)
params_diode = {
    'Is': 1e-12,    # rryma e ngopjes (A)
    'n': 1.7,       # faktori i idealitetit
    'VT': 0.02585,  # tensioni termik në 300 K (V)
    'Rs': 10        # rezistenca serike (Ohm)
}

# Parametra për një Zener tipik (1N4733A, 5.1 V)
params_zener = {
    'Is': 1e-12,
    'n': 1.5,
    'VT': 0.02585,
    'Rs': 15,
    'Vz': 5.1,       # tensioni Zener (V)
    'Iz_max': 50e-3  # rryma maksimale në breakdown (A)
}

# Vargu i tensioneve: nga -6 V në 1 V
V = np.linspace(-6, 3, 1000)

# Modeli eksponencial me rezistencë serike për diodën normale
def diode_current(V, Is, n, VT, Rs):
    I = np.zeros_like(V)
    for i, v in enumerate(V):
        I_guess = 1e-12
        for _ in range(100):
            try:
                V_guess = I_guess * Rs + n * VT * np.log(I_guess / Is + 1)
                dVdI = Rs + n * VT / (I_guess + Is)
                I_guess -= (V_guess - v) / dVdI
                if I_guess < 0:
                    I_guess = 0
                    break
            except:
                I_guess = 0
                break
        I[i] = I_guess
    return I

# Modeli për Zener-in: përpara përdor Shockley, në të kundërt modelon breakdown-in në -Vz
def zener_current(V, params):
    Is = params['Is']
    n = params['n']
    VT = params['VT']
    Rs = params['Rs']
    Vz = params['Vz']
    Iz_max = params['Iz_max']

    I = np.zeros_like(V)
    for i, v in enumerate(V):
        if v >= 0:
            I[i] = diode_current([v], Is, n, VT, Rs)[0]
        else:
            if v > -Vz:
                I[i] = -Is  # rrjedhje e vogël para breakdown-it
            else:
                I[i] = (v + Vz) / Rs  # rënie në prishje (rrymë negative)
                if I[i] < -Iz_max:
                    I[i] = -Iz_max
    return I

# Llogaritja e rrymave
I_diode = diode_current(V, **params_diode)
I_zener = zener_current(V, params_zener)

# Vizatim
plt.figure(figsize=(10, 6))
plt.plot(V, I_diode * 1e3,'-k', lw=2, label='Diodë Silici (1N4148)')
plt.plot(V, I_zener * 1e3,'-r', lw=2, label='Diodë Zener (1N4733A 5.1V)')
plt.axvline(0, color='black', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("Tensioni $V_D$ (V)")
plt.ylabel("Rryma $I_D$ (mA)")
plt.title("Karakteristika V–I e Diodës Silici dhe Diodës Zener (Model i Përmirësuar)")
plt.grid(True)
plt.legend()
plt.ylim(-50, 80)
plt.xlim(-6, 3)
plt.tight_layout()
plt.show()

````

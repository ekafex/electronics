Kjo diagramë paraqet katër elementët themelorë dy-terminalësh në elektronikë: **rezistenca**, **kondensatori**, **bobina** dhe **memristori**. Këto elementë lidhin katër madhësi elektrike bazë: **tensionin ($v$)**, **rrymën ($i$)**, **ngarkesën ($q$)** dhe **fluksin magnetik ($\varphi$)**.

**Lidhjet midis madhësive elektrike për secilin element:**

1. **Rezistenca (R):**
   - **Lidhja:** Tension - rrymë
   - **Ekuacioni:** $v = R \cdot i $

2. **Kondensatori (C):**
   - **Lidhja:** Tension - ngarkesë
   - **Ekuacioni:** $q = C \cdot v $

3. **Bobina (L):**
   - **Lidhja:** Rrymë - fluks magnetik
   - **Ekuacioni:** $\varphi = L \cdot i $

4. **Memristori (M):**
   - **Lidhja:** Ngarkesë - fluks magnetik
   - **Ekuacioni:** $\varphi = M(q) \cdot q $

Në këtë leksion do merremi me qarqet që përmbajnë R, L dhe C. Duhet te dini qe ekziston nje element i katert pasiv qe quhet memristor. Memristori fillimisht u parashikua si një element që plotëson lidhjen e munguar midis ngarkesës dhe fluksit magnetik, duke e bërë atë elementin e katërt themelor në teorinë e qarqeve elektrike. Ndërsa rezistenca, kondensatori dhe bobina janë përdorur gjerësisht në praktikë, memristori është një koncept më i ri me aplikime potenciale në fushën e memories dhe sistemeve neuromorfike.

Memristori është parashikuar në vitin 1971 nga Leon Chua dhe i realizuar eksperimentalisht për herë të parë rreth vitit 2008 nga HP Labs. Emri vjen nga bashkimi i fjalëve **"memory"** dhe **"resistor"**, pra rezistencë me memorie.

#### **Karakteristikat fizike të memristorit**

- **Rezistenca e tij ndryshon në varësi të historikut të rrymës** që ka kaluar nëpër të. Ai “mban mend” se sa ngarkesë ka kaluar.
- Nuk kërkon energji për të ruajtur gjendjen e vet – pra është **jo-fluturues (non-volatile)**.
- Nëse e ndërpret qarkun dhe e rindez më vonë, **ruan gjendjen rezistive të mëparshme**.

#### **Sjellje e thjeshtë e memristorit**

Në mënyrë të thjeshtë:

- Nëse kalon rrymë në një drejtim, **rezistenca zvogëlohet**.
- Nëse kalon rrymë në drejtimin e kundërt, **rezistenca rritet**.
- Kur ndalon rryma, **rezistenca ngel konstante** – pra komponenti ka “kujtesë”.

#### **Zbatime dhe potenciale**

1. **Memorie jo-fluturues (non-volatile memory)** memorie e qëndrueshme (alternative ndaj flash memory).
2. **Rrjete neuromorfike** – memristorët imitojnë sinapset në trurin biologjik.
3. **Kompjuter analoge** – ku gjendja e sistemit përfaqëson zgjidhjen e një problemi.
4. **Hardware për Machine Learning** – realizim fizik i "weight matrices" (disa startups që po zhvillojnë këtë teknologji këto 5-vitet e fundit për të kursyer energji dhe për të bërë AI më eficient.
5. etj...


#### **Pse është interesant në elektronikë dhe fizikë**

- Memristori **shton një dimension të ri** në analizën e qarqeve – veçanërisht kur përdorim sisteme **jo-lineare** dhe me **memorie**.
- Është një **komponent jo-linear** dhe që **varion në kohë** (time-variant), i cili sfidon modelet klasike të analizës së qarqeve.

# RLC

## **Kondensatori**
Kondensatorët janë elementë që **ruajnë energji në fushën elektrike**. Ato janë si “depo tensioni”, që rezistojnë ndaj ndryshimeve të shpejta të tensionit.

- Në qark, ato **bllokojnë DC** dhe **kalojnë AC**, në varësi të frekuencës.
- Përdoren për **filtrim** (p.sh. në qarqe furnizimi për të hequr valëzimet), **kuplim sinjali** (midis dy nivele tensioni), **integrues** dhe **kondensatorë të kthimit në zero**.
- Sjellja e tyre përshkruhet nga ligji:
  $$ i(t) = C \frac{dv(t)}{dt} $$ 
  ku çdo ndryshim në tension shkakton një rrymë.

## **Bobina**
Bobinat (induktorët) janë komponentë që **ruajnë energji në fushën magnetike**. Janë si “depo rryme”, që rezistojnë ndaj ndryshimeve të shpejta të rrymës.

- **Kalojnë DC** por **bllokojnë AC**, për frekuenca të larta.
- Janë të dobishme në filtrat e furnizimeve, rezonatorë, transformatorë dhe qarqe rezonante.
- Sjellja e tyre përshkruhet nga:
  $$ v(t) = L \frac{di(t)}{dt} $$
  ku ndryshimet në rrymë gjenerojnë tension.

## Qarqet me Kondensatorë dhe Bobina

### Ngarkimi dhe shkarkimi i një kondensatori të vetëm

#### 1. **Modeli fizik i kondensatorit real**

Një **kondensator ideal** ruan ngarkesën pa asnjë humbje. Por në realitet, çdo kondensator ka **rrjedhje të ngarkesës** me kalimin e kohës.

Prandaj modeli i tij **real** është:

- **Kondensator ideal** $C$
- **Rezistencë parazitare** $R_p$ **në paralel**, e quajtur edhe **rezistenca e rrjedhjes (leakage)**

**Qëllimi i $R_p$**: përfaqëson humbjet për shkak të materialit dielektrik apo rrymave të rrjedhjes nëpër izolator.


#### 2. **Ngarkimi i kondensatorit**

**Qarku**: Një burim DC $V_0$, një rezistencë $R$, dhe një kondensator $C$ në seri.

**Ekuacioni diferencial**:
$$
V_0 = v_C(t) + R \cdot i(t), \quad i(t) = C \cdot \frac{dv_C(t)}{dt}
$$

**Zgjidhja**:
$$
v_C(t) = V_0 \left(1 - e^{-t/RC} \right)
\quad
i(t) = \frac{V_0}{R} e^{-t/RC}
$$

**Interpretim fizik**:
- Në $t = 0$, kondensatori është bosh, rryma është maksimale.
- Me kohën, tensioni mbi kondensatorin rritet dhe rryma bie.
- Në $t \to \infty$, kondensatori arrin $V_0$ dhe rryma bëhet zero.


#### 3. **Shkarkimi i kondensatorit – roli i $R_p$**

Në mungesë të një rruge të jashtme për shkarkim, kondensatori **shkarkohet përmes $R_p$**.

**Ekuacioni për shkarkim natyror**:
$$
v_C(t) = V_0 e^{-t/(R_p C)}
$$

**Interpretim**:
- Në kondensatorët me $R_p $ shumë të madh (p.sh. 10 MΩ), shkarkimi ndodh **ngadalë**.
- Ky është efekti që vëzhgojmë te **kondensatorët elektrolitikë (film ose qeramike)** që ruajnë ngarkesën për minuta apo orë.


<br>
<br>


### **Analizë e detajuar e qarkut RC**

**Qarku:** Një burim konstant DC me tension $V_0$, një rezistencë $R$ dhe një kondensator $C$ në seri.

##### **Ekuacion diferencial për qarkun RC:**
Nga ligji i Kirchhoffit për tensionet:
$$
V_0 = v_R(t) + v_C(t), \quad v_R(t) = R i(t), \quad i(t)=C \frac{dv_C}{dt}
$$

Kështu ekuacioni diferencial është:
$$
V_0 = RC \frac{dv_C}{dt} + v_C(t)
$$

##### **Zgjidhja:**
Për një kondensator fillimisht bosh, \(v_C(0)=0\), zgjidhja është:
$$
v_C(t) = V_0 (1 - e^{-t/RC}), \quad i(t) = \frac{V_0}{R} e^{-t/RC}
$$

<br>

## **Analizë e qarkut RL**

**Qarku:** Burim konstant DC $V_0$, rezistencë $R$, dhe bobinë $L$ në seri.

##### **Ekuacion diferencial për qarkun RL:**
Nga ligji i Kirchhoffit:
$$
V_0 = v_R(t) + v_L(t), \quad v_R(t) = Ri(t), \quad v_L(t)=L\frac{di}{dt}
$$

Kështu ekuacioni diferencial:
$$
V_0 = R i(t) + L \frac{di}{dt}
$$

##### **Zgjidhja:**
Me bobinë fillimisht pa rrymë $i(0)=0$:
$$
i(t) = \frac{V_0}{R}\left(1-e^{-Rt/L}\right), \quad v_L(t) = V_0 e^{-Rt/L}
$$

<br>


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrat
V0 = 5  # Volt
R = 1000  # Ohm
C = 1e-6  # Farad
tau = R*C
t = np.linspace(0, 0.01, 1000)

# Tensioni dhe rryma
v_C = V0*(1 - np.exp(-t/tau))
i = (V0/R)*np.exp(-t/tau)

# Grafikët
plt.figure(figsize=(10,5))
plt.plot(1e3*t, v_C,'-k',lw=2, label=r'Tensioni mbi kondensator ($V_C$)')
plt.plot(1e3*t, 1e3*i,'-r',lw=2, label=r'Rryma në qark ($i$)')
plt.xlabel('Koha [ms]')
plt.ylabel('Tensioni [V], Rryma [mA]')
plt.title('Ngarkimi i kondensatorit në qark RC')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Parametrat
V0 = 5  # Volt
R = 1000  # Ohm
L = 0.5  # Henry
tau = L/R
t = np.linspace(0, 0.01, 1000)

# Rryma dhe tensioni
i_L = (V0/R)*(1 - np.exp(-t/tau))
v_L = V0 * np.exp(-t/tau)

# Grafikët
plt.figure(figsize=(10,5))
plt.plot(1e3*t, v_L,'-k',lw=2, label=r'Tensioni në bobinë ($V_L$)')
plt.plot(1e3*t, 1e3*i_L,'-r',lw=2, label=r'Rryma në bobinë ($i$)')
plt.xlabel('Koha [ms]')
plt.ylabel('Tensioni [V], Rryma [mA]')
plt.title('Ngarkimi i bobinës në qark RL')
plt.grid(True)
plt.legend()
plt.show()
```

# **Lexim: Analiza e përgjithshme e ekuacioneve diferenciale. Koncepti i përgjithësuar i impedancës $Z$ (Impendanca e Laplace-it)**

#### **Hapi 1: Përdor ligjet e Kirchhoff-it**

Në analizën e qarqeve elektrike me elementë pasivë (**rezistencë R, kondensator C, bobinë L**) dhe burim tensioni $V(t)$ (jo detyrimisht AC), përdorim ligjet e Kirchhoff-it për të shkruar ekuacione diferenciale që përshkruajnë sjelljen e qarkut.

Për shembull, në qark RC më lart, gjetëm:
$$ V(t) = RC\frac{dv_C(t)}{dt} + v_C(t) $$

#### **Hapi 2: Përdor transformimin Laplace**

Transformimi i Laplace-it ($\mathcal{L}$) është një metodë e fuqishme që shndërron ekuacionet diferenciale në ekuacione algjebrike:

$$ \mathcal{L}\left\{\frac{dv_C(t)}{dt}\right\}=sV_C(s)-v_C(0),\quad \mathcal{L}\{V(t)\}=V(s) $$

Kemi përdorur në mënyrë eksplicite vetitë e transformimit Laplace ($s=\sigma +j\omega$):

- $\mathcal{L}\{\frac{dv_C}{dt}\}= sV_C(s)-v_C(0)$
- $\mathcal{L}\{V(t)\}=V(s)$


**Në rastin e qarkut RC me kushte fillestare jo zero pra kur $v_C(0)=V_{C0}$ (dmth kur kondensatori në castin $t=0$ ka voltazh $V_{C0}$)**, ekuacioni bëhet:

$$ V(s) = (1+RC\,s)V_C(s)-RC\,V_{C0} $$


#### **Hapi 3: Zgjidh ekuacioni algjebrik per kondensatorin**

Zgjidhim ekuacionin në mënyrë eksplicite për $V_C(s)$, gjithashtu shënojmë me $\tau=R\, C$:

$$ V_C(s) = \frac{V(s)+\tau\,V_{C0}}{1+\tau\,s} $$


Për një zgjidhje eksplicite konsiderojmë rastin e një burimi tensioni te tipit AC

$$ V(t)=V_0\cos(\omega t)\quad\Longrightarrow\quad V(s)=\frac{sV_0}{s^2+\omega^2} $$

rrjedhimisht, tensioni mbi kondensator bëhet:

$$ V_C(s)=\frac{\frac{sV_0}{s^2+\omega^2}+\tau\,V_{C0}}{1+\tau\,s} $$


#### **Hapi 4: Apliko transformimin invers Laplace (zgjidhja në domenin kohorë)**

Zgjidhja e përgjithshme në domenin kohorë është  

$$ V_C(s)=\frac{sV_0}{(1+\tau\,s)(s^2+\omega^2)}+\frac{\tau\,V_{C0}}{1+\tau\,s} $$

ku e kemi ndarë në dy pjesë për lehtësi integrimi. Gjejmë invers transformimin për termin e parë

$$ v_C(t)=\frac{V_0}{1+(\omega\, \tau)^2}\left[\cos(\omega t)+\omega \tau\,\sin(\omega t)\right]-\frac{V_0 e^{-t/\tau}}{1+(\omega\, \tau)^2} $$

Kjo shprehje përmban dy zgjidhje, njëra për gjëndjen e stacionare (**steady-state**) vetëm me kosinuse, dhe tjetra me termat me exponencialë që quhen termat kalimtarë (**tranzient**). Termi i dytë jep:

$$ V_C(s)=\frac{\tau\,V_{C0}}{1+\tau\,s}\quad\Longrightarrow\quad v_C(t)=V_{C0}e^{-t/\tau} $$

Ky term jep thjeshtë zgjidhjen exponenciale rënëse, pra është një term tranzient.

#### **Hapi 5: Zgjidhja e plotë**

Zgjidhja e plotë e tensionit të kondensatorit për kushte fillestare të ndryshme nga zero është:

$$ v_C(t)= \left(V_{C0}- \frac{V_0}{1+(\omega \,\tau)^2} \right) \,e^{-t/\tau} + \frac{V_0}{1+(\omega \,\tau)^2}\left[\cos(\omega t)+\omega\,\tau\,\sin(\omega t)\right] $$

Ky rezultat tregon qartë që zgjidhja ka:

- Një komponente kalimtare (tranziente) nga voltazhi fillestarë i kondensatorit.
- Një komponente kalimtare nga aplikimi i voltazhit fillestarv i burimit të tensionit.
- Përgjigjen afatgjatë me gjëndjen stacionare (steady-state), pasi komponentet tranziente janë shuar.


#### **Rasti i vecantë 1: Zgjidhja pa burim tensioni, vetëm me kushte fillestare**

Në rast kur $V(t)=0$, kemi thjeshtë përgjigjen/zgjidhjen natyrale:

$$ v_C(t)=V_{C0}\,e^{-t/\tau} $$

#### **Rasti i vecantë 2: Zgjidhja me burim tensioni por pa kushte fillestare**

Në rastin kur $v_c(t=0)=V_{C0} =0$, dhe për kohë mjaftueshmërisht të madhe $t\gg \tau$ zgjidhja kthehet në zgjidhjen stacionare

$$ v_C(t)= \frac{V_0}{1+(\omega \, \tau)^2}\left[\cos(\omega t)+\omega\, \tau\,\sin(\omega t)\right] $$

duke shënuar me $A_c=\frac{V_0}{\sqrt{1+(\omega\,\tau)^2}}$ amplitudën e tensionit në kondensator, dhe me $\varphi=\arctan(\omega\,\tau)$ fazën, përftojmë:

$$ v_C(t)= A_c \, \cos(\omega t - \varphi) $$

Pra me fjalë të tjera, nëse qarku stimulohet me një tension sinusoidal në një kohë shumë të gjatë, efekti i kushteve fillestare dhe i termave kalimtarë (**tranzient**) zhduket. Në qarkun RC, për $t \gg RC=\tau$, mbetet vetëm pjesa sinusoidale e **zgjidhjes stacionare**. Në këto kushte ekuacionet e Laplace-it për zgjidhjen e qarkut reduktohen në formë sinusoidale, pra ne analizën Fourier ku $s = j\omega$. Ky është kalimi kryesor nga Laplace-it në analizën e gjëndjes stacionare sinusoidale (Fourier ose AC) që ne do përdorim këtej e tutje.

### **Supozimi kryesorë për përdorimin e analizes Fourier është:**

1. **Kushte fillestare zero (dmth për $t=0$ kondensatori ka fushë elektrike zero dhe bobina ka fushë magnetike zero).**
2. **Kohë mjaftueshmërisht e gjatë pasi është vendosur gjëndja stacionare.**

**Në këto kushte $s=\sigma+j\omega$ për shkak të mungesës së termave exponenciale tranziente në zgjidhjen për $t\to\infty$ duhet që $\sigma\to 0$ rrjedhimisht $s\to j\omega$.**

### Impemdamca

Impendanca përkufizohet si:

$$ Z(s) = \frac{V(s)}{I(s)} $$

Nëse $s=\sigma +j\omega$ atëherë ky është përkufizimi i përgjithshëm që merr parasysh edhe fenomenet tranzitore (nga efekti i kushteve fillestare të komponenteve apo burimit) si dhe zgjidhjen stacionare që është harmonike. <u>Analiza AC konsideron që në qark është vendosur gjëndja stacionare, rrjedhimisht $s=j\omega$, pra:</u>

$$ Z(j\omega) = \frac{V(j\omega)}{I(j\omega)} $$

ose ndryshe, impendanca është komplekse dhe funksion i frekuencës:

$$ Z(\omega) = \frac{V(\omega)}{I(\omega)}$$


#### Përmbledhje e kalimit Laplace–Fourier (AC):

| Domeni        | Variabli        | Kushte fillestare  | Gjendja kalimtare (tranzient) |
|---------------|-----------------|--------------------|--------------------------------|
| Laplace        | $s=\sigma+j\omega$ | Po, të përgjithshme | Po, trajton tranzientët        |
| Fourier (AC)  | $s=j\omega$ | Zero (gjëndje stacionare) | Jo, vetëm gjëndje stacionare |

Për ta bërë këtë kalim, supozojmë që koha është mjaftueshëm e gjatë dhe efektet kalimtare janë zhdukur plotësisht.


### Interpretimi fizik i Impendancës

- **Impedanca** është një koncept themelor në analizën e gjendjes së qëndrueshme sinusoidale që lehtëson shumë analizën e qarqeve në frekuenca të ndryshme.
- Analiza me transformime të Laplasit na lejon trajtimin e përgjithshëm të qarqeve elektrike, duke përfshirë edhe gjendjet kalimtare dhe kushtet fillestare.
- Kalimi në analizën AC (Fourier) përfshin supozimin e zhdukjes së transientëve dhe gjendjen e stacionare sinusoidale.

## **Shembuj llogaritje e Impendancës**

Më poshtë po kryejmë llogaritjen e impendancës për disa qarqe të thjeshta, duke nisur nga rasti i përgjithshëm dhe duke konvergjuar te gjëndja stacionare (pra kur analiza e Laplace-it konvertohet në analiza Fourier) për të cilën ka rëndësi praktike.

# Qarku RC në seri

Një rezistencë $R$ në seri me një kondensator $C$ dhe një burim tensioni AC:
$$ V(t) = V_0 \cos(\omega t) $$

Nga ligji Kirchhoff-it për tensionet (KVL), kemi:
$$ V(t) = v_R(t) + v_C(t), $$

për C dhe R kemi:
$$ i(t) = C \frac{d v_C(t)}{dt},\quad v_R(t) = R\,i(t)$$

Zëvëndësojmë dhe kemi:
$$
V_0 \cos(\omega t) = R\,C \frac{d v_C(t)}{dt} + v_C(t)
$$

Ky është ekuacioni diferencial (ODE) në domenin e kohës që përcakton sjelljen e qarkut RC në seri.

##### Aplikojme Transformimet e Laplace-it:

- $\mathcal{L}\{v_C(t)\} = V_C(s)$,
- $\mathcal{L}\{ \frac{d v_C(t)}{dt}\} = s V_C(s) - v_C(0)$ për $ v_C(0) = 0$.
- $\mathcal{L}\{\cos(\omega t)\} = \frac{s}{s^2 + \omega^2}$.

Aplikojmë transformimin e Laplace-it për ekuacionin diferencial, dhe marrim:
$$
V_0 \frac{s}{s^2+\omega^2} = R\,C [s\,V_C(s)] + V_C(s)
$$

Faktorizojmë $V_C(s)$:
$$
V_0 \frac{s}{s^2+\omega^2} = V_C(s)(R\,C\,s + 1)
$$

##### Gjejmë impedancën $Z_{RC}(s)$:

Impedanca përkufizohet si $Z(s)=V(s)/I(s)$. Rryma që përshkon kondensatorin është:
$$ I(s) = C s V_C(s) $$

Kështuqë impendanca totale është:
$$
Z_{RC}(s) = \frac{V(s)}{I(s)} = \frac{V_C(s)(R\,C\,s + 1)}{C s V_C(s)} = R + \frac{1}{sC}
$$

##### Nga Laplace tek Fourier (AC steady-state):

Për të përftuar impendancën AC vendosim $s = j\omega$:
$$
Z_{RC}(\omega) = R + \frac{1}{j\omega C}
$$



## Qarku RL në seri

Njëlloj kemi rezistencën $R$ dhe bobinën $L$ dhe burimin AC:
$$ V(t) = V_0 \cos(\omega t) $$

Nga KVL:
$$
V(t) = v_R(t) + v_L(t) = R\,i(t) + L \frac{di(t)}{dt}
$$

Transformimi i Laplace-it:

- $\mathcal{L}\{i(t)\}=I(s)$,
- $\mathcal{L}\{\frac{di(t)}{dt}\} = s I(s)$, 
- $\mathcal{L}\{\cos(\omega t)\}=\frac{s}{s^2+\omega^2}$

Kështu:
$$
V_0\frac{s}{s^2+\omega^2}= R\,I(s)+L\,s\,I(s)
$$

Faktorizo $I(s)$:
$$
V_0\frac{s}{s^2+\omega^2}=I(s)(R+L s)
$$

##### Impedanca e qarkut RL është:
$$
Z_{RL}(s)=\frac{V(s)}{I(s)}=R+Ls
$$

për analizën AC (steady-state) $s=j\omega $:
$$ Z_{RL}(\omega)=R+j\omega L $$


## RLC në Seri

KVL jep:
$$
V(t)=R i(t)+L\frac{di(t)}{dt}+\frac{1}{C}\int i(t)\,dt
$$

Derivojmë njëherë të eleminojmë integralin, dhe kemi:
$$
\frac{dV(t)}{dt}=R\frac{di(t)}{dt}+L\frac{d^2i(t)}{dt^2}+\frac{1}{C}i(t)
$$

$$
-V_0\omega\sin(\omega t)=R\frac{di(t)}{dt}+L\frac{d^2i(t)}{dt^2}+\frac{1}{C}i(t)
$$

Transformi Laplace-it:

- $\mathcal{L}\{i(t)\}=I(s)$,
- $\mathcal{L}\{\frac{di}{dt}\}=sI(s)$,
- $\mathcal{L}\{\frac{d^2i}{dt^2}\}=s^2 I(s)$,
- $\mathcal{L}\{-V_0\omega\sin(\omega t)\}=\frac{-V_0\omega^2}{s^2+\omega^2}$.

Tani kemi:
$$
\frac{-V_0\omega^2}{s^2+\omega^2}=R s I(s)+L s^2 I(s)+\frac{1}{C}I(s)
$$

Faktorizo $I(s)$:
$$
\frac{-V_0\omega^2}{s^2+\omega^2}=I(s)(L s^2+R s+\frac{1}{C})
$$

Me një ri-rregullim, gjejmë që impendanca është:
$$
Z_{RLC}(s)=L\, s + R + \frac{1}{sC}
$$

vendos $s=j\omega$:
$$
Z_{RLC}(j\omega)=R+j\omega L+\frac{1}{j\omega C}=R+j\left(\omega L-\frac{1}{\omega C}\right)
$$


## RLC në Paralel

Nga lidhja në paralel kemi:
$$
i(t) = i_R(t) + i_L(t) + i_C(t)
$$

ku:

- $i_R(t) = \frac{V(t)}{R}$
- $i_L(t) = \frac{1}{L} \int V(t)\,dt$
- $i_C(t) = C \frac{dV(t)}{dt}$

Njëlloj si RLC ne seri, kryejmë edhe një derivat që të eleminojmë integralin pastaj aplikojmë transformimin e Laplace-it:
$$
I(s) = \frac{V(s)}{R} + \frac{V(s)}{sL} + sC\,V(s)
$$

Faktorizojmë $V(s)$:
$$
I(s) = V(s) \left[\frac{1}{R} + \frac{1}{sL} + sC\right]
$$

Gjejmë impendancën nga përkufizimi si:
$$
Z(s) = \frac{V(s)}{I(s)} = \frac{1}{\frac{1}{R} + \frac{1}{sL} + sC}
$$

Pra:
$$
Z_{RLC}^{\text{parallel}}(s) = \left[\frac{1}{R} + \frac{1}{sL} + sC\right]^{-1}
$$

Në analizën AC, marrim $s = j\omega$ dhe kemi:
$$
Z_{RLC}^{\text{parallel}}(\omega) = \left[\frac{1}{R} + \frac{1}{j\omega L} + j\omega C\right]^{-1}
$$

Mund ta shkruajmë si:
$$
Z(\omega) = \frac{1}{\frac{1}{R} + j\left(\omega C - \frac{1}{\omega L}\right)}
$$


## **Permbledhje:**

| Circuit | Impedanca Laplace-it $Z(s)$ | Impendanca AC (Fourier) $Z(j\omega)$|
|---------|----------------------------|----------------------------|
| RC      | $R+\frac{1}{sC}$         | $R+\frac{1}{j\omega C}$  |
| RL      | $R+sL$                   | $R+j\omega L$            |
| RLC në seri | $R+sL+\frac{1}{sC}$   | $R+j(\omega L-\frac{1}{\omega C})$ |
| RLC në paralel | $\left[\frac{1}{R} + \frac{1}{sL} + sC\right]^{-1}$ | $\left[\frac{1}{R} + \frac{1}{j\omega L} + j\omega C\right]^{-1}$ |

- Extrakto impendancën nga $Z(s)=V(s)/I(s)$.
- Supozimi që qarku është në gjëndjen stacionare bën të zëvëndësosh $s=j\omega$ që na lejon të gjejmë $Z(\omega)$.

# Qarku RLC në seri

## **Ekuacionet e Kirchhoffit**

Le të konsiderojmë një qark të thjeshtë RLC me komponentët në seri si në figurë:

<img src="figs/RLC_seri.svg" align="center" width=400 style="margin-left:auto; margin-right:auto"/>

Nga ligji i Kirchhoffit për tensionet (KVL):

$$ v(t) = v_R(t) + v_L(t) + v_C(t) $$

Duke i shprehur tensionet në mënyrë komplekse, marrim:
$$
V_0 e^{j\omega t} = R i(t) + L\frac{di(t)}{dt} + \frac{1}{C}\int i(t) dt
$$

## **Analiza në domenin kompleks**

Në regjimin sinusoidal, rryma ka formën:
$$ i(t)= I_0 e^{j\omega t} $$

- Rezistenca: $v_R = R i(t)$
- Bobina: $v_L = j\omega L i(t)$
- Kondensatori: $v_C = \frac{i(t)}{j\omega C}$

Pra ekuacioni bëhet (për amplitudat komplekse):
$$
V = R I + j\omega L I + \frac{I}{j\omega C}
$$

## **Impedanca e përgjithshme $Z$**

Impedanca totale në seri është:
$$
Z = R + j\omega L + \frac{1}{j\omega C} = R + j\left(\omega L - \frac{1}{\omega C}\right)
$$

Mund ta shkruajmë edhe:
$$
Z = R + jX, \quad \text{ku} \quad X = \omega L - \frac{1}{\omega C}
$$

Këtu:

- $R$ është **rezistenca aktive**,
- $X$ është **rezistenca reaktive**.


## **Rryma dhe tensionet**

Rryma në qark është:
$$
I = \frac{V}{Z} = \frac{V_0}{\sqrt{R^2+\left(\omega L - \frac{1}{\omega C}\right)^2}} e^{j(\omega t - \phi)}
$$

ku këndi i fazës është:
$$
\phi = \tan^{-1}\frac{\omega L - \frac{1}{\omega C}}{R}
$$

Tensionet në secilin komponent:

- Në rezistencë: 
$$ V_R = IR $$

- Në bobinë: 
$$ V_L = I (j\omega L) $$

- Në kondensator:
$$ V_C = I\frac{1}{j\omega C} $$

## **Rezonanca në qarkun RLC**

Në frekuencën e rezonancës, rezistenca reaktive është zero:
$$
X = \omega_0 L - \frac{1}{\omega_0 C} = 0 \quad \Rightarrow \quad \omega_0 = \frac{1}{\sqrt{LC}}
$$

Në këtë pikë, qarku është **rezonant**, dhe rryma arrin vlerën maksimale:
$$ I_{max}=\frac{V_0}{R}$$

## **Dallimi në fazë**

- Në rezonancë ($\omega = \omega_0$): **rryma dhe tensioni janë në fazë**.
- Për $\omega < \omega_0$, qarku ka natyrë **kapacitive**, rryma paraprin tensionin.
- Për $\omega > \omega_0$, qarku ka natyrë **induktive**, rryma vonohet ndaj tensionit.


### **Përkufizimi i faktorit të cilësisë (Q) Q-factor**

Faktori i cilësisë përcaktohet si raporti midis energjisë së ruajtur në sistem dhe energjisë së humbur në çdo cikël të lëkundjes:

$$
Q = 2\pi \frac{\text{Energji e ruajtur në çdo cikël}}{\text{Energjia e humbur në çdo cikël}}
$$

Në kontekstin e qarqeve elektrike **RLC në seri**, faktori i cilësisë përcaktohet nga komponentët e qarkut:

$$
Q = \frac{\omega_0 L}{R} = \frac{1}{\omega_0 C R}
$$

ku: $\omega_0$ është frekuenca e rezonancës: $\omega_0 = \frac{1}{\sqrt{LC}}$. Shiko seksionin më poshtë për llogaritjen e $Q$-së.

### **Interpretimi fizik i Q**

- **Q i lartë**:
  - Humbjet e energjisë janë të vogla.
  - Kurba rezonante është **e ngushtë dhe e mprehtë**.
  - Qarku ruan mirë energjinë e tij.
  - Rrezonon në një interval të ngushtë frekuencash.

- **Q i ulët**:
  - Humbje të mëdha energjie (rezistencë e lartë).
  - Kurba rezonante është **më e gjerë dhe e sheshtë**.
  - Qarku disipon shpejt energjinë.
  - Rrezonon në një interval më të gjerë frekuencash.

### **Lidhja me bandwidth (gjerësinë e brezit)**

Faktori i cilësisë lidhet me gjerësinë e brezit $\Delta \omega$:

$$ \Delta \omega = \frac{\omega_0}{Q} $$

Kjo do të thotë se sa më i lartë të jetë faktori Q, aq më e ngushtë është gjerësia e rezonancës.

#### **Zbatime praktike**

Në praktikë, faktorë të lartë të cilësisë përdoren në qarqet selektive, siç janë:

- **Radio-sintonizuesit** – zgjedhja e një frekuence specifike mes shumë frekuencave.
- **Filtrat selektivë** – zgjedhja e sinjaleve të dëshiruara me precizion të lartë.
- **Qarqe rezonante në pajisjet matëse** – ku kërkohet saktësi e lartë.


#### **Pra për qarkun RLC në seri**

| Madhësia                  | Shprehja                                     |
|---------------------------|----------------------------------------------|
| Impedanca totale          | $Z = R + j(\omega L - \frac{1}{\omega C})$ |
| Frekuenca e rezonancës    | $\omega_0 = \frac{1}{\sqrt{LC}}$           |
| Këndi i fazës             | $\phi = \tan^{-1}\frac{\omega L - \frac{1}{\omega C}}{R}$ |
| Faktori i cilësisë        | $Q= \omega_0 L/R$                          |
| Gjerësinë e bandës        | $\Delta \omega = \frac{\omega_0}{Q}$       |


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Funksioni që llogarit dhe grafikon rrymën
def rlc_current(R, L, C):
    V0 = 10  # Amplituda e burimit (Volt)
    C  *= 1e-6 # Konverto microFarad ne Farad
    f = np.logspace(1, 4, 1000)  # Frekuencat (Hz)
    omega = 2 * np.pi * f

    Z = R + 1j * (omega * L - 1/(omega * C))
    I = V0 / np.abs(Z)

    plt.figure(figsize=(9,5))
    plt.plot(f, I, '-k', lw=2)
    plt.title('Përgjigja e rrymës në frekuencë për qarkun RLC')
    plt.xlabel('Frekuenca [Hz]')
    plt.ylabel('Rryma [A]')
    plt.grid(True)
    plt.ylim(0, np.max(I)*1.1)
    plt.xscale('log')
    plt.show()

# Sliders interaktiv për R, L dhe C
interact(rlc_current,
         R=FloatSlider(min=10, max=1000, step=10, value=100, description=r'$R (\Omega)$'),
         L=FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description=r'$L (H)$'),
         C=FloatSlider(min=0.1, max=100, step=0.1, value=10, description=r'$C (\mu F)$')
        );
```

# **Analiza në domainin e kohës e qarkut RLC në seri**

Duke përdorur ligjin e Kirchhoff-it:

$$
v(t) = v_R(t) + v_L(t) + v_C(t)
$$

Në terma të rrymës $i(t)$:

$$
v(t) = R i(t) + L\frac{di(t)}{dt} + \frac{1}{C}\int i(t) dt
$$

Duke diferencuar një herë, marrim ekuacionin diferencial të rendit të dytë:

$$
L\frac{d^2 i(t)}{dt^2}+R\frac{d i(t)}{dt}+\frac{1}{C}i(t)=\frac{d v(t)}{dt}
$$

## **Rasti pa burim (lëkundje të lira)**

Nëse burimi është hequr ($v(t) = 0$), ekuacioni bëhet homogjen:

$$
L\frac{d^2 i(t)}{dt^2}+R\frac{d i(t)}{dt}+\frac{1}{C}i(t)=0
$$

E shkruajmë në formën standarde:

$$
\frac{d^2 i(t)}{dt^2}+2\alpha \frac{d i(t)}{dt}+\omega_0^2 i(t)=0
$$

ku:

- Frekuenca natyrore (pa frenim): 
$$ \omega_0=\frac{1}{\sqrt{LC}} $$

- Faktori i frenimit ($\alpha$):
$$ \alpha=\frac{R}{2L} $$

## **Llojet e frenimit (shuarjes)**

Zgjidhja e këtij ekuacioni homogjen varet nga raporti mes $\alpha$ dhe $\omega_0$:

1. **Frenimi i dobët** ($\alpha<\omega_0$):
   - Zgjidhja:
   $$ i(t)=e^{-\alpha t}[A\cos(\omega_d t)+B\sin(\omega_d t)],\quad \omega_d=\sqrt{\omega_0^2-\alpha^2} $$
   - Sjellje: Sinusoidale me amplitudë eksponencialisht në rënie.

2. **Frenimi kritik** ($\alpha=\omega_0$):
   - Zgjidhja:
   $$ i(t)=(A+Bt)e^{-\alpha t}$$
   - Sjellje: Rënie më e shpejtë pa oshilacione.

3. **Frenimi i fortë** ose mbifrenimi ($\alpha>\omega_0$):
   - Zgjidhja:
   $$
   i(t)=A e^{s_1 t}+B e^{s_2 t},\quad s_{1,2}=-\alpha \pm\sqrt{\alpha^2-\omega_0^2}
   $$
   - Sjellje: Rënie eksponenciale pa oshilacione.


```python
# Parametrat
R = 10  # Ohm
L = 0.2  # Henry
C = 10e-6  # Farad
alpha = R/(2*L)
omega_0 = 1/np.sqrt(L*C)

t = np.linspace(0, 0.05, 1000)

# Kontrollojme frenimin
if alpha < omega_0:
    omega_d = np.sqrt(omega_0**2 - alpha**2)
    i_t = np.exp(-alpha*t)*np.cos(omega_d*t)
    title = "Frenim i dobët (me oscilacione)"
elif alpha == omega_0:
    i_t = (1 + alpha*t)*np.exp(-alpha*t)
    title = "Frenim kritik"
else:
    s1 = -alpha + np.sqrt(alpha**2 - omega_0**2)
    s2 = -alpha - np.sqrt(alpha**2 - omega_0**2)
    i_t = 0.5*(np.exp(s1*t) + np.exp(s2*t))
    title = "Frenim i fortë (pa oscilacione)"

# Grafik
plt.figure(figsize=(10, 5))
plt.plot(t, i_t, '-k', lw=2)
plt.title(f"Lëkundjet e rrymës në qarkun RLC ({title})")
plt.xlabel('Koha [s]')
plt.ylabel('Rryma (A)')
plt.grid(True)
plt.show()
```

# **Qarku RLC në paralel**

Në një qark RLC në paralel, të gjitha elementet lidhen në paralel me një burim tensioni sinusoidal $ v(t) = V_0 e^{j\omega t} $ secili komponent ka të njëjtin tension $v(t)$ mbi të.

<img src="figs/RLC_paral.svg" align="center" width=400 style="margin-left:auto; margin-right:auto"/>

## **Ekuacionet dhe impedancat**

- Rezistenca: $I_R = \frac{V}{R}$
- Kondensatori: $I_C = j\omega C V$
- Bobina: $I_L = \frac{V}{j\omega L}$

**Rryma totale** në qark është:
$$ I_{tot} = I_R + I_C + I_L = V\left(\frac{1}{R}+j\omega C+\frac{1}{j\omega L}\right)$$

**Admitanca totale (Y)** është inversi i impedancës (si përcjellshmëria $1/R$):

$$ Y = \frac{1}{Z} = \frac{1}{R}+j\omega C+\frac{1}{j\omega L} = \frac{1}{R}+j\left(\omega C - \frac{1}{\omega L}\right) $$

## **Frekuenca e rezonancës ($\omega_0$)**

Në rezonancë, admitanca imagjinare është zero:

$$ \omega_0 C - \frac{1}{\omega_0 L}=0 \quad \Rightarrow \quad \omega_0 = \frac{1}{\sqrt{LC}}$$

Në këtë pikë, admitanca është minimale, impedanca maksimale, dhe rryma nga burimi është minimale.

## **Faktori i cilësisë ($Q$) në qarkun paralel**

Faktori i cilësisë në qarkun paralel është:
$$ Q = R \sqrt{\frac{C}{L}} = \frac{\omega_0}{\Delta \omega} $$

ku gjerësia e brezit (bandwidth) është:
$$
\Delta \omega = \frac{\omega_0}{Q} = \frac{1}{R C}
$$

## **Ekuacioni diferencial (analiza në kohë)**

Duke shkruar ligjin e Kirchhoffit për rrymat në nyjën e përbashkët, marrim:
$$
i_R(t)+i_C(t)+i_L(t)=0
$$

Në terma të tensionit, marrim ekuacionin diferencial:
$$
C\frac{d^2 v}{dt^2}+\frac{1}{R}\frac{dv}{dt}+\frac{1}{L}v=0
$$

Kjo shprehje mund të shkruhet në formën standarde:
$$
\frac{d^2 v}{dt^2}+2\alpha \frac{dv}{dt}+\omega_0^2 v=0
$$

ku:

- $\alpha=\frac{1}{2RC}$ është faktori i frenimit.
- $\omega_0=\frac{1}{\sqrt{LC}}$ është frekuenca e rezonancës.


## **Interpretimi fizik i qarkut RLC paralel**

- Në rezonancë, bobina dhe kondensatori shkëmbejnë energjinë mes tyre pa marrë energji nga burimi.
- Rezistenca humb energjinë dhe përcakton frenimin (shuarjen).
- Sa më e madhe të jetë $R$, aq më i lartë $Q$, dhe më e ngushtë kurba e rezonancës.



```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def parallel_rlc(R, L, C):
    V0 = 10  # Volt
    f = np.logspace(1, 4, 1000)  # Hz
    omega = 2 * np.pi * f
    C *= 1e-6 # konverto nga mikro Farad ne Farad
    
    Y = 1/R + 1j*(omega*C - 1/(omega*L))
    Z = 1/Y
    I = V0 / np.abs(Z)

    plt.figure(figsize=(9,5))
    plt.plot(f, I, '-k', lw=2)
    plt.title('Përgjigja e rrymës në frekuencë për qarkun RLC paralel')
    plt.xlabel('Frekuenca [Hz]')
    plt.ylabel('Rryma [A]')
    plt.grid(True)
    plt.ylim(0, np.max(I)*1.1)
    plt.xscale('log')
    plt.show()

interact(parallel_rlc,
         R=FloatSlider(min=10, max=2000, step=10, value=500, description=r'$R (\Omega)$'),
         L=FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description=r'$L (H)$'),
         C=FloatSlider(min=0.1, max=100, step=0.1, value=10, description=r'$C (\mu F)$')
        );
```

# **Qarku Diferencues dhe Qarku Integrues RC**

## Qarku Diferencues
Në qarkun diferencues, sinjali hyrës $v_{in}(t)$ vendoset në kondensator dhe sinjali i daljes merret mbi rezistencën.

<img src="figs/RC_dif.svg" align="center" width=400 style="margin-left:auto; margin-right:auto"/>

Nga ligji i Kirchhoff-it për rrymat në nyjën e daljes kemi $ i_C(t) = i_R(t)$ duke ditur që:
$$
i_C(t) = C\frac{d}{dt}[v_{in}(t)-v_{out}(t)], \quad i_R(t)=\frac{v_{out}(t)}{R}
$$

**Nëse zgjedhim rezistencën të vogël, dalja do jetë e vogël, dhe supozojmë $v_{out} \ll v_{in}$ **, ekuacioni bëhet:
$$ C\frac{dv_{in}(t)}{dt} \approx \frac{v_{out}(t)}{R} $$

Pra sinjali ne dalje është **proporcionale me derivatin** e sinjalit në hyrje:
$$ v_{out}(t)=RC\frac{dv_{in}(t)}{dt} $$

<br>

## **Qarku Integrues RC**

Në qarkun integrues, sinjali hyrës $v_{in}(t)$ vendoset në rezistencë dhe dalja merret mbi kondensatorin.

<img src="figs/RC_int.svg" align="center" width=400 style="margin-left:auto; margin-right:auto"/>

Rryma që kalon në rezistencë shkon në kondensator $ i_R(t)=i_C(t)$ duke ditur që:
$$ i_R(t)=\frac{v_{in}(t)-v_{out}(t)}{R}, \quad i_C(t)=C\frac{dv_{out}(t)}{dt} $$

Në qark ideal, $v_{out}\ll v_{in}$, marrim:
$$ \frac{v_{in}(t)}{R}\approx C\frac{dv_{out}(t)}{dt} $$

Pra sinjali në dalje është **proporcionale me integralin** e sinjalit në hyrje:
$$ v_{out}(t)=\frac{1}{RC}\int v_{in}(t)\, dt $$


## **Analiza Fourier**

Në domenin e frekuencës, impedancat tregojnë qartë:

- **Qarku diferencues** ka impedancën:
$$ Z = R + \frac{1}{j\omega C} \approx \frac{1}{j\omega C}, \quad (\omega \text{ e madhe}) $$
Dalja rritet me rritjen e frekuencës.

- **Qarku integrues** ka impedancën:
$$ Z = R + \frac{1}{j\omega C} \approx R, \quad (\omega \text{ e madhe})$$
Dalja zvogëlohet me rritjen e frekuencës.


## **Aplikimet Praktike*

- Diferencues: detektor impulsesh, gjenerues impulsi.
- Integrues: prodhues sinjali trekëndor, filtrues.



## **Këshilla për Implementimin Praktik**

#### **1. Për Qarku Diferencues**

Në qarkun diferencues, dalja është e përafërt me derivatin e hyrjes në kushtin që frekuenca e sinjalit hyrës të jetë shumë më e vogël se frekuenca karakteristike $f_c=1/RC$:

$$ f_{in} \ll f_c = \frac{1}{2\pi R C} $$

**Këshilla praktike për zgjedhjen e R dhe C:**

- Përcakto frekuencën më të lartë të sinjalit tënd hyrës $f_{max}$.
- Zgjidh $f_c$ të jetë rreth **10 herë më e madhe** se frekuenca maksimale e hyrjes:
  $$ f_c \approx 10 \cdot f_{max} $$

- Nga kjo frekuencë, zgjedh:
  - Një kondensator standard (p.sh. nga 1 nF deri në 100 nF).
  - Pastaj llogarit R nga:
    $$ R = \frac{1}{2\pi f_c C} $$

- Praktikisht:
  - Për sinjale audio (10–20 kHz), zakonisht përdorim $C = 10\,\text{nF}$ dhe $R$ nga 1 kΩ deri në 10 kΩ.
  - Për sinjale me frekuencë më të lartë (MHz), përdorim kondensatore më të vegjël (p.sh. 10–100 pF).


#### **2. Qarku Integrues**

Në qarkun integrues, dalja është përafërsisht integrali i sinjalit hyrës në kushtin që frekuenca e sinjalit hyrës të jetë shumë më e madhe se frekuenca karakteristike $f_c$:

$$ f_{in} \gg f_c = \frac{1}{2\pi R C} $$


**Këshilla praktike për zgjedhjen e R dhe C:**

- Përcakto frekuencën më të ulët që duhet integruar, $f_{min}$.
- Zgjidh frekuencën karakteristike $f_c$ të jetë rreth **10 herë më e vogël** se $f_{min}$:
  $$ f_c \approx \frac{f_{min}}{10} $$

- Zgjidh:
  - Një rezistencë standarde (1 kΩ deri 100 kΩ).
  - Kapacitetin llogarite me:
    $$ C = \frac{1}{2\pi f_c R} $$

- Praktikisht:
  - Për frekuenca audio, mund të zgjedhësh rezistencën rreth 10 kΩ deri 100 kΩ, dhe kondensatorin nga 10 nF deri 1 µF.
  - Për frekuenca më të larta, përdor kondensatorë më të vegjël (nF ose pF).


#### **3. Kujdes për zgjedhjen e vlerave:**

- Kapacitorët shumë të mëdhenj (> 1 µF) mund të kenë probleme me rrjedhje (leakage current).
- Rezistencat e mëdha (> 1 MΩ) mund të fusin zhurmë (noise) dhe ndikohen nga ndërhyrjet elektromagnetike.
- Zgjidhni komponentë me tolerancë të vogël (1–5%) për rezultate të sakta.



### **Shembull praktik:**

Nëse sinjali hyrës është një valë sinusoidale me frekuencë maksimale 1 kHz, për qarkun diferencues:

- Zgjidh $f_c \approx 10\,\text{kHz}$.
- Nëse zgjedhim $C=10\,\text{nF}$, atëherë:
$$ R = \frac{1}{2\pi f_c C} = \frac{1}{2\pi \cdot 10000 \cdot 10\times10^{-9}} \approx 1.6\,\text{k}\Omega $$

Në qarkun integrues për të njëjtën hyrje 1 kHz (si frekuenca minimale):

- Zgjidh $f_c \approx 100\,\text{Hz}$.
- Nëse marrim $R=10\,\text{k}\Omega$, atëherë:
$$C = \frac{1}{2\pi f_c R} = \frac{1}{2\pi \cdot 100 \cdot 10000} \approx 0.16\,\mu F
$$

Kapaciteti më i afërt standard që i afrohet është: 0.15 µF ose 0.22 µF. Zgjidh njërin sepse e plotësojne kushtin!



```python
t = np.linspace(0, 0.01, 1000)
vin = np.sin(2*np.pi*1000*t)  # 1kHz sinjal
RC = 1e-4

# Diferencuesi
vout_diff = RC * 2*np.pi*1000 * np.cos(2*np.pi*1000*t)

# Integruesi
vout_int = (-1/(RC * 2*np.pi*1000)) * np.cos(2*np.pi*1000*t)

plt.figure(figsize=(10,5))
plt.plot(t, vin, '-k', lw=2, label='Vin (sin)')
plt.plot(t, vout_diff, '-r', lw=2, label='Diferencues (cos)')
plt.plot(t, vout_int, '-b', lw=2, label='Integrues (-cos)')
plt.xlabel('Koha [s]')
plt.ylabel('Tensioni [V]')
plt.title('Sinjali hyrës dhe përgjigjet e integruesit dhe diferencuesit')
plt.legend()
plt.grid(True)
plt.show()
```

# Lexim: Llogaritja e faktorit $Q$ për qarkun RLC

## 1. Përkufizimi fizik i $Q$

Faktori i cilësisë përcaktohet si:

$$
Q = 2\pi \cdot \frac{\text{Energjia e ruajtur në qark}}{\text{Energjia e disipuar për një cikël}}
$$

Ky raport mat **sa mirë ruan energjinë qarku**, krahasuar me sa shumë e humbet për një cikël rrotull rezonancës.


## 2. Shembull: Qarku **RLC në seri**

Qarku:

$$
V(t) \to R \text{ — } L \text{ — } C
$$

Rryma në qark:
$$
i(t) = I_0 \cos(\omega t - \varphi)
$$

Në rezonancë ($\omega = \omega_0 = \frac{1}{\sqrt{LC}}$), tensionet mbi bobinën dhe kondensatorin anulojnë njëra-tjetrën, dhe kemi:

- Impedanca minimale: $Z = R$
- Rryma maksimale: $i(t) = \frac{V_0}{R} \cos(\omega_0 t)$

## 3. Llogaritja e energjisë së ruajtur

### a) Në **bobinë**:

Energjia ruajtur në fushën magnetike:
$$ W_L = \frac{1}{2} L i^2(t) $$

Amplituda maksimale:
$$ W_{L,\text{max}} = \frac{1}{2} L \left(\frac{V_0}{R}\right)^2 $$

### b) Në **kondensator**:

Energjia ruajtur në fushën elektrike:
$$ W_C = \frac{1}{2} C v_C^2(t) $$

Por në rezonancë, tensioni mbi $C$ dhe $L$ është shumë më i madh se ai i burimit, sepse ka qarkullim të energjisë. Vlera maksimale e energjisë është mesatarja midis $W_L$ dhe $W_C$, pra konsiderojmë vetëm njërën për përllogaritje, p.sh. $W_L$.

## 4. Llogaritja e energjisë së harxhuar për cikël

Në rezistencë, energjia humbet si nxehtësi:
$$ P(t) = i^2(t) \cdot R $$

Energji e humbur për një cikël:
$$
\Delta W = \int_0^T i^2(t) \cdot R \, dt = R \int_0^T \left(\frac{V_0}{R}\right)^2 \cos^2(\omega_0 t)\,dt
$$

Shkurtimisht:
$$
\Delta W = \frac{V_0^2}{R} \cdot \int_0^T \cos^2(\omega_0 t)\,dt = \frac{V_0^2}{R} \cdot \frac{T}{2}
$$

ku $T = \frac{2\pi}{\omega_0}$ është perioda.


## 5. Tani llogarisim $Q$:

Energjia ruajtur maksimale:
$$
W = \frac{1}{2} L \left(\frac{V_0}{R}\right)^2
$$

E harxhuar për cikël:
$$
\Delta W = \frac{V_0^2}{R} \cdot \frac{T}{2}
$$

Nga përkufizimi i $Q$:
$$
Q = 2\pi \cdot \frac{W}{\Delta W} = 2\pi \cdot \frac{\frac{1}{2} L \left(\frac{V_0}{R}\right)^2}{\frac{V_0^2}{R} \cdot \frac{T}{2}} = 2\pi \cdot \frac{L}{R} \cdot \frac{1}{T}
$$

Por $$T = \frac{2\pi}{\omega_0}$$, prandaj:

$$
Q = \omega_0 \cdot \frac{L}{R}
\quad \Rightarrow \quad
\boxed{Q = \frac{\omega_0 L}{R}}
$$

## 6. Forma alternative për $Q$ e RLC në seri:

Duke përdorur që $\omega_0 = \frac{1}{\sqrt{LC}}$, zëvendësojmë:

$$
Q = \frac{1}{R} \cdot \sqrt{\frac{L}{C}} = \frac{1}{\omega_0 C R}
$$


## 7. Qarku **RLC në paralel**

Kemi konfigurimin:

$$
V(t) \to \left( R \parallel L \parallel C \right)
$$

Analiza bëhet në të njëjtën mënyrë, por për qarkun paralel përdorim:

- Rryma alternative që rrjedh në bobinë dhe kondensator, ndërsa rezistenca shërben për harxhimin e energjisë.
- Llogaritet energjia e ruajtur (në $L$ ose $C$) dhe energjia e disipuar në $R$.

Në fund arrijmë:
$$
\boxed{Q = R \cdot \sqrt{\frac{C}{L}}}
$$

ose në mënyrë ekuivalente:

$$
Q = \frac{1}{R} \cdot \sqrt{\frac{L}{C}}
$$

(të dyja janë të vlefshme në varësi të përkufizimit të funksionit që analizojmë – tension ose rrymë)


#### Përfundim

| Qarku         | Formula e $Q$                             |
|---------------|---------------------------------------------|
| RLC në seri   | $Q = \frac{\omega_0 L}{R} = \frac{1}{\omega_0 C R}$ |
| RLC në paralel| $Q = R \sqrt{\frac{C}{L}} = \frac{1}{R} \sqrt{\frac{L}{C}}$ |

# Filtrat

## Përkufizimi i filtrit elektronik

Një **filtër elektronik** është një sistem linear që lejon kalimin e disa komponentëve frekuencorë të një sinjali dhe dobëson ose ndalon të tjerat. Ai vepron si një përzgjedhës i frekuencave, duke kontrolluar sjelljen e sistemit në varësi të frekuencës së sinjalit hyrës.


### Qëllimi dhe funksioni i filtrit

Filtri modifikon spektrin e frekuencave të një sinjali:

- Lejon kalimin e frekuencave të dëshiruara (banda e kalimit),
- Dobëson frekuenca të padëshiruara (banda e ndalimit).

Nëpërmjet kësaj, ai mund të përzgjedhë sinjale, të largojë zhurma, të formësojë llojin e valës ose të realizojë operacione të ngjashme si derivimi dhe integrimi.


### Përshkrimi matematikor

Sjellja e filtrit përshkruhet nga **funksioni i transferimit** në domenin e Laplasit:

$$
H(s) = \frac{V_{out}(s)}{V_{in}(s)}
$$

Ku, $V_{in}$ dhe $V_{out}$ janë sinjalet hyrëse dhe dalëse nga filtri, respektivisht. Në analizën e gjendjes së qëndrueshme sinusoidale (AC), përdoret $H(j\omega)$, ku:

- Moduli i $H(j\omega)$ përfaqëson amplifikimin (ose dobësimin) për frekuencën $\omega$,
- Faza e $H(j\omega)$ tregon zhvendosjen e fazës së sinjalit.


### Karakteristikat kryesore të një filtri

- **Frekuenca e prerjes (kritike)** ($f_c$): frekuenca ku sinjali fillon të bie ndjeshëm (zakonisht 3 dB, dmth $V_{out}/V_{in}=\sqrt(2)$).
- **Banda e kalimit**: brezi i frekuencave që transmetohen me pak ose aspak dobësim.
- **Banda e ndalimit**: brezi i frekuencave që dobësohen në mënyrë të theksuar.
- **Rënia (roll-off)**: shkalla me të cilën magnituda bie jashtë bandes së kalimit.
- **Faza**: vonesa ose avancimi i sinjalit për frekuenca të ndryshme.
- **Selektiviteti**: aftësia për të selektuar banden e lejuar (lidhet me faktorin e cilësisë $Q$).


### Lloje të zakonshme filtrash

- **Low-pass**: lejon frekuencat e ulëta, ndalon të lartat.
- **High-pass**: lejon frekuencat e larta, ndalon të ulëtat.
- **Band-pass**: lejon një brez frekuencash mes dy kufijve.
- **Band-stop (notch)**: ndalon një brez frekuencash.


### Aplikime praktike të filtrave

| Fusha | Qëllimi i përdorimit |
|-------|-----------------------|
| Audio | Përcaktimi i tonit (bass/treble), reduktim zhurme |
| Telekomunikacion | Përzgjedhja e frekuencës së kanalit |
| Instrumentacion | Filtrim i sinjaleve matëse nga zhurma |
| Elektronikë dixhitale | Formësim i pulsit, eliminim harmonikash |


### Klasifikim i përgjithshëm

- **Pasivë**: përbëhen vetëm nga elementë R, L dhe C, nuk amplifikojnë sinjalin.
- **Aktivë**: përfshijnë përforcues operacionalë (OpAmp) dhe mund të realizojnë përforcim dhe filtrim më fleksibël.
- **Digjitalë**: implementohen përmes algoritmeve në mikroprocesorë (llogaritës).

## **Grafiku Bode** 

### Përkufizimi

**Grafiku Bode** përfaqëson funksionin e transferimit $H(j\omega)$ të një sistemi në dy grafikë të ndarë:

1. **Magnituda**:  
   $$ 20 \log_{10}\left(|H(j\omega)|\right) $$
   paraqitet në funksion të frekuencës $\omega$ (zakonisht në shkallë logaritmike dmth $\log(\omega))$).

2. **Faza**:  
   $$ \arg H(j\omega) $$
   paraqitet po ashtu ndaj frekuencës logaritmike, në gradë ose radian.


## Qëllimi dhe përdorimi

Grafiku Bode është një mjet i fuqishëm për:

- **Analizën frekuencore të filtrave**: për të parë se cilat frekuenca transmetohen dhe cilat dobësohen.
- **Karakterizimin e sjelljes së sistemeve dinamike**: për shembull, në inxhinieri të kontrollit (gain margin, phase margin).
- **Vlerësimin e rënies së sinjalit** në dB/dekadë, që tregon sjelljen asimptotike të filtrit.
- **Zgjedhjen e frekuencës së prerjes (kritike)** dhe projektimin e parametrave të komponentëve elektronike.


## Pse përdorim shkallë logaritmike?

- Shumë sinjale mbulojnë frekuenca në disa rende madhësie (Hz deri në MHz).
- Rëniet dhe amplifikimet zakonisht janë të tipit proporcional në logaritëm të frekuencës.
- Logaritmi lejon paraqitje të qartë të raportit të fuqive dhe sjelljes eksponenciale.

# Filtri **RC Kalim-i-Ultë** (Low-pass)

<img src="figs/RC_LowPass.svg" width=300/>


### Analiza:

Përdorim domenin e Laplasit:

- Rezistenca: $Z_R = R$
- Kondensatori: $Z_C = \frac{1}{sC}$

$$
H(s) = \frac{V_{out}(s)}{V_{in}(s)} = \frac{Z_C}{Z_R + Z_C} = \frac{\frac{1}{sC}}{R + \frac{1}{sC}} = \frac{1}{1 + sRC}
$$

Në analizën AC: $s = j\omega $

$$
H(j\omega) = \frac{1}{1 + j\omega RC}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

R = 1e3      # Ohm
C = 1e-6     # Farad

num = [1]
den = [R*C, 1]

system = signal.TransferFunction(num, den)
w, mag, phase = signal.bode(system)

f = w/2/np.pi

plt.figure()
plt.semilogx(f, mag)
plt.title('RC Low-pass - Magnituda')
plt.xlabel('Frekuenca (Hz)')
plt.ylabel('Gain [dB]')
plt.grid(True)

plt.figure()
plt.semilogx(f, phase)
plt.title('RC Low-pass - Faza')
plt.xlabel('Frekuenca (Hz)')
plt.ylabel('gradë')
plt.grid(True)

plt.show()
```

# Filtri RLC **kalim-bande** (band-pass)

<img src="figs/RCL_BandPass.svg" width=300/>

## **Analiza e qarkut**

Impedanca e paralelit LC është:

$$
Z_{LC}(s) = \left( \frac{1}{sL} + sC \right)^{-1} = \frac{sL}{s^2 LC + 1}
$$

Impedanca totale e qarkut:

$$
Z_{\text{tot}}(s) = R + Z_{LC}(s) = R + \frac{sL}{s^2 LC + 1}
$$

Funksioni i transferimit:

$$
H(s) = \frac{V_{out}(s)}{V_{in}(s)} = \frac{Z_{LC}(s)}{R + Z_{LC}(s)} = \frac{\frac{sL}{s^2 LC + 1}}{R + \frac{sL}{s^2 LC + 1}}
$$

Duke thjeshtuar:

$$
H(s) = \frac{sL}{s^2 LC + sR + 1}
$$

Tani marrim $s=j\omega$ dhe gjejme:
$$
H(\omega) = \frac{j\omega L}{j\omega L + R(1 - \omega^2 LC)}
$$

Kjo është formë standarde për një **filtri kalim-bande (band-pass)** me qendër në frekuencën:
$$
\omega_0 = \frac{1}{\sqrt{LC}}, \quad Q = \frac{1}{R} \sqrt{\frac{L}{C}}
$$


## 3. **Faktori i cilësisë $Q$**

Faktori i cilësisë për këtë konfigurim është:
$$
Q = \frac{\omega_0 L}{R} = \frac{1}{R} \sqrt{\frac{L}{C}}
$$

- $Q$ tregon **sa "e ngushtë" dhe e përqendruar është rezonanca**.
- Sa më e madhe të jetë $Q$, aq më pak energji humbet qarku gjatë një cikli → rezonancë më selektive.


## 4. **Gjerësia e brezit** $\Delta\omega$

Gjerësia e brezit përcaktohet si:

$$
\Delta\omega = \omega_2 - \omega_1
$$

ku $\omega_1$ dhe $\omega_2$ janë frekuencat për të cilat magnituda bie në $\frac{1}{\sqrt{2}} $ të vlerës maksimale → pra rënia me **-3 dB**.

Formula e lidhur me $Q$ është:

$$
\Delta\omega = \frac{\omega_0}{Q}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parametrat e komponentëve
R = 0.1         # Rezistencë në Ohm
L = 1e-2     # Induktancë në Henry
C = 1e-5      # Kapacitet në Farad

# Funksioni i transferimit për qarkun: R në seri me L||C, Vout merret mbi L||C
# Z_L = sL, Z_C = 1/(sC), Z_parallel = (Z_L * Z_C) / (Z_L + Z_C)
# H(s) = Z_parallel / (R + Z_parallel) = (sL)/(s^2*LC + sR + 1)

# H(s) = (sL)/(s^2*LC + sR + 1)
num = [L, 0]
den = [L*C, R, 1]
system = signal.TransferFunction(num, den)

# Bode plot
w, mag, phase = signal.bode(system)

f = w/2/np.pi
Q = np.sqrt(L/C)/R
f0 = 1/np.sqrt(L*C)/2/np.pi 

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.semilogx(f, mag)
plt.title(r'Magnituda - Band-pass ($Q=%.1f$, $f_0=%.1f$ Hz)'%(Q,f0))
plt.xlabel('Frekuenca [Hz]')
plt.ylabel('Gain [dB]')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.semilogx(f, phase)
plt.title('Faza - Band-pass: R - (L||C)')
plt.xlabel('Frekuenca [Hz]')
plt.ylabel('Faza [gradë]')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# RLC Band-block

<img src="figs/RCL_BandBlock.svg" width=300/>


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrat e komponentëve
R = 100         # Rezistencë (Ohm)
L = 10e-3       # Induktancë (H)
C = 1e-6        # Kapacitet (F)

# Frekuencat (omega) në rad/s
omega = np.logspace(2, 6, 1000)
f = omega/2/np.pi

# Impedancat përkatëse
jw = 1j * omega
Z_L = jw * L
Z_C = 1 / (jw * C)
Z_LC = Z_L + Z_C         # Seri LC
Z_R = R
Z_total = Z_R + Z_LC     # Total impedance

# Funksioni i transferimit: Vout merret mbi LC
# H(jw) = Z_LC / (R + Z_LC)
H = Z_LC / Z_total

# Magnituda (dB) dhe faza (gradë)
gain = 20 * np.log10(np.abs(H))
phase = np.angle(H, deg=True)

# Vizatim i grafikëve Bode
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.semilogx(f, gain)
plt.title('Magnituda - Band-block: R - (L + C)')
plt.xlabel('Frekuenca [Hz]')
plt.ylabel('Gain [dB]')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.semilogx(f, phase)
plt.title('Faza - Band-block: R - (L + C)')
plt.xlabel('Frekuenca [Hz]')
plt.ylabel('Faza [gradë]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

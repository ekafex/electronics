# **Seritë Fourier**  
Seria Fourier është një përfaqësim i një funksioni periodik si një **shumë e pafundme** e funksioneve trigonometrike (sinuse dhe kosinuse). Kjo metodë përdoret për të analizuar dhe sintetizuar sinjale periodike.

Një funksion periodik $f(t)$ me periodë $T$ mund të shprehet si:

$$ f(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t) \right) $$

ku $\omega_0 = \frac{2\pi}{T}$ është frekuenca themelore, $a_0,\; a_n,\; b_n $ janë koeficientët Fourier, të dhënë nga:

$$ a_0 = \frac{1}{T} \int_{0}^{T} f(t) dt $$
$$ a_n = \frac{2}{T} \int_{0}^{T} f(t) \cos(n\omega_0 t) dt $$
$$ b_n = \frac{2}{T} \int_{0}^{T} f(t) \sin(n\omega_0 t) dt $$


### **Shembull**: Seria Fourier e një Funksion Drejtkëndor  
Marrim funksionin periodik:

$$ f(t) =
\begin{cases} 
1, & 0 \leq t < T/2 \\
-1, & T/2 \leq t < T
\end{cases}
$$

Koeficientët Fourier për këtë valë katrore jepen nga:

$$ a_n = 0, \quad b_n = \frac{4}{n\pi} \sin\left(\frac{n\pi}{2}\right)$$

Duke përdorur këto koeficientë, valën katrore mund ta shprehim si:

$$ f(t) \approx \frac{4}{\pi} \left( \sin(\omega_0 t) + \frac{1}{3} \sin(3\omega_0 t) + \frac{1}{5} \sin(5\omega_0 t) + \dots \right) $$

Kjo është pikërisht ajo që **Transformimi Fourier** na mundëson të bëjmë.

### **Arsyet Kryesore për Përdorimin e Transformimit Fourier**

#### 1. **Shpërbërja e Sinjaleve Kompleksë në Komponentë të Thjeshtë**
Nëse vëzhgoni evolucionin e sinjaleve të ndryshme në jetën e përditëshme -- psh. valët zanore, sinjalet elektrike, imazhet etj -- si funksion i kohës duken shumë të ndërlikuar dhe kompleksë. Në kërkim të një analize më të thjeshtë të sinjalit, ne mund të transformojmë sinjalin nga funksion i kohës $f(t)$ në një funksion të një variabli të ri $\chi$. Për rrjedhojë funksioni i transformuar $F(\chi)$ paraqet $f(t)$ në hapësirën e $\chi$-ve. Nësë variabli $\chi$ është zgjedhur me kujdes, atëherë, $F(\chi)$ shfaq rregullsi dhe na lejon të kuptojmë më thellë sjelljen e $f(t)$-së. Transformimi Fourier është një nga këto transformime ku variabli $\chi$ do paraqes frekuencën e lëkundjes. Ky transformim na lejon t’i përshkruajmë sinjalet si shuma të **sinusoidave** (valëve sinus dhe kosinus), gjë që na mundëson jo vetëm analizimin më të thjeshtë të sinjalit, por edhe kryerjen e manipulimeve më lehtë mbi sinjalin.

#### 2. **Kuptimi i Përmbajtjes në Frekuencë**
- Në përpunimin e audios, na ndihmon të kuptojmë se cilat frekuenca janë të pranishme në një sinjal—e dobishme për ekuilibrimin, heqjen e zhurmave dhe kompresimin.
- Në elektronikë, ndihmon në projektimin e filtrave që lejojnë vetëm disa frekuenca të kalojnë, ndërsa bllokojnë të tjerat, p.sh., filtra kalues të lartë (high-pass) dhe të ulët (low-pass).
- Në fizikë, përdoret për të analizuar fenomenet lëkundëse dhe valore, nga mekanika klasike, optka deri te fenomenet kuantike.

#### 3. **Zgjidhja më e Lehtë e Ekuacioneve Diferenciale**
Shumë sisteme fizike përshkruhen nga ekuacione diferenciale, të cilat mund të jenë të vështira për t’u zgjidhur në domenin e kohës. Transformimi Fourier i shndërron këto ekuacione diferenciale në ekuacione algjebrike në domenin e frekuencës, duke i bërë ato më të lehta për t'u trajtuar.

#### 4. **Kompresimi dhe Transmetimi i të Dhënave**
- Formati **JPEG** për imazhe dhe **MP3** për audio mbështeten në transformimin e sinjaleve në domenin e frekuencës, duke eliminuar komponentët e panevojshëm për të zvogëluar madhësinë e skedarëve.
- Në sistemet e komunikimit, analiza Fourier është thelbësore për **modulimin dhe demodulimin e sinjaleve** në radio, televizion dhe rrjete wifi.

#### 5. **Përpunimi dhe Filtrimi i Sinjaleve**
Shumë komponentë të padëshiruar (si zhurma) në një sinjal shfaqen në frekuenca specifike. Duke përdorur filtrat në domenin e frekuencës, ne mund të heqim zhurmën ose të nxjerrim informacionin e dëshiruar në mënyrë efikase.

#### 6. **Aplikimet në Imazheri Mjekësore dhe Fizikë**
- **MRI** dhe **CT Scan** mbështeten në Transformimin Fourier për të rindërtuar imazhet nga të dhënat e papërpunuara.
- **Spektroskopia** në fizikë dhe kimi përdor analizën Fourier për të identifikuar materialet në bazë të vetive të tyre spektrale.

## **Transformimi i Fourier-it dhe Inversi i Tij**

Transformimi i Fourier-it i një funksioni $f(t)$ është i dhënë nga:

$$ F(\omega) = \int\limits_{-\infty}^{\infty} f(t)\, e^{-j\omega t}\, dt $$

Ky transformim shndërron një funksion nga domeni i kohës $t$ në domenin e frekuencës $\omega$, duke na dhënë një përfaqësim alternativ të sinjalit.

Inversi i Transformimit Fourier është:

$$f(t) = \frac{1}{2\pi} \int\limits_{-\infty}^{\infty} F(\omega)\, e^{j\omega t}\, d\omega $$

Ky integral na lejon të rikthejmë sinjalin origjinal nga përfaqësimi në domenin e frekuencës.

### **Vetitë Kryesore të Transformimit të Fourier-it**
1. **Lineariteti**
   - Nëse kemi dy funksione $f(t)$ dhe $g(t)$, atëherë:
   $$ \mathcal{F}\{ a\, f(t) + b\, g(t) \} = a\, F(\omega) + b\, G(\omega) $$
   ku $\mathcal{F}$ paraqet transformimin e Fourier-it.

2. **Përhapja në Kohë dhe Frekuencë**
   - Një sinjal i ngushtuar në kohë ka një spektër të gjerë në frekuencë dhe anasjelltas.

3. **Teorema e Konvolucionit**
   - Konvolucioni i dy funksioneve në domenin e kohës bëhet një shumëzim në domenin e frekuencës:
   $$\mathcal{F}\{ f(t) * g(t) \} = F(\omega)\, G(\omega)$$

4. **Derivimi dhe Integrimi**
   - Transformimi i Fourier-it e shndërron derivatin në një shumëzim me frekuencën:
   $$ \mathcal{F} \left\{ \frac{d}{dt} f(t) \right\} = j\, \omega\, F(\omega) $$

5. **Zhvendosja në Kohë**
   - Një zhvendosje në kohë përkthehet në një shumëzim eksponencial në domenin e frekuencës:
   $$ \mathcal{F}\{ f(t - t_0) \} = e^{-j\, \omega\, t_0}\, F(\omega) $$

---

### **Tabela e Funksioneve më të Përdorura dhe Transformimet e Tyre Fourier**

| Funksioni në kohë $f(t)$ | Transformimi Fourier $F(\omega)$ |
|-------------------------------|---------------------------------|
| $\delta(t)$ (Delta e Dirakut) | $1$ |
| $1$ (Funksioni konstant) | $ 2\pi \delta(\omega) $ |
| $ e^{-at} u(t) $, $ a > 0 $ | $ \frac{1}{a + j\omega} $ |
| $ \cos(\omega_0 t) $ | $ \pi [\delta(\omega - \omega_0) + \delta(\omega + \omega_0)] $ |
| $ \sin(\omega_0 t) $ | $ j\pi [\delta(\omega - \omega_0) - \delta(\omega + \omega_0)] $ |
| Funksioni drejtkëndor $ \mathsf{rect}(t/T) $ | $ T \text{sinc}(T\omega) $ |
| Funksioni Gaussian $ e^{-t^2} $ | $ \sqrt{\pi} e^{-\omega^2/4} $ |

## **Përfundim**
Transformimi i Fourier-it është një mjet themelor që **ndryshon mënyrën se si e shohim një sinjal**—nga analiza në domenin e kohës në analizën në domenin e frekuencës. Ky këndvështrim i dyfishtë na jep njohuri të fuqishme dhe aplikime praktike në inxhinieri, fizikë dhe shumë fusha të tjera.



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# Përkufizimi i domenit të kohës
T = 1.0  # Periudha totale e kohës
N = 500  # Numri i mostrave
t = np.linspace(0, T, N, endpoint=False)

# Përkufizimi i një funksioni: Shuma e sinusoideve
f = np.sin(2 * np.pi * 5 * t) + 0.8 * np.sin(2 * np.pi * 10 * t)

# Llogaritja e Transformimit të Fourier-it
F = fft(f)
frequencies = fftfreq(N, T/N)

# Grafiku i sinjalit në domenin e kohës
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t, f)
plt.title("Sinjali në Domenin e Kohës")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")

# Grafiku i spektrit të frekuencës
plt.subplot(1, 2, 2)
plt.plot(frequencies[:N//2], np.abs(F[:N//2]))
plt.title("Spektri i Frekuencës (Transformimi i Fourier-it)")
plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Madhësia")
plt.xlim(0,30)

plt.tight_layout()
plt.show()
````


# **Shembuj Transformimesh Fourier**

### **Shembulli 1: Puls Drejtkëndor**
$$
f(t) =
\begin{cases} 
1, & |t| \leq \tau/2 \\
0, & \text{ndryshe}
\end{cases}
$$

Duam të gjejmë **Transformimin Fourier**:

$$ F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-j\omega t} dt$$

Zëvendësojmë $f(t)$ në integral:
$$ F(\omega) = \int_{-\tau/2}^{\tau/2} e^{-j\omega t} dt$$

Ky është një integral i thjeshtë eksponencial:

$$ F(\omega) = \left[ \frac{e^{-j\omega t}}{-j\omega} \right]_{-\tau/2}^{\tau/2}$$

Zëvendësojmë kufijtë:
$$F(\omega) = \frac{e^{-j\omega (\tau/2)} - e^{j\omega (\tau/2)}}{-j\omega}$$

Përdorim formulën e sinuseve komplekse:
$$ e^{-j\theta} - e^{j\theta} = -2j \sin(\theta) $$

$$ F(\omega) = \frac{-2j \sin(\omega \tau /2)}{-j\omega}$$

Përfundimisht marrim:
$$ F(\omega) = \tau \text{sinc}(\omega \tau /2)$$

ku **sinc(x)** është funksioni:
$$ \text{sinc}(x) = \frac{\sin(x)}{x}$$

---

### **Shembulli 2: Sinjali Eksponencial**
$$ f(t) = e^{-at} \theta(t), \quad a > 0 $$
ku **$\theta(t)$ është funksioni Heaviside** që është $\theta(t)=0$ për $t<0$ dhe $\theta(t)=1$ për $t>0$.
$$ F(\omega) = \int_{-\infty}^{\infty} e^{-at} \theta(t) e^{-j\omega t} dt $$

Përshkak të $\theta(t)$ integrali bëhet:
$$ F(\omega) = \int_{0}^{\infty} e^{-(a + j\omega)t} dt $$

Ky është një integral i formës:
$$ \int e^{-\lambda t} dt = \frac{e^{-\lambda t}}{-\lambda} $$

ku $\lambda = a + j\omega$. Aplikojmë kufijtë:
$$ F(\omega) = \left[ \frac{e^{-(a + j\omega)t}}{-(a + j\omega)} \right]_{0}^{\infty} $$

Në $t = \infty$, eksponenciali bëhet zero, dhe në $t = 0$ mbetet:
$$ F(\omega) = \frac{1}{a + j\omega} $$

Transformimi Fourier për funksionin eksponencial është:
$$ F(\omega) = \frac{1}{a + j\omega}, \quad a > 0 $$

---

### **Shembulli 3: Sinjal Sinusoidal**
Le të marrim funksionin:
$$ f(t) = \cos(\omega_0 t) $$

Transformimi i tij është:
$$ F(\omega) = \int_{-\infty}^{\infty} \cos(\omega_0 t) e^{-j\omega t} dt $$

Përdorim formulën e Eulerit:
$$ \cos(\omega_0 t) = \frac{e^{j\omega_0 t} + e^{-j\omega_0 t}}{2} $$
Prandaj:
$$ F(\omega) = \int_{-\infty}^{\infty} \frac{e^{j\omega_0 t} + e^{-j\omega_0 t}}{2} e^{-j\omega t} dt $$

Shpërndajmë eksponencialet:
$$ F(\omega) = \frac{1}{2} \int_{-\infty}^{\infty} e^{-j(\omega - \omega_0) t} dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j(\omega + \omega_0) t} dt $$

Dihet që:
$$ \int_{-\infty}^{\infty} e^{-j\alpha t} dt = 2\pi \delta(\alpha) $$

Prandaj:
$$ F(\omega) = \frac{1}{2} \left[ 2\pi \delta(\omega - \omega_0) + 2\pi \delta(\omega + \omega_0) \right] $$

Përgjigja përfundimtare:
$$ F(\omega) = \pi \left[ \delta(\omega - \omega_0) + \delta(\omega + \omega_0) \right] $$

**Interpretimi**: Transformimi i Fourier-it i një sinusoidi është **dy impulse Dirac** në $\pm \omega_0$, që tregon se një sinjal sinusoidal ka vetëm **dy përbërës frekuencorë**.

### **Shembulli 4: Funksioni Gaussian**
Funksioni Gaussian është një nga rastet më interesante pasi **transformimi i tij është përsëri një Gaussian**.
$$ f(t) = e^{-a t^2}, \quad a > 0 $$
Duam të llogarisim:
$$ F(\omega) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j\omega t} dt $$
Kombinojmë eksponentët:
$$ F(\omega) = \int_{-\infty}^{\infty} e^{-a t^2 - j\omega t} dt $$

Përdorim formulën e plotësimit të katrorit:
$$ a t^2 + j\omega t = a \left( t^2 + \frac{j\omega}{a} t \right) $$

Duke shtuar dhe zbritur $\left(\frac{j\omega}{2a}\right)^2$:
$$ a \left(t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a} $$

Rikrijojmë eksponencialin:
$$ e^{-a t^2 - j\omega t} = e^{-a \left( t + \frac{j\omega}{2a} \right)^2} e^{-\frac{\omega^2}{4a}} $$

Integrali i parë është një integral Gaussian standard:
$$ \int_{-\infty}^{\infty} e^{-a(t+b)^2} dt = \sqrt{\frac{\pi}{a}} $$

Përgjigja përfundimtare:
$$ F(\omega) = \sqrt{\frac{\pi}{a}} e^{-\frac{\omega^2}{4a}} $$

**Rezultati**: **Transformimi i Fourier-it i një Gaussian është një tjetër Gaussian!** Kjo është një pronë unike që e bën Gaussian-in shumë të përdorshëm në përpunimin e sinjaleve.

---

### **Shembulli 5: Funksioni Delta Dirac**
Delta e Diracut është një funksion i veçantë që vepron si një shpërndarje matematikore dhe ka një **transformim shumë të thjeshtë**.
$$ f(t) = \delta(t) $$

Transformimi Fourier jepet nga:
$$ F(\omega) = \int_{-\infty}^{\infty} \delta(t) e^{-j\omega t} dt $$
Nga vetia e funksionit Delta:
$$ \int_{-\infty}^{\infty} \delta(t) g(t) dt = g(0) $$

Zëvendësojmë $g(t) = e^{-j\omega t}$:
$$ F(\omega) = e^{-j\omega \cdot 0} = 1 $$

**Rezultati**:
$$ F(\omega) = 1 $$

Pra, transformimi Fourier i një **impulsi Dirac është një sinjal konstant**. Kjo përputhet me faktin se një impuls përmban të gjitha frekuencat në mënyrë të barabartë.

# **Transformimi i Laplace-it**  

Transformimi i Laplace-it është një teknikë matematikore e përdorur për **analizën e sistemeve dinamike**, **zgjidhjen e ekuacioneve diferenciale** dhe **analizën e sinjaleve**. Ai e shndërron një funksion në domenin e kohës $f(t)$ në një funksion në domenin e frekuencës komplekse $F(s)$.  

## **1. Përkufizimi**  
Transformimi i Laplace-it i një funksioni $f(t)$ është i dhënë nga:  
$$ F(s) = \mathcal{L} \{ f(t) \} = \int_0^\infty f(t) e^{-st} dt $$

ku **$s$** është një numër kompleks $s = \sigma + j\omega$, ku $\sigma$ është pjesa reale dhe $\omega$ është pjesa imagjinare. **$F(s)$** është transformimi i Laplace-it në domenin e frekuencës komplekse.  

## **2. Transformimi i Inversit të Laplace-it**  
Për të rikthyer $f(t)$ nga $F(s)$, përdoret **transformimi i kundërt i Laplace-it**:
$$ f(t) = \mathcal{L}^{-1} \{ F(s) \} = \frac{1}{2\pi j} \int_{\sigma - j\infty}^{\sigma + j\infty} F(s) e^{st} ds $$

Ky integral kërkon **mjetet e analizës komplekse** për të gjetur $f(t)$.

## **3. Vetitë Kryesore**
Transformimi i Laplace-it ka **disa veti të dobishme**:

| **Vetitë** | **Shprehja Matematike** |
|------------|------------------------|
| Lineariteti | $\mathcal{L} \{ a f(t) + b g(t) \} = a F(s) + b G(s)$ |
| Derivimi |$ \mathcal{L} \{ f'(t) \} = sF(s) - f(0) $ |
| Integrimi | $ \mathcal{L} \left\{ \int_0^t f(\tau) d\tau \right\} = \frac{F(s)}{s} $ |
| Zhvendosja në kohë | $ \mathcal{L} \{ f(t - T) u(t - T) \} = e^{-sT} F(s) $ |
| Teorema e Konvolucionit | $ \mathcal{L} \{ f(t) * g(t) \} = F(s) G(s) $ |


## **4. Transformime të Disa Funksioneve të Njohura**

| **Funksioni në kohë $f(t)$** | **Transformimi i Laplace-it $F(s)$** |
|----------------------------------|------------------------------------|
| $1$ | $\frac{1}{s}$ |
| $e^{-at}$ | $\frac{1}{s + a} $ |
| $t$ | $\frac{1}{s^2}$ |
| $\sin(\omega t)$ | $\frac{\omega}{s^2 + \omega^2}$ |
| $\cos(\omega t)$ | $\frac{s}{s^2 + \omega^2}$ |


## **5. Dobia e Transformimit të Laplace-it**

- **Zgjidhja e ekuacioneve diferenciale** → Kthimi i ekuacioneve diferenciale në ekuacione algjebrike në domenin $s$.  
- **Analiza e sistemeve dinamike** → Përdoret në **teorinë e kontrollit**, **qarqet elektrike**, dhe **mekanikën e sistemit të vibrimeve**.  
- **Filtrimi dhe analizimi i sinjaleve** → Përdoret për të **analizuar sinjale dhe sisteme në elektronikë**.  
- **Stabiliteti i sistemeve** → Analiza e **poleve dhe zerove** në domenin $s$.

# **Përmbledhje e Transformimit Z**  

Transformimi **Z** është një mjet matematikor i përdorur për **analizën e sistemeve diskrete** dhe **përpunimin e sinjaleve dixhitale**. Ai shndërron një sinjal diskret $x[n]$ në një përfaqësim algjebrik në domenin e frekuencës komplekse $X(z)$, duke lehtësuar **analizën dhe zgjidhjen e ekuacioneve të ndryshme**. Transformimi **Z** është versioni diskret is transformimit të Laplace-it.

## **1. Përkufizimi**  
Transformimi **Z** i një funksioni diskret \( x[n] \) është i dhënë nga:  
$$ X(z) = \mathcal{Z} \{ x[n] \} = \sum_{n=-\infty}^{\infty} x[n] z^{-n} $$

ku: **$x[n]$** është sinjali diskret në domenin e kohës. **$z$** është një numër kompleks. **$X(z)$** është transformimi Z në domenin e frekuencës komplekse.  

## **2. Transformimi i Kundërt i Z**
Për të rikthyer $x[n]$ nga $X(z)$, përdoret **integrali i konturit**:
$$ x[n] = \frac{1}{2\pi j} \oint_C X(z) z^{n-1} dz $$

ku **konturi $C$** është një rreth i mbyllur brenda rajonit të konvergjencës së $X(z)$.

## **3. Vetitë Kryesore të Transformimit Z**
Transformimi **Z** ka **veti të ngjashme me Transformimin e Laplace-it**, por i aplikuar për sinjale diskrete:

| **Vetitë** | **Shprehja Matematike** |
|------------|------------------------|
| Lineariteti | $\mathcal{Z} \{ a x_1[n] + b x_2[n] \} = a X_1(z) + b X_2(z)$ |
| Zhvendosja në kohë | $\mathcal{Z} \{ x[n - k] \} = z^{-k} X(z)$ |
| Derivimi në domenin Z | $\mathcal{Z} \{ n x[n] \} = -z \frac{dX(z)}{dz}$ |
| Teorema e Konvolucionit | $\mathcal{Z} \{ x[n] * h[n] \} = X(z) H(z)$ |
| Vlera Fillestare | $x[0] = \lim_{z \to \infty} X(z)$ |


## **4. Transformime të Disa Funksioneve të Njohura**
| **Funksioni Diskret $x[n]$** | **Transformimi Z $X(z)$** | **Rajoni i Konvergjencës (ROC)** |
|----------------------------------|---------------------------------|-------------------------------|
| $\delta[n]$ | $1$ | $|z| > 0$ |
| $u[n]$ | $\frac{1}{1 - z^{-1}}$ | $|z| > 1$ |
| $a^n u[n]$ | $\frac{1}{1 - az^{-1}}$ | $|z| > |a|$ |
| $n u[n]$ | $\frac{z^{-1}}{(1 - z^{-1})^2}$ | $|z| > 1$ |
| $\cos(\omega_0 n) u[n]$ | $\frac{1 - z^{-1} \cos(\omega_0)}{1 - 2z^{-1} \cos(\omega_0) + z^{-2}}$ | $ |z| > 1$ |


## **5. Dobia e Transformimit Z?**
- **Analiza e sistemeve dixhitale** → Përdoret në **sistemet e përpunimit të sinjaleve** dhe **sistemet e kontrollit dixhital**.  
- **Zgjidhja e ekuacioneve me ndryshorë diskretë** → Lejon transformimin e ekuacioneve të diferencave në ekuacione algjebrike në domenin **Z**.  
- **Filtrimi dixhital** → Përdoret në projektimin dhe analizën e **filtrave dixhitalë (IIR, FIR)**.  
- **Stabiliteti i sistemeve** → Analiza e **poleve dhe zerove** në planin $z$ për të përcaktuar **stabilitetin e sistemeve dixhitale**.  


### **6. Lidhja midis Transformimit Z dhe Transformimit të Laplace-it**
Transformimi **Z** është ekuivalenti diskret i **Transformimit të Laplace-it**. Lidhja e tyre vjen përmes **zëvendësimit**:
$$ z = e^{sT}$$
ku $T$ është periudha e mostrimit në një sistem diskret. Ky zëvendësim lidhet me **diskretizimin e sistemeve të vazhdueshme**.

# **Përmbledhje e Konvolucionit**  

Konvolucioni është një operacion matematikor i përdorur gjerësisht në **përpunimin e sinjaleve, analizën e sistemeve dhe zgjidhjen e ekuacioneve diferenciale**. Ai përshkruan sesi një **sinjal (ose funksion)** modifikohet nga një **sistem** dhe është thelbësor në **analizën e sinjaleve, filtrimin dhe teorinë e sistemeve**.

## **1. Përkufizimi i Konvolucionit**
### **Konvolucioni i Sinjaleve të Vazhdueshme**
Për dy funksione të kohës së vazhdueshme $f(t)$ dhe $g(t)$, konvolucioni përcaktohet si:
$$ (f * g)(t) = \int_{-\infty}^{\infty} f(\tau) g(t - \tau) d\tau $$

ku: **$f(t)$** është sinjali hyrës, **$g(t)$** është përgjigja impulsive e sistemit, **$(f * g)(t)$** është sinjali dalës i sistemit.

### **Konvolucioni i Sinjaleve Diskrete**
Për sinjale diskrete $x[n]$ dhe $h[n]$, konvolucioni përcaktohet si:
$$ (y[n] = x[n] * h[n] = \sum_{k=-\infty}^{\infty} x[k] h[n - k]) $$

ku: **$x[n]$** është hyrja e sistemit, **$h[n]$** është përgjigja impulsive, **$y[n]$** është dalja e sistemit.

## **2. Vetitë Kryesore të Konvolucionit**
Konvolucioni ka disa veti të rëndësishme:

| **Vetitë** | **Shprehja Matematike** |
|------------|------------------------|
| **Komutativiteti** | $f * g = g * f$ |
| **Asociativiteti** | $(f * g) * h = f * (g * h)$ |
| **Distributiviteti** | $f * (g + h) = (f * g) + (f * h)$ |
| **Përhapja në kohë** | $f(t - t_0) * g(t) = (f * g)(t - t_0)$ |
| **Konvolucioni me një Delta Dirac** | $f(t) * \delta(t) = f(t)$ |

## **3. Lidhja me Transformimet Matematike**
Konvolucioni ka një **lidhje të drejtpërdrejtë me Transformimet e Fourier-it dhe të Laplace-it**:

| Transformimi | Lidhja me Konvolucionin |
|-------------|-------------------------|
| **Transformimi i Fourier-it** | $\mathcal{F} \{ f * g \} = \mathcal{F} \{ f \} \cdot \mathcal{F} \{ g \}$ |
| **Transformimi i Laplace-it** | $\mathcal{L} \{ f * g \} = \mathcal{L} \{ f \} \cdot \mathcal{L} \{ g \}$ |
| **Transformimi Z** | $\mathcal{Z} \{ x[n] * h[n] \} = \mathcal{Z} \{ x[n] \} \cdot \mathcal{Z} \{ h[n] \}$ |


## **4. Përdorime Kryesore të Konvolucionit**
- **Analiza e Sistemeve Linearë dhe të Pandryshueshëm në Kohë (LTI)**  
  Konvolucioni përcakton daljen e një sistemi LTI për një hyrje të caktuar.
  
- **Filtrimi në Përpunimin e Sinjaleve**  
  Në filtrat dixhitalë dhe analogë, konvolucioni modelon sesi një sinjal ndikohet nga një filtër.

- **Vizion Artificial dhe Përpunimi i Imazheve**  
  Operacionet e turbullimit, skicimit dhe përmirësimit të imazheve kryhen me konvolucion.

- **Zgjidhja e Ekuacioneve Diferenciale**  
  Konvolucioni përdoret për të gjetur përgjigjet e sistemeve dinamike.

- **Analiza e Sinjaleve në Komunikime**  
  Sinjalet e transmetuara kalojnë përmes kanaleve që mund të modelohen me konvolucion.

## **5. Shembull i Konvolucionit në Domenin e Kohës**
Le të marrim dy sinjale:
$$ f(t) = e^{-t} \theta(t), \quad g(t) = \theta(t) $$

Ku **$\theta(t)$** është funksioni Heaviside.

### **Hapat e Llogaritjes**
1. **Zëvendësojmë në formulën e konvolucionit**:
$$ (f * g)(t) = \int_0^t e^{-\tau} d\tau $$

2. **Llogarisim integralin**:
$$ (f * g)(t) = \left[ -e^{-\tau} \right]_0^t = 1 - e^{-t} $$

3. **Përfundimi**:
$$ y(t) = (1 - e^{-t}) \theta(t) $$

Ky është një **sinjal i butë**, që fillon në zero dhe rritet gradualisht në 1.


## **6. Shembull i Konvolucionit në Domenin e Frekuencës**
Nëse kemi:
$$ f(t) = e^{-at} \theta(t), \quad g(t) = e^{-bt} \theta(t) $$

Në **Transformimin e Laplace-it**, kemi:
$$ \mathcal{L} \{ f(t) \} = \frac{1}{s + a}, \quad \mathcal{L} \{ g(t) \} = \frac{1}{s + b} $$

Meqenëse konvolucioni në domenin e kohës bëhet **shumëzim në domenin e frekuencës**:
$$ Y(s) = \frac{1}{(s + a)(s + b)}$$

Ky është një **filtër kalues i ulët** në analizën e sinjaleve.


### **Përfundim**
- **Konvolucioni** është një operacion themelor në analizën e sinjaleve dhe sistemeve.
- **Në domenin e kohës**, ai modelon sesi një sinjal transformohet nga një sistem.
- **Në domenin e frekuencës**, ai bëhet një shumëzim i thjeshtë.
- **Ka aplikime të gjera** në **përpunimin e sinjaleve, filtrimin dixhital, analizën e sistemeve dhe fizikë**.


### **7. Konvolucioni në Python**
Mund të llogarisim konvolucionin e dy sinjaleve diskrete në **Python** duke përdorur `numpy.convolve`:

````python
import numpy as np
import matplotlib.pyplot as plt

# Sinjalet hyrëse
x = np.array([1, 2, 3, 4])
h = np.array([0.2, 0.5, 0.2])

# Konvolucioni
y = np.convolve(x, h, mode='full')

# Vizualizimi
plt.stem(y)
plt.title("Konvolucioni i dy sinjaleve")
plt.xlabel("Koha (n)")
plt.ylabel("Amplituda")
plt.grid()
plt.show()
````

# Mesatarizimi i Sinjaleve Harmonikë

Le të kemi një **sinjal harmonik** të përgjithshëm të formës:

$$f(t) = A \cos(\omega t + \phi)$$

Ku $A$ është **amplituda**, $\omega$ është **frekuenca këndore** ($\omega = 2\pi f$), $\phi$ është **faza fillestare**. Mesatarja në kohë mbi një periudhë $T$ jipet nga:

$$\langle f(t) \rangle = \frac{1}{T} \int_{0}^{T} f(t) dt$$

Duke zëvendësuar $f(t)$:
$$\langle f(t) \rangle = \frac{1}{T} \int_{0}^{T} A \cos(\omega t + \phi) dt$$

Duke përdorur faktin se mesatarja e një kosinusi mbi një periudhë të plotë është **zero**, përftojmë:
$$\langle f(t) \rangle = 0$$

Kjo do të thotë se sinjalet harmonike **nuk kanë komponent mesatar** kur mesatarizohen në një periudhë të plotë.


### **Mesatarja e Fuqisë së Sinjalit**
Megjithatë, në shumë raste na intereson **fuqia mesatare** e një sinjali, e cila është e dhënë nga mesatarja kohore e $f^2(t)$:

$$\langle f^2(t) \rangle = \frac{1}{T} \int_{0}^{T} A^2 \cos^2(\omega t + \phi) dt$$

Duke përdorur identitetin trigonometrik:

$$\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}$$

kemi:

$$\langle f^2(t) \rangle = \frac{A^2}{T} \int_{0}^{T} \frac{1 + \cos(2\omega t + 2\phi)}{2} dt$$

Integrali i $\cos(2\omega t + 2\phi)$ në një periudhë të plotë është **zero**, kështu që mbetet:

$$\langle f^2(t) \rangle = \frac{A^2}{2}$$

Pra si përfundim: **Mesatarja kohore e një sinjali harmonik është gjithmonë zero** mbi një periudhë të plotë. **Mesatarja e fuqisë së sinjalit është $A^2/2$**, një rezultat i rëndësishëm për analizën e fuqisë së sinjaleve në elektronikë, mekanikë dhe përpunim sinjalesh.



### **Mesatarja e Produktit të Dy Sinjaleve Harmonike**

Le të kemi dy sinjale harmonike:

$$f_1(t) = A_1 \cos(\omega_1 t + \phi_1)$$

$$f_2(t) = A_2 \cos(\omega_2 t + \phi_2)$$

Produkti i këtyre dy sinjaleve është:

$$f_1(t) f_2(t) = A_1 A_2 \cos(\omega_1 t + \phi_1) \cos(\omega_2 t + \phi_2)$$

Mesatarja kohore mbi një periudhë $T$ jepet nga:

$$\langle f_1(t) f_2(t) \rangle = \frac{1}{T} \int_{0}^{T} A_1 A_2 \cos(\omega_1 t + \phi_1) \cos(\omega_2 t + \phi_2) dt$$

Përdorim identitetin trigonometrik:

$$\cos(\alpha) \;\cos(\beta) = \frac{1}{2} [\cos(\alpha - \beta) + \cos(\alpha + \beta)]$$

Duke e aplikuar këtë për rastin tonë:

$$\cos(\omega_1 t + \phi_1) \cos(\omega_2 t + \phi_2) = \frac{1}{2} \left[\cos((\omega_1 - \omega_2)t + (\phi_1 - \phi_2)) + \cos((\omega_1 + \omega_2)t + (\phi_1 + \phi_2)) \right]$$

Duke e zëvendësuar këtë në shprehjen e mesatares kohore:

$$\langle f_1(t) f_2(t) \rangle = \frac{A_1 A_2}{2T} \int_{0}^{T} \left[ \cos((\omega_1 - \omega_2)t + (\phi_1 - \phi_2)) + \cos((\omega_1 + \omega_2)t + (\phi_1 + \phi_2)) \right] dt$$

Integrali i një funksioni kosinus mbi një periudhë të plotë është **zero**, përveç rasteve kur argumenti i tij është konstant, pra kur:

1. $\omega_1 = \omega_2$ (frekuencat janë të barabarta)
2. $\omega_1 \neq \omega_2$ (frekuencat janë të ndryshme)

Për rastin **$\omega_1 \neq \omega_2$**, të dy kosinusët përmbajnë frekuenca jozero dhe mesatarja bëhet **zero**, pra:

$$\langle f_1(t) f_2(t) \rangle = 0$$

Kjo do të thotë se dy sinjale harmonike me **frekuenca të ndryshme** nuk kanë një komponent konstant mesatar kur shumëzohen dhe mesatarizohen në kohë.

Për rastin **$\omega_1 = \omega_2$**, kemi:

$$\langle f_1(t) f_2(t) \rangle = \frac{A_1 A_2}{2} \cos(\phi_1 - \phi_2)$$

Ky është një rezultat shumë i rëndësishëm: kur dy sinjale kanë **të njëjtën frekuencë**, mesatarja e produktit të tyre është **gjysma e produktit të amplitudave të tyre të shumëzuar me kosinusin e ndryshimit të fazës**.

Për ti përmbledhur:
1. **Nëse $\omega_1 \neq \omega_2$** → mesatarja kohore e produktit është **zero**.
2. **Nëse $\omega_1 = \omega_2$** → mesatarja kohore është:

   $$\frac{A_1 A_2}{2} \cos(\phi_1 - \phi_2)$$

Kujtoni se integrali i formës (për $T \to \infty$):
$$ \int_0^T \cos(\omega t + \phi) dt =
\begin{cases}
0,           & \text{nëse } \omega_1 \neq \omega_2 \\
T \cos(\phi), & \text{nëse } \omega_1 = \omega_2
\end{cases}$$


Ky rezultat ka **shumë aplikime**, për shembull:
- **Në sinjalistikë dhe komunikime** – për të përcaktuar se si sinjalet me frekuenca të ndryshme ndërveprojnë.
- **Në analizën e fuqisë së sinjaleve AC** – ku komponentët me frekuenca të ndryshme nuk kontribuojnë në fuqinë mesatare.
- **Në interferencën e valëve** – për të kuptuar se si dy valë me të njëjtën frekuencë interferojnë me njëra-tjetrën.

Pra:
$$ \langle f_1(t) f_2(t) \rangle =
\begin{cases}
0, & \text{nëse } \omega_1 \neq \omega_2 \\
\frac{A_1 A_2}{2} \cos(\phi_1 - \phi_2), & \text{nëse } \omega_1 = \omega_2
\end{cases}
$$

Ky rezultat ka **shumë aplikime**, për shembull:
- **Në sinjalistikë dhe komunikime** – për të analizuar se si sinjalet me frekuenca të ndryshme ndërveprojnë.
- **Në analizën e fuqisë së sinjaleve AC** – ku komponentët me frekuenca të ndryshme nuk kontribuojnë në fuqinë mesatare.
- **Në ndërhyrjen e valëve** – për të kuptuar se si dy valë me të njëjtën frekuencë interferojnë me njëra-tjetrën.



# **Shkalla dB (Decibel)**  

**Shkalla decibel (dB)** është një mënyrë logaritmike për të matur raportin e dy madhësive, shpesh përdorur për të shprehur **fuqinë, tensionin, intensitetin e zërit** etj, në elektronikë, akustikë dhe telekomunikacion...

## **1. Përkufizimi i Decibel-it**  
Një madhësi **$X$** shprehet në **dB** duke përdorur një shprehje logaritmike:

- **Për fuqi (P)**:
  $$ L_{\text{dB}} = 10 \log_{10} \left( \frac{P}{P_{\text{ref}}} \right) $$

- **Për tension (V) ose rrymë (I)**:
  $$ L_{\text{dB}} = 20 \log_{10} \left( \frac{V}{V_{\text{ref}}} \right) $$

ku **$P_{\text{ref}}$** ose **$V_{\text{ref}}$** janë vlerat referuese.

## **2. Pse Përdoret Shkalla dB?**
- **Përfaqëson raporte shumë të mëdha ose të vogla në një format kompakt**.
- **Përputhet me mënyrën se si perceptojmë zërin dhe intensitetin e valëve**.
- **Lehtëson llogaritjet për sistemet kaskadë**, pasi shumëzimi kthehet në mbledhje.

## **3. Shembuj të Zakonshëm**
| Raporti i Fuqisë $P/P_{\text{ref}}$ | Fuqia në dB |
|----------------------|------------|
| $10^3$ (1000x)  | 30 dB |
| $10^2$ (100x)   | 20 dB |
| $10^1$ (10x)    | 10 dB |
| $10^0$ (1x)     | 0 dB |
| $10^{-1}$ (0.1x) | -10 dB |
| $10^{-2}$ (0.01x) | -20 dB |

## **4. Vlera Referuese të Njohura**
Në shumë aplikime, përdoren **vlera standarde referuese**:

- **dBm** – fuqia në raport me **1 mW**.
- **dBW** – fuqia në raport me **1 W**.
- **dBV** – tensioni në raport me **1 V**.
- **dBu** – tensioni në raport me **0.775 V** (standard audio).

### **5. Shembull Praktik**
**Një sinjal ka tensionin $V = 2V$ dhe referenca është $V_{\text{ref}} = 1V$. Cila është vlera në dB?**
$$ L_{\text{dB}} = 20 \log_{10} \left( \frac{2V}{1V} \right) = 20 \log_{10} (2) = 6.02 \text{ dB} $$

### **6. Përfundim**
- **Shkalla dB** është **logaritmike**, që e bën atë ideale për të trajtuar **raporte të mëdha**.
- **dB përdoret në shumë fusha**, përfshirë **elektronikën, inxhinierinë e komunikimeve dhe akustikën**.
- **10 dB ≈ 10x në fuqi, ndërsa 20 dB ≈ 10x në tension**.

## **Ushtrimi 1: Mesatarizimi në Kohë i Një Sinjali Sinusoidal**
**Qëllimi**: Të verifikohet eksperimentalisht që mesatarja kohore e një sinjali sinusoidal është **zero**, por mesatarja e fuqisë nuk është.

**Udhëzime**:
1. Gjenero një sinjal sinusoidal me frekuencë $f = 5 Hz$ dhe amplitudë $A = 2$.
2. Llogarit mesataren kohore të sinjalit në një periudhë të plotë.
3. Llogarit mesataren e fuqisë $f^2(t)$.

**Kodi Python**:
```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrat e sinjalit
A = 2
f = 5  # Hz
T = 1 / f  # Periudha
t = np.linspace(0, 2*T, 1000)  # 2 periudha

# Gjenerimi i sinjalit
signal = A * np.cos(2 * np.pi * f * t)

# Llogaritja e mesatares kohore dhe fuqisë
mean_signal = np.mean(signal)
mean_power = np.mean(signal**2)

# Vizualizimi i sinjalit
plt.plot(t, signal, label="Sinjali")
plt.axhline(mean_signal, color='r', linestyle='--', label=f"Mesatarja = {mean_signal:.2f}")
plt.title("Mesatarja Kohore e një Sinjali Sinusoidal")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")
plt.legend()
plt.show()

print(f"Mesatarja e sinjalit: {mean_signal:.4f}")
print(f"Mesatarja e fuqisë: {mean_power:.4f} (pritet A^2/2 = {A**2/2:.2f})")
```
**Pritet që** mesatarja të jetë **afërsisht zero**, ndërsa mesatarja e fuqisë të jetë $A^2/2$.

**Qëllimi**: Të tregohet se mesatarja e produktit të dy sinjaleve sinusoidale është **zero**, përveç rasteve kur ato kanë të njëjtën frekuencë.

**Udhëzime**:
1. Gjenero dy sinjale sinusoidale:
   - Njëri me frekuencë $f_1 = 5 Hz$.
   - Tjetri me $f_2 = 10 Hz$.
2. Llogarit mesataren e produktit të tyre për frekuenca të ndryshme.

# Parametrat e sinjaleve
A1, A2 = 2, 3
f1, f2 = 5, 10  # Hz

signal1 = A1 * np.cos(2 * np.pi * f1 * t)

mean_product = np.mean(product_signal)

plt.plot(t, signal1, label="Sinjali 1")
plt.plot(t, signal2, label="Sinjali 2")

## **Ushtrimi 3: Transformimi i Fourier-it dhe Analiza e Spektrit**
**Qëllimi**: Të gjenerohet një sinjal sinusoidal dhe të vëzhgohet spektri i tij duke përdorur **FFT (Fast Fourier Transform)**.

**Udhëzime**:
1. Gjenero një sinjal sinusoidal me frekuencë $f = 10 Hz$.
2. Përdor **FFT** për të analizuar spektrin e tij.
3. Shfaq sinjalin dhe spektrin.

**Kodi Python**:
```python
from scipy.fftpack import fft, fftfreq

# Parametrat e sinjalit
Fs = 100  # Frekuenca e mostrimit
T = 1 / Fs  # Intervali i mostrimit
N = 1000  # Numri i mostrave
t = np.linspace(0, N*T, N, endpoint=False)

# Sinjali
f_sin = 10  # Hz
signal = np.sin(2 * np.pi * f_sin * t)

# Llogaritja e FFT-së
F = fft(signal)
freqs = fftfreq(N, T)

# Vizualizimi i sinjalit dhe spektrit
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title("Sinjali në Domenin e Kohës")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")

plt.subplot(1, 2, 2)
plt.plot(freqs[:N//2], np.abs(F[:N//2]))  # Pjesa pozitive e spektrit
plt.title("Spektri i Frekuencës")
plt.xlabel("Frekuenca (Hz)")
plt.ylabel("Magnituda")

plt.tight_layout()
plt.show()
```
**Pritet që spektri të tregojë një vijë të fortë në $10 Hz$.**

## **Ushtrimi 4: Simulimi i Filtrit Kalues të Ulët**
**Qëllimi**: Të zbatohet një **filtrim në domenin e frekuencës** duke përdorur **Transformimin e Fourier-it**.

**Udhëzime**:
1. Gjenero një sinjal me dy komponentë:
   - Njëri me frekuencë **10 Hz**.
   - Tjetri me frekuencë **50 Hz**.
2. Përdor FFT për të analizuar spektrin.
3. Zbato një **filtrim kalues të ulët** për të mbajtur vetëm sinjalin e 10 Hz.

**Kodi Python**:
```python
# Gjenerimi i sinjalit me dy frekuenca
signal = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 50 * t)

# FFT e sinjalit
F = fft(signal)
freqs = fftfreq(N, T)

# Aplikimi i filtrit (heqim frekuencat mbi 20 Hz)
F[np.abs(freqs) > 20] = 0

# Rikthimi i sinjalit me inversin e FFT
filtered_signal = np.real(np.fft.ifft(F))

# Vizualizimi
plt.plot(t, signal, label="Sinjali Origjinal")
plt.plot(t, filtered_signal, label="Sinjali i Filtruar", linestyle="dashed")
plt.title("Filtrimi Kalues i Ulët në Domenin e Frekuencës")
plt.xlabel("Koha (s)")
plt.ylabel("Amplituda")
plt.legend()
plt.show()
```
**Pritet që komponenti 50 Hz të hiqet.**

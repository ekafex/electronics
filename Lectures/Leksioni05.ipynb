# **Fizika e gjysmëpërçuesve**

## **1. Struktura atomike dhe kristalore e gjysmëpërçuesve**

### 1.1 Struktura atomike – Gjysmëpërçuesit si elementë të grupit IV
Gjysmëpërçuesit klasikë si **siliciumi (Si)** dhe **germaniumi (Ge)** ndodhen në **grupin IV të tabelës periodike**. Kjo do të thotë që kanë **katër elektrone në shtresën valencë**.

Në një kuptim mekaniko-kuantik:
- Elektronet rrethojnë bërthamën në **nivele energjie** të lejueshme, që përshkruhen nga zgjidhjet e ekuacionit të Schrödinger për potencialin elektrostatik bërthamë-elektron.
- Në atom të vetëm, këto nivele janë **diskrete** (energjitë janë të kuantizuara).

#### Për shembull për Siliciumin:
- Konfigurimi elektronik është:  
  $$
  \text{Si: } 1s^2 2s^2 2p^6 3s^2 3p^2
  $$
- Katër elektronët në nivelin $n=3$ janë ata që marrin pjesë në lidhje kimike – **elektronët e valencës**.


### 1.2 Lidhja kovalente dhe rrjeti kristalor

Në fazën e ngurtë, atomet e siliciumit organizohen në një **strukturë kristalore me renditje të rregullt**, konkretisht në **strukturën diamante** (e ngjashme me atë të karbonit në diamant).

#### Struktura kristalore:
- Çdo atom është i lidhur me **katër fqinj të afërt** përmes **lidhjeve kovalente**.
- Elektronet ndahen në çift mes atomeve, duke krijuar një **rrjet shumë të qëndrueshëm**.
- Njësia themelore përsëritëse në hapësirë quhet **qeliza elementare** dhe ka simetri **kubike me fytyrë të trupëzuar** (FCC me dy atome për qendër). Shiko strukturen kristaline ne psh [VESTA](https://jp-minerals.org/vesta/en/).

### 1.3 Nga nivele diskrete në breza energjie

Në një **atom të izoluar**, elektronet lejohen vetëm në disa nivele energjie specifike.

Por, në një **kristal me një numër shumë të madh atomesh shumë afër njëri-tjetrit**, orbitalet atomike të jashtme (të valencës) **mbivendosen** dhe ndërveprojnë. Kjo sjell pasoja të mëdha:

#### Interferenca e niveleve:
- Nga mekanika kuantike, kur $N$ atome identike afrohen dhe ndërveprojnë, nivelet diskrete të energjisë **ndarësohen** në $N$ nivele shumë të afërta.
- Këto nivele formojnë një **brez të vazhdueshëm** energjie (quhet **brez energjie**).
  
Për shembull:
- Niveli 3p për çdo atom jep një **brez valencë** me miliona nivele shumë të afërta.
- Po ashtu formohet një **brez përçueshmërie** me nivele më të larta.

### 1.4 Energjia potenciale periodike në kristal

Energjia që ndjen një elektron në një kristal është **periodike**, për shkak të vendosjes së rregullt të bërthamave:
$$ V(x) = V(x + a) $$
ku $a$ është konstanta e rrjetit.

#### Zgjidhja e ekuacionit të Schrödinger:
- Ekuacioni për një elektron në potencial periodik jep **zgjidhje të formës së Bloch**:
$$ \psi_k(x) = u_k(x) e^{ikx} $$
- Kjo strukturë energjie krijon **zona të ndaluara (band gaps)** dhe **zona të lejuara (breza)**.


### Përmbledhje:
- Gjysmëpërçuesit janë materialë ku strukturat atomike (me 4 elektrone valencë) lidhen në rrjete shumë të rregullta.
- Orbitalet kovalente ndërveprojnë dhe formojnë breza energjie.
- Brezat e energjisë përfshijnë: brezin e valencës, brezin e përçueshmërisë, dhe një **hapësirë të ndaluar energjie (band gap)** ndërmjet tyre.
- Kjo hapësirë energjie është çelësi për të kuptuar sjelljen elektrike të materialit.

## **2. Formimi i Brezave të Energjisë dhe Band Gap në Gjysmëpërçues**

### **2.1 Origjina kuantike e brezave të energjisë**

Kur atomet ndodhen në izolim, çdo elektron ndodhet në një nivel energjie të caktuar, të përcaktuar nga ekuacioni i Schrödinger për një atom të vetëm. Por në një **material të ngurtë kristalor**, kur shumë atome janë të afërta dhe formojnë lidhje kovalente, ndodh ndërveprim orbital.

#### Orbitalet e valencës ndërveprojnë:
- Në një kristal me $N \sim 10^{23}$ atome, orbitalet si $3s$, $3p$ (në Si) ndërveprojnë dhe **ndarësohen** në $N$ nivele shumë të afërta.
- Nivelet nuk janë më të dallueshme si të izoluara → formojnë një **brez energjie të vazhdueshëm**.

> *Ky është një rezultat i **parimit të përjashtimit të Pauli**: nuk ka dy elektrone që mund të kenë të njëjtin grup numrash kuantik në një sistem – kështu që secilit elektron i “takon” një nivel pak më ndryshe.*


### **2.2 Brezi i valencës dhe brezi i përçueshmërisë**

Pas formimit të brezave, kemi dy breza të rëndësishëm:

#### Brezi i valencës (*Valence band*):
- Brezi që rezulton nga ndërveprimi i **orbitalëve të valencës** (p.sh. 3s dhe 3p për Si).
- Në temperaturë të ulët, ky brez është **i mbushur plotësisht** me elektrone.

#### Brezi i përçueshmërisë (*Conduction band*):
- Brezi më i lartë energjetik, që rrjedh nga nivele të tjera më të larta (p.sh. 4s ose nivele më të larta të sistemit).
- Në gjysmëpërçues, ky brez është **i zbrazët në $T=0 K$**, por mund të popullohet me elektrone që fitojnë energji.

#### Hapësira ndërmjet dy brezave – **band gap $E_g$**:
- Është energjia minimale që duhet t’i jepet një elektroni për të kaluar nga brezi i valencës në brezin e përçueshmërisë:
$$ E_g = E_c - E_v $$
ku $E_c$ është fundi i brezit të përçueshmërisë dhe $E_v$ është maja e brezit të valencës.


### **2.3 Shembuj konkretë të band gap-it**

| Materiali       | Lloji          | Band Gap (eV) |
|----------------|----------------|----------------|
| Germanium (Ge) | Gjysmëpërçues  | 0.66           |
| Silicium (Si)  | Gjysmëpërçues  | 1.12           |
| Gallium Arsenide (GaAs) | Gjysmëpërçues | 1.42 |
| Karboni (diamant) | Izolator     | ∼5.5           |
| Bakër (Cu)     | Përçues        | 0 (banda mbivendosen) |

#### Vëzhgim:
- Sa më i vogël band gap-i, aq më **lehtë elektronët kalojnë në brez përçueshmërie**.
- Për temperaturat e dhomës (∼300 K), energjia termike është:
  $$ E_{termike} \approx k_B T \approx 0.025 \text{ eV} $$
  Pra vetëm gjysmëpërçuesit me band gap të vogël mund të kenë përçueshmëri për shkak të agjitimit termik.


### **2.4 Roli i band gap-it në përçueshmëri**

#### a) **Përçuesit (metalet)**:
- Brezi i valencës dhe ai i përçueshmërisë **janë të mbivendosur** → nuk ka band gap.
- Elektronet lëvizin lirisht → **përçueshmëri shumë e lartë**.

#### b) **Izolatorët**:
- Band gap shumë i madh (> ∼4-5 eV) → elektronët nuk kalojnë dot.
- **S’ka bartës ngarkese të lirë** → **përçueshmëri ∼0**.

#### c) **Gjysmëpërçuesit**:
- Band gap i mesëm (∼0.6 – 2 eV).
- Në temperaturë dhome, një përqindje e vogël e elektroneve kapërcejnë band gap-in → krijojnë **çifte elektron–vrimë** → bëhen **bartës të përkohshëm**.

Në kristal, këto breza përshkruhen zakonisht si funksion i momentit $k$, përmes **strukturës së brezit** $E(k)$, dhe përshkruhen me formalizma si **modeli i elektronit efektiv**, që mund ta shpjegojmë më vonë në kontekstin e përçueshmërisë.

## **3. Bartësit e ngarkesës dhe dopimi në gjysmëpërçues**

### **3.1 Përçueshmëria në gjysmëpërçues të pastër (intrinsik)**

Në një gjysmëpërçues të pastër, si **silici i pastër**, në temperaturë të ulët (0 K) brezi i përçueshmërisë është i zbrazët, ndërsa brezi i valencës është i mbushur.

Por me **rritjen e temperaturës**, disa elektrone marrin **energji termike** $(k_B T \sim 0.025 \text{ eV}$ në 300 K) dhe **kalojnë në brezin e përçueshmërisë**. Kjo sjell dy pasoja të rëndësishme:

1. **Krijohet një elektron i lirë** në brezin e përçueshmërisë → bartës negativ.
2. **Lë pas një “vrimë”** në brezin e valencës → që sillet si ngarkesë pozitive.

Këto çifte quhen **çifte elektron-vrimë**.
$$ \text{e}^- \quad \leftrightarrow \quad \text{vrimë}^+ $$

#### Kufizimi:
- Numri i bartësve në material intrinsik është relativisht i ulët → përçueshmëri e dobët.


### **3.2 Dopimi – Papastërtia e kontrolluar për të rritur përçueshmërinë**

Për të përmirësuar përçueshmërinë, ne **shtojmë qëllimisht papastërti shumë të vogla (∼1 në 10⁷ atome)** që ndryshojnë ndjeshëm numrin e bartësve të ngarkesës.

Kjo quhet **dopim** dhe materiali bëhet **gjysmëpërçues ekstrinsik**.


### **3.3 Dopimi i tipit $n$ (përdorim i elementeve dhurues-donor)**

- Shtojmë **elementë të grupit V**, si **fosfor (P), arsenik (As)**.
- Ata kanë **5 elektrone valence** → 4 formojnë lidhje si zakonisht, por **1 mbetet i lirë**.
- Ky elektron i tepërt futet shumë pranë brezit të përçueshmërisë (∼0.05 eV) dhe **lehtësisht ionizohet** në temperaturë dhome.

$$ \text{P}^0 \rightarrow \text{P}^+ + e^- $$

- Elektronet bëhen **bartës shumicë**, vrimat janë shumë të pakta → **gjysmëpërçues tip-n**.

**Energjia e jonizimit e vogël** → çdo dopant kontribuon direkt me një elektron të lirë.


### ➖ **3.4 Dopimi i tipit $p$ (përdorim i elementeve pranues-akseptor)**

- Shtojmë **elementë të grupit III**, si **bor (B), alumin (Al)**.
- Ata kanë **vetëm 3 elektrone valencë** → krijojnë **lidhje të paplota**, duke lënë një **vrimë**.

- Vrima sillet si një **nivel bosh** shumë afër majës së brezit të valencës (∼0.05 eV).
$$ \text{B}^0 + e^- \rightarrow \text{B}^- + \text{vrimë}^+ $$

- Vrimat bëhen **bartës shumicë**, elektronet janë pak → **gjysmëpërçues tip-p**.


### **3.5 Koncentrimet dhe përçueshmëria totale**

Përçueshmëria elektrike përshkruhet nga:
$$ \sigma = q (n \mu_n + p \mu_p) $$
ku:
- $n$ dhe $p$ janë përqëndrimet e elektroneve dhe vrimave, respektivisht.
- $\mu_n$, $\mu_p$ janë lëvizshmëritë (mobilitetet) e tyre.
- $q$ ngarkesa elektronit.

Në gjysmëpërçues të dopuar të:
- tipit $n$ → $n \gg p$, përçueshmëria dominohet nga elektronet.
- tipit $p$ → $p \gg n$, përçueshmëria dominohet nga vrimat.


### **3.6 Karakteristikat V-I dhe sjellja elektrike**

#### Për një copë gjysmëpërçues të dopuar:
- Sjellja është **lineare në fillim** si një rezistencë ohmike, por ndryshon në varësi të formimit të bashkimeve p-n.
- Tek diodat p-n, karakteristika V-I është **jo-lineare dhe asimetrike**, do ta shohim në kapitull më vonë.


### Diagrama ilustrative për gjendjen energjetike:

```
n-tip:
Brezi përçues                      ————————
Niveli donor  (~0.05 eV)           - - - - - -
Brezi i valencës                   ————————

p-tip:
Brezi përçues                      ————————
Niveli akceptor  (~0.05 eV)        - - - - - -
Brezi i valencës                   ————————
```

### Përmbledhje

| Tipi           | Dopanti         | Bartës shumicë | Niveli energjie | Përçueshmëri |
|----------------|------------------|----------------|------------------|---------------|
| Intrinsik      | -                | e⁻ dhe vrima    | vetë nga $k_BT$ | e ulët        |
| **tip-n**       | P, As (grupi V)  | elektronë       | pranë $E_c$     | e lartë       |
| **tip-p**       | B, Al (grupi III)| vrima           | pranë $E_v$     | e lartë       |


Në pjesën tjetër do të kalojmë tek:
- **Struktura e bashkimeve p-n**
- **Si lind drejtimi i rrjedhjes së rrymës në dioda**
- **Përdorimi në pajisje si dioda, transistorë, fotodioda, etj.**

## **4. Sjellja elektrike dhe përçueshmëria e gjysmëpërçuesve**

### **4.1 Përçueshmëria në materiale të ndryshme**

#### a) **Përçuesit (metale)**
- Elektronet janë shumë të lirë → lëvizin lehtësisht në fushë elektrike.
- **Rryma është lineare me tensionin** (Ligji i Ohmit).
- **Përçueshmëri e lartë** (∼10⁶ – 10⁸ S/m).

#### b) **Izolatorët**
- Band gap shumë i madh → elektronët nuk kalojnë dot në brezin e përçueshmërisë.
- **Përçueshmëri shumë e ulët** (∼10⁻¹² – 10⁻⁸ S/m).

#### c) **Gjysmëpërçuesit**
- Band gap i vogël → një pjesë e elektroneve termike kalojnë në $E_c$.
- Në silicium në 300 K: përqendrimi i bartësve $n_i \sim 1.5 \times 10^{10} \text{ cm}^{-3}$



### **4.2 Përçueshmëria e gjysmëpërçuesit intrinsik**

Nga statistika Fermi–Dirac dhe modelet e brezave:
$$ \sigma = q \cdot n_i \cdot (\mu_n + \mu_p) $$

ku:
- $n_i$: përqendrimi i bartësve në gjysmëpërçues të pastër
- $\mu_n, \mu_p $: mobiliteti i elektroneve dhe vrimave
- $q$: ngarkesa elementare

#### Shembull:
Për silicium në 300 K:
- $n_i \approx 1.5 \times 10^{10} \text{ cm}^{-3}$
- $\mu_n \approx 1350 \text{ cm}^2/\text{V·s}$
- $\mu_p \approx 480 \text{ cm}^2/\text{V·s}$
- $$   \sigma \approx 1.6 \times 10^{-19} \cdot 1.5 \cdot 10^{10} \cdot (1350 + 480) \approx 2.2 \times 10^{-4} \text{ S/cm}  $$

Pra shumë më pak se një metal, por ende përcjell pak rrymë.


### **4.3 Sjellja V-I e gjysmëpërçuesve të pastër dhe të dopuar**

#### a) **Gjysmëpërçues i pastër (intrinsik):**
- Karakteristikë V-I është **simetrike** dhe jo-lineare për tensione të vogla, sepse përqendrimi i bartësve ndryshon me temperaturën dhe fushën elektrike.
- Kur rritet tensioni, më shumë elektrone kalojnë në $E_c$, duke rritur përçueshmërinë.

#### b) **Gjysmëpërçues të dopuar (ekstrinsik):**
- Përçueshmëria rritet ndjeshëm.
- Pjesa e sipërme e karakteristikës I-V bëhet **më afër lineare**, si një rezistencë më e vogël.
- Në shumë aplikacione (si në dioda), përdorim **bashkime p-n**, ku karakteristika bëhet **jashtëzakonisht jo-lineare dhe e njëanshme**.


### **4.4 Ilustrim grafik i karakteristikës V-I**

#### a) Intrinsik:
$$ I \propto e^{-E_g/2kT} $$
- për tensione të vogla, rryma është shumë e ulët
- rritet eksponencialisht me temperaturën

#### b) tip $n$ dhe $p$:
$$ I = \frac{V}{R_{eff}} \quad \text{(sjellje më ohmike)} $$
ku $R_{eff} \ll R_{intrinsik}$


### **Eksperiment për laborator:**

**Qëllimi:** Krahasoni karakteristikën V-I për:
1. Silici të pastër
2. Silic me dopim donor (tip $n$)
3. Silic me dopim akseptor (tip $p$)

**Pajisjet:**
- Burim tensioni (0–5V)
- Rezistenca të njohura për kufizim rryme
- Ampermetër dhe voltmetër
- Mostra të përgatitura të tre llojeve

**Rezultatet e pritura:**
- I_{n} > I_{p} ≫ I_{intrinsik} për të njëjtin tension

### Tabelë krahasimi

| Materiali         | Përqendrim bartësish | Lëvizshmëri | Sjellje V-I             | Përdorim praktik         |
|------------------|----------------------|-------------|--------------------------|--------------------------|
| Silicium i pastër| shumë e ulët         | e moderuar  | jo-lineare, e dobët     | Sensorë termikë, fotodioda |
| Silicium n-tip   | i lartë (elektronë)  | e lartë     | më afër lineares        | Transistorë, dioda       |
| Silicium p-tip   | i lartë (vrima)      | më e ulët   | më afër lineares        | Dioda, LED, sensora      |


### Përmbledhje:

- Gjysmëpërçuesit e pastër kanë **përçueshmëri të kufizuar** për shkak të band gap-it.
- Me **dopim të lehtë**, përçueshmëria rritet me disa **rendi madhësie**, duke krijuar **materiale me bartës shumicë**.
- Sjellja elektrike ndryshon nga jo-lineare në më afër lineare (ohmike) me dopimin.
- Karakteristika V-I shërben si mënyrë e drejtpërdrejtë për të krahasuar ndikimin e dopimit në përçueshmëri.

## **5. Bashkimi p-n**

### **5.1 Çfarë është një bashkim p-n?**

Kur vendosim një gjysmëpërçues të **tipit $p$** dhe një tjetër të **tipit $n$** pranë njëri-tjetrit, formohet një **bashkim p-n**. Kjo krijon një ndërfaqe/barrierë ku ndodhin disa fenomene shumë interesante kuantike dhe elektrostatike.


### **5.2 Çfarë ndodh në kontaktin p-n?**

#### a) Difuzioni i bartësve:
- Elektronet (nga zona n) **difuzojnë** drejt zonës p, ku ka përqendrim më të ulët elektronësh.
- Vrimat (nga zona p) **difuzojnë** drejt zonës n.

#### b) Rikombinimi:
- Kur një elektron nga zona n takon një vrimë në zonën p, ato **rikombinohen** → nuk kemi më bartës të lirë aty.
- Si rezultat, formohet një **zonë e zbrazët nga bartësit e lirë** pranë ndërfaqes/bashkimit → kjo quhet **zona e zbrazët (depletion region)**.

#### c) Formimi i fushës elektrike të brendshme:
- Rikombinimi lë pas **jone të palëvizshme të dopantëve**:
  - në anën p: jone negative nga dopimi me psh bor
  - në anën n: jone pozitive nga dopimi me psh fosfor

- Kjo krijon një **fushë elektrike të brendshme** e cila kundërshton më tej difuzionin.


### **5.3 Barrierë potenciali – tensioni i ndalimit**

Fusha elektrike në zonën e zbrazët krijon një **barrierë potenciali** që pengon më tej lëvizjen e bartësve të rrymës.
$$ V_0 \approx \frac{kT}{q} \ln\left( \frac{N_a N_d}{n_i^2} \right) $$

- $N_a$, $N_d$: përqendrimet e dopantëve në anën p dhe n (akseptorë dhe donorë, përkatësisht)
- $n_i$: bartësit në materialin intrinsik.

Për silicin, ky tension është zakonisht ∼**0.6–0.7 V**.


## **6. Polarizimi i bashkimit p-n**

### **6.1 Polarizimi përpara (forward bias)**

- Anoda (p) lidhet me +, katoda (n) me -.
- Fusha e jashtme **kundërshton** fushën e brendshme → **zona e zbrazët ngushtohet**.
- Elektronet nga zona n dhe vrimat nga zona p **kalojnë pengesën dhe rikombinohen**.
- Rryma kalon → **dioda përçon**.

#### Karakteristikë V-I:

$$ I \approx I_0 \left( e^{qV/kT} - 1 \right) $$

- Rritje **eksponenciale** e rrymës me tensionin në lidhje të drejtë.


### **6.2 Polarizimi kundërt (reverse bias)**

- Anoda (p) lidhet me -, katoda (n) me +.
- Fusha e jashtme **shton** barrierën → **zona e zbrazët zgjerohet**.
- Bartësit parësorë pengohen, por bartësit e dytësorë intrisik (pak fare) kontribuojnë → rrymë **shumë e vogël** (rryma e kundert $I_0$).


### **6.3 Tensioni i prishjes (breakdown)**

Nëse tensioni i kundërt tejkalon një vlerë kritike:
- Elektronet përshpejtohen aq shumë sa krijojnë **ionizim** → godasin atome dhe çlirojnë elektrone të tjera.
- Ky efekt quhet **prishje me goditje** (avalanche breakdown) ose **tunelim** (Zener breakdown).
- Diodat Zener përdorin këtë efekt për **rregullim të tensionit**.


### Eksperiment praktik me diodë

- Lidh një diodë me burim tensioni dhe rezistencë.
- Mat tensionin dhe rrymën në të dy drejtimet.
- Përdor grafikun për të vizatuar karakteristikën V-I.
- Përdor një LED për të vëzhguar polarizimin përpara (LED ndizet).


### Përmbledhje:

| Polarizimi         | Zona e zbrazët | Rryma | Sjellja            |
|--------------------|----------------|--------|---------------------|
| **Përpara**        | Ngushtohet     | e lartë | Dioda përçon        |
| **Prapë**          | Zgjerohet      | shumë e vogël | Bllokohet rryma     |
| **Prishje**        | Thehet barriera| shumë e lartë | Rrezik ose funksional (Zener) |



````python
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(-1, 1, 400)
q = 1.6e-19  # C
k = 1.38e-23  # J/K
T = 300  # K
n = 1.5
Is = 1e-12

I = Is * (np.exp((q * V) / (n * k * T)) - 1)

plt.figure(figsize=(8, 4))
plt.plot(V, I * 1e6)  # mikroampera
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.title("Karakteristika I-V e një diode gjysmëpërçuese")
plt.xlabel("Tensioni V (V)")
plt.ylabel("Rryma I (μA)")
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

Vin = np.linspace(-1, 1.5, 500)
R = 1000  # Ohm
Vd_th = 0.7
I_out = np.where(Vin > Vd_th, (Vin - Vd_th) / R, 0)

plt.figure(figsize=(8, 4))
plt.plot(Vin, I_out * 1e3)  # mA
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.title("Rryma në një qark me diodë dhe rezistencë")
plt.xlabel("Tensioni hyrës Vin (V)")
plt.ylabel("Rryma I (mA)")
plt.grid(True)
plt.show()
````
